Here lies Cock Robin,  
Dead and cold;  
This book his end  
Will soon unfold.  
这里躺着知更鸟,  
他死了,身体还透着凉;  
这是他的死亡之书,  
现在,即将为你展开.  

![0008](http://www.gutenberg.org/files/22921/22921-h/images/illus-0008-1.jpg)  

---  

Who kill'd Cock Robin?  
I, said the Sparrow,  
With my bow and arrow,  
And I kill'd Cock Robin.  
谁杀死了知更鸟?  
我,麻雀说,  
用我的弓和箭,  
我杀死了知更鸟.  

This is the Sparrow,  
With his bow and arrow.  
这是那只麻雀,  
他正带着他的弓和箭.  

![0009](http://www.gutenberg.org/files/22921/22921-h/images/illus-0009-1.jpg)  

---  

Who saw him die?  
I, said the fly,  
With my little eye,  
And I saw him die.  
谁见证了他的死亡?  
我,苍蝇说,  
用我的小眼睛,  
我见证了他的死亡.  

This is the Fly,  
With his little eye.  
这是那只苍蝇,  
还有他的小眼睛.  

![0012](http://www.gutenberg.org/files/22921/22921-h/images/illus-0012-1.jpg)  

---  

Who caught his blood?  
I, said the Fish,  
With my little dish,  
And I caught his blood.  
谁取走了他的血?  
我,鱼说,  
用我的小碟子,  
我取走了他的血.  

This is the Fish,  
That held the dish.  
这是那条鱼,  
正举着他的小碟子.  

![0013](http://www.gutenberg.org/files/22921/22921-h/images/illus-0013-1.jpg)  

---  

Who'll make his **shroud**?  
I, said the Beetle,  
With my thread and needle,  
And I'll make his shroud.  
谁将为他做**寿衣**?  
我,甲虫说,  
用我的针和线,  
我将为他把寿衣做来.  

This is the Beetle,  
With his thread and needle.  
这是那只甲虫,  
正拿着他的针和线.  

![0016](http://www.gutenberg.org/files/22921/22921-h/images/illus-0016-1.jpg)  

---  

Who'll dig his grave?  
I, said the Owl,  
With my spade and shovel,  
And I'll dig his grave.  
谁将为他掘墓?  
我,猫头鹰说,  
我将用我的铲和锨;  
为他把墓掘来.  

This is the Owl so brave,  
That dug Cock Robin's grave.  
这是勇敢的猫头鹰,  
正在为知更鸟掘墓.  

![0017](http://www.gutenberg.org/files/22921/22921-h/images/illus-0017-1.jpg)  

---  

Who'll be the **Parson**?  
I, said the Rook,  
With my little book,  
And I'll be the Parson.  
谁将来做他的**牧师**?  
我,**乌鸦**说,  
拿着我的小本子,  
我来为他做牧师.  

Here's Parson Rook,  
Reading his book.  
这是乌鸦牧师,  
他正在念着他的小本子.  

![0020](http://www.gutenberg.org/files/22921/22921-h/images/illus-0020-1.jpg)  

---  

Who'll be the Clerk?  
I, said the Lark,  
If 'tis not in the dark,  
And I'll be the Clerk.  
谁将来做簿记员?  
我,云雀说,  
如果不是在黑暗中工作,  
我来当他的簿记员.  

Behold the little Lark,  
Says Amen like a Clerk.  
请看这只小云雀,  
就像教士一样说着阿门.  

![0021](http://www.gutenberg.org/files/22921/22921-h/images/illus-0021-1.jpg)  

---  

Who'll carry him to his grave?  
I, said the Kite,  
If 'tis not in the night,  
I'll carry him to his grave.  
谁将来为他抬棺?  
我,鸢(yuān)说,  
如果不走夜路,  
我来为他抬棺.  

Behold the noble Kite,  
About to take his flight.  
请看这只高贵的鸢,  
他正准备起飞呢.  

![0024](http://www.gutenberg.org/files/22921/22921-h/images/illus-0024-1.jpg)  

---  

Who'll carry the link?  
I, said the **Linnet**,  
Will fetch it in a minute,  
And I'll carry the link.  
谁将来持火把?  
我,**朱顶雀**说,  
我立刻将它取来.  
我来手持火把.  

Here's the Linnet with the light,  
Although it is not night.  
这是举着火把的朱顶雀,  
尽管这还不是晚上.  

![0025](http://www.gutenberg.org/files/22921/22921-h/images/illus-0025-1.jpg)  

---  

Who'll be **chief mourner**?  
I, said the Dove;  
For I mourn for my love;  
And I'll be chief mourner.  
谁是**死者最亲近的人**?  
我,鸽子说,  
我将要哀悼我的挚爱,  
我是他最亲近的人.  

Here's the pretty Dove,  
That mourns for her love.  
这是漂亮的鸽子,  
她将悼念她的挚爱.  

![0028](http://www.gutenberg.org/files/22921/22921-h/images/illus-0028-1.jpg)  

---  

Who'll bear the **pall**?  
We, say the **Wrens**,  
Both the cock and the hen,  
And we'll bear the pall.  
谁来为棺盖上**帷幕**?  
我们,**鹪鹩**(jiāo liáo)们说,  
我们夫妇一起,  
将会为棺盖上帷幕.  

Here are the Wrens so small,  
Who bore Cock Robin's pall.  
这是小鹪鹩们,  
他们正提着帷幕.  

![0029](http://www.gutenberg.org/files/22921/22921-h/images/illus-0029-1.jpg)  

---  

Who'll sing a psalm?  
I, said the **Thrush**,  
As he sat in the bush;  
And I'll sing a psalm.  
谁将唱诵赞美诗?  
我,**画眉**说,  
我将站在灌木丛上,  
为他唱诵赞美诗.  

Here's a fine Thrush,  
Singing psalms in a bush.  
这是好心的画眉,  
正在灌木丛上歌唱.  

![0032](http://www.gutenberg.org/files/22921/22921-h/images/illus-0032-1.jpg)  

---  

Who'll toll the bell?  
I, said the Bull,  
Because I can pull;  
So Cock Robin, farewell.  
谁将敲响丧钟?  
我,老牛说,  
因为我有力气,  
我来敲响丧钟.  

Here is the Bull,  
Who said he could pull.  
这是老牛,  
他说他可以拉响丧钟.  

![0033](http://www.gutenberg.org/files/22921/22921-h/images/illus-0033-1.jpg)  

---  

All the birds in the air  
Fell to sighing and sobbing,  
When they heard the bell toll  
For poor Cock Robin.  
所有的鸟儿都停在空中,  
全都为他的死叹息哭泣,  
当他们听到丧钟声响起,  
一起铭记可怜的知更鸟.
