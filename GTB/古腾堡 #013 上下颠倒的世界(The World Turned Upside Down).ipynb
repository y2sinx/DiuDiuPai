# The World Turned Upside Down,    
Or,    
No News, And Strange News.    

Adorned With Cuts    

York:    
Printed By J. Kendrew, Colliergate.    
***    

**Frontispiece**:    
![上下颠倒的世界](https://gutenberg.org/files/10965/10965-h/ud-front.png)    

Here you may see what's very rare,    
The world turn'd upside down;    
A tree and castle in the air,    
A man walk on his crown.

[-3-]()    
***    

To see a butcher kill a hog, is no news;    
But to see a hare run after a dog, is strange indeed!    
![野兔追狗](https://gutenberg.org/files/10965/10965-h/ud-06ill.png)    

This hare hunts the dog,    
Tho' all of you know,    
Most dogs hunt the hare—    
But here it's not so.    

[-6-]()    
***    

To see a poor man and a rich, is no news;    
But to see the devil **hugging/抱** a witch, is strange indeed!    
![恶魔抱女巫](https://gutenberg.org/files/10965/10965-h/ud-07ill.png)    

Is your name Nick, sir,    
Or Old Harry,    
I insist you tell before    
We marry.    

[-7-]()    
***    

To see a cat catching a mouse, is no news;    
But to see a rat building a house, is strange indeed!    
![老鼠建房子](https://gutenberg.org/files/10965/10965-h/ud-08ill.png)    

Some rats take delight to gnaw    
Houses down—    
I want to build a good    
House of my own.    

[-8-]()    
***    

To see a bird picking at fruit, is no news;    
But to see a dog playing the flute, is strange indeed!    
![狗吹笛子](https://gutenberg.org/files/10965/10965-h/ud-09ill.png)    

You see I am playing here,    
Too, too, too, too;    
When I've done with my flute,    
I'll give it to you.    

[-9-]()    
***    

To see a **greyhound/灰狗** catch a hare, is no news;    
But to see a lamb hunting a bear, is strange indeed!    
![小羊猎熊](https://gutenberg.org/files/10965/10965-h/ud-10ill.png)    

The bear runs away, the    
Lamb is pursuing;    
If he catches the bear,    
There'll be terrible doing.    

[-10-]()    
***    

To see an eagle spread her wings, is no news;    
But to see an old man in leading strings, is strange indeed!    
![老人引绳](https://gutenberg.org/files/10965/10965-h/ud-11ill.png)    

Take care, grandpapa, **lest/以免**    
You should fall,    
And if you want your    
Chair—pray call.    

[-11-]()    
***    

To see a weaver throw his shuttle, is no news;    
But to see a man get in a quart bottle, is strange indeed!    
![男人坐在夸脱瓶](https://gutenberg.org/files/10965/10965-h/ud-12ill.png)    

You'll all be amazed to    
See me get in,    
And much more surprised    
If I get out again.    

[-12-]()    
***    

To hear of a Frenchman eating a frog, is no news;    
But to see a butcher stuck by a hog, is strange indeed!    
![猪捅屠夫](https://gutenberg.org/files/10965/10965-h/ud-13ill.png)    

Don't hollo!    
'Tis only a knife,    
You never stuck a pig    
So easy in your life.    

[-13-]()    
***    

To see a gardener gather a salad, is no news;    
But to see an ass singing a **ballad/民谣**, is strange indeed!    
![驴子唱民谣](https://gutenberg.org/files/10965/10965-h/ud-14ill.png)    

My voice is so fine,    
And so charming the tune,    
I expect for to sing at    
The opera soon.    

[-14-]()    
***    

To see a man get a boat, is no news;    
But to see a man jump down his throat, is strange indeed!    
![想要跳进自己喉咙的男人](https://gutenberg.org/files/10965/10965-h/ud-15ill.png)    

If I once get my legs in    
As far as my knees,    
The rest will slip down    
With a great deal of ease.    

[-15-]()    
***    

To see a dog **baiting/诱惑** a bull, is no news;    
But to see a ram spinning of wool, is strange indeed!    
![公羊纺羊毛](https://gutenberg.org/files/10965/10965-h/ud-16ill.png)    

Good lack a daisy!    
How you grin,    
To see a poor ram    
Try to spin.    

[-16-]()    
***    

To see a tailor making a coat, is no news;    
But to see a hog rowing a boat, is strange indeed!    
![猪划船](https://gutenberg.org/files/10965/10965-h/ud-17ill.png)    

Step into my boat,    
I'll give you a row,    
I'm an excellent waterman,    
All of you know.    

[-17-]()    
***    

To see a man fish in a **moat/护城河**, is no news;    
But to see a monkey shaving a goat, is strange indeed!    
![猴子给山羊剃毛](https://gutenberg.org/files/10965/10965-h/ud-18ill.png)    

My **lather/肥皂泡** is good, and    
My razor is keen,    
In two minutes time,    
I'll shave you quite clean.    

[-18-]()    
***    

To see a miller **grinding/磨** corn, is no news;    
But to see an ox blowing a horn, is strange indeed!    
![公牛吹喇叭](https://gutenberg.org/files/10965/10965-h/ud-19ill.png)    

I blow and I run,    
And I run and I blow,    
And which I do best,    
I'm sure I don't know.    

[-19-]()    
***    

To see a shoemaker hammer his leather, is no news;    
But to see a hound and a buck drinking together, is strange indeed!    
![猎狗和公鹿在喝酒](https://gutenberg.org/files/10965/10965-h/ud-20ill.png)    

Mr. Buck, I wish you luck.    
Mr. Hound, your most profound.    

[-20-]()    
***    

To see a good boy read his book, is no news;    
But to see a goose roasting a cook, is strange indeed!    
![大鹅烤厨子](https://gutenberg.org/files/10965/10965-h/ud-21ill.png)    

I'll roast **ye/你**, and **baste/浇卤汁于** ye,    
But who will may taste ye.    

[-21-]()    
***    

To see a beau at his toilet dress, is no news;    
But to see two horses playing at chess, is strange indeed!    
![两只马在下棋](https://gutenberg.org/files/10965/10965-h/ud-22ill.png)    

Mr. Horse, you don't play    
Fair, but cheat:    
Mr. Nag, you say so,    
Now you're beat.    

[-22-]()    
***    

To see a **high-bred/纯种的** horse when **prancing/以后脚腾跃**, is no news;    
But to see a cat **fiddling/拉小提琴** and mice all dancing, is strange indeed!    
![猫拉小提琴,老鼠跳舞](https://gutenberg.org/files/10965/10965-h/ud-23ill.png)    

Come foot it, my dears,    
And when you've done,    
I'll eat you for supper,    
Aye, every one.    

[-23-]()    
***    

To see a lady drinking of tea, is no news;    
But to see a bird shoot a man in a tree, is strange indeed!    
![鸟射人](https://gutenberg.org/files/10965/10965-h/ud-24ill.png)    

First of all, I'll shoot this clown,    
And then I'll bring his bow wow down.    

[-24-]()    
***    

To see a boy swim in a brook, is no news;    
But to see a fish catch a man with hook, is strange indeed!    
![鱼钓人](https://gutenberg.org/files/10965/10965-h/ud-25ill.png)    

Spare me, good Mr. Fish,    
I didn't **molest/骚扰** you.    
I'll spare you no longer    
Than till I dress you.    

[-25-]()    
***    

To see a cat steal milk from a pan, is no news;    
But to see a buck hunting a man, is strange indeed!    
![鹿猎人](https://gutenberg.org/files/10965/10965-h/ud-26ill.png)    

Hark forward, **huzza/欢呼,喝彩**,    
He can't get away:    
And he's not buck enough    
To stand at a bay.    

[-26-]()    
***    

To see a **cobbler/补鞋匠** mending a shoe, is no news;    
But to see a goat cry old clothes like a **jew/珠宝**, is strange indeed!    
![山羊心疼旧衣服](https://gutenberg.org/files/10965/10965-h/ud-27ill.png)    

'Tis alvays my vay,    
To cheat ven I can,    
Yet for all that I be    
A very goot man.    

[-27-]()    
***    

To see a haymaker using of **rakes/耙子**, is no news;    
But to see a bear making plum cakes, is strange indeed!    
![熊做糕点](https://gutenberg.org/files/10965/10965-h/ud-28ill.png)    

What I have got in the pan,    
I shall eat if I can,    
And this cake on the shelf,    
I shall keep for myself.    

[-28-]()    
***    

To see a barrel made by a **cooper/制桶工**, is no news,    
But to see a goat act the part of a **trooper/骑兵**, is strange indeed!    
![山羊当骑兵](https://gutenberg.org/files/10965/10965-h/ud-29ill.png)    

Hur will fight hur foes    
For the honour of Wales,    
And if the French come,    
They shall **turn their tails/逃跑**.    

[-29-]()    
***    

To see **wrestlers/摔跤的人** kicking **shins/小腿**, is no news;    
But to see cats playing at **nine pins/一种类似于保龄球的游戏**, is strange indeed!    
![猫玩保龄球](https://gutenberg.org/files/10965/10965-h/ud-30ill.png)    

I'll lay a penny, you don't get many.    
I'll knock down all, and swallow the ball.    

[-30-]()    
***    

To hear a parrot say, pretty Poll, is no news;    
But to see a sow with a parasol, is strange indeed!    
![母猪撑遮阳伞](https://gutenberg.org/files/10965/10965-h/ud-31ill.png)    

Like a lady I shine,    
I'm so fat and so fine;    
I've a right I suppose,    
To a shade for my nose.    

[-31-]()    
***    

![猴子骑熊](https://gutenberg.org/files/10965/10965-h/ud-32ill.png)

Little Folks may be supplied with a great variety of Little Books.
