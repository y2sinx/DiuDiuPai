Their only son Fukutaro(F) also was of a gentle disposition. When he was old enough to take a wife, his parents concluded to get him one, transfer their property to him, and seek retirement. Fortunately, one of their relatives named Chudayu(C) had a lovely daughter called Hatsuka. Accordingly a **go-between** was employed to enter into negotiations    
和 O 都非常的**谨慎**. 他们从不在白天(甚至是夜晚)去**客厅**或者厨房, 因此他们过着**平静的**远离猫猫的危险生活. 他们唯一的儿子 F 亦有一个平和的性情. 当他到了该娶妻的年龄时,他的父母决定给他找一位,并决定把自己的财富转交给他,然后退休. 幸运的是,他们有一位亲戚名叫 C, C 有一位可爱的女儿名叫 H. K 雇了一位**媒人**

他们选了一个良辰吉日,并做好所有有关把新娘送到新郎家前的准备. 新娘的衣服需要裁剪,她需要的物品需要购买,新娘的母亲忙的的一塌糊涂.

The parents made their daughter Hatsuka blacken her teeth as a sign that she would not marry a second husband; they also carefully taught her that she must obey her husband, be dutiful to her father-in-law, and love her mother-in-law.    
新娘的父母把新娘的牙齿涂黑,表示她此生不侍二夫; 并认真的教她要顺从自己的丈夫,顺从自己的公婆.    

Kanemochi on his part cleaned up his house inside and out, made preparation for the marriage ceremony and feast, assembled his relatives and friends, and sent out many of his servants to meet the bride on her way, and to give notice of her approach, that all might be prepared for her reception.    
K 则把家的里里外外打扫的干干净净,并准备婚礼和盛宴,他召集了他的亲朋好友,并吩咐他的仆人去新娘那儿打探,并随时通知新娘准备到哪了,所有的一切都要为接待新娘做好准备.

Soon the bride came in her **palanquin** with her boxes carried before her, and a long train of attendants following her. Kanemochi went out as far as the gate to meet her, and **ushered** her into the parlor.    
不久之后,新娘乘着**轿子**来了,她的箱子架在她的前面, 而她的后面跟着一队侍从. K 立马跑到了门口接她,并把她**领进了**客厅.    

At a signal from the go-between the bride and bridegroom, to confirm the marriage bond, exchanged between themselves three cups of sake, drinking three times from each cup in turns.    
媒人示意后,新郎和新娘便结为了连理,然后喝起了交杯酒.
