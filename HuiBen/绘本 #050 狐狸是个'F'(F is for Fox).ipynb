### 狐狸是个'F'    

马克·帕滕登    

![sm.png](https://s.pc.qq.com/tousu/img/20210304/2423656_1614801615.jpg)    

---

世界上有许多种不同的狐狸(foxes), 如: 红狐(red fox), 大耳狐(fennec fox), 蝠耳狐(bat-eared fox), 北极狐(Arctic fox), 灰狐(grey fox), 迅敏狐(swift fox).    

![02-03](https://s.pc.qq.com/tousu/img/20210304/1178948_1614801331.jpg)    

---

一些狐狸生活在森林里的洞穴(dens)中.    

![04](https://s.pc.qq.com/tousu/img/20210304/6677644_1614801333.jpg)    

---

厚厚的毛皮(thick fur)可以让北极狐在雪中保持温暖.    

![05](https://s.pc.qq.com/tousu/img/20210304/1442838_1614801324.jpg)    

---

狐狸习惯于在夜间(nighttime)猎食(hunt for food).    

![06-07](https://s.pc.qq.com/tousu/img/20210304/1178474_1614801336.jpg)    

---

狐狸有四只覆盖着毛皮的脚(furry feet).    

![08](https://s.pc.qq.com/tousu/img/20210304/2332472_1614801325.jpg)    

---

每只狐狸都有一条蓬松多毛的(fluffy,bushy)的尾巴, 叫做狐尾(brush).    

![09](https://s.pc.qq.com/tousu/img/20210304/8928863_1614801325.jpg)    

---

狐狸幼崽(cubs)与他们的兄弟姐妹们一起玩耍.    

![10-11](https://s.pc.qq.com/tousu/img/20210304/5101720_1614801335.jpg)    

---

狐狸喜欢吃多汁的(juicy)的浆果和其他的水果.    

![12-13](https://s.pc.qq.com/tousu/img/20210304/5114332_1614801346.jpg)    

---

狐狸可以生活在很多其他的地方.     

![14-15](https://s.pc.qq.com/tousu/img/20210304/7937586_1614801334.jpg)    

---

狐狸真是一种了不起(fantastic)的动物!    

![16](https://s.pc.qq.com/tousu/img/20210304/7669355_1614801324.jpg)
