###  爱上野马的女孩    

保罗·戈布尔(述+绘)    

Bradbury Press    

![04](https://ae04.alicdn.com/kf/Hceedbe0b22634209b566b792b0138246x.jpg)    

---

献给爱马的邦妮(Bonnie)和珍妮特(Janet).    

![05](https://sc04.alicdn.com/kf/H1376829da3d240978e574ffe1c5350eaN.jpg)    

---

印第安人(the people)总是跟着野牛群(herds of buffalo)从一个地方搬到另一个地方去(moving from place to place). 他们养了很多的马(had many horses)用以搬帐篷(tipis)和其他的东西(belonging). 他们还会训练跑的最快的马去猎野牛.    

![06](https://sc01.alicdn.com/kf/H951fd99c1b66410d90ece08b02e4c906v.jpg)    

---

村子里有一个爱马的女孩. 她总是在破晓时分(daybreak)就起床, 那时的鸟儿还在为渐渐升起的太阳而歌唱. 她让马儿去河边饮水. 她低声喃喃着, 马儿就会跟着她.    

人们注意到她对马儿有一种特殊的理解(understand)方式. 她知道马儿最喜欢什么样的草, 她知道去哪寻找躲避冬天暴风雪(winter blizzards)的避难所(shelter). 如果有马儿受了伤, 她会亲自照顾他们.    

![07](https://ae03.alicdn.com/kf/H9708934a698246eea99256759ba0e6ec8.jpg)    

---

每天, 当她帮母亲打完水(carry water)和收集完木柴(firewood), 她总是跑去跟马儿待在一起(be with the horses). 她在牧场上(meadows)细心地看着马儿, 不会让他们跑的离家太远(beyond sight of home).    

在一天炎热的正午(sun was overhead), 她感觉非常地困(sleepy). 于是她铺了一张毯子(spread a blanket)并躺了下来. 听着马儿在花丛间吃草并缓缓地移动, 她感觉非常的美好. 不久她就睡着了.    

远雷(distant thunder)微弱的隆隆响(faint rumble)并没有把她惊醒. 愤怒的乌云携带着闪电(lightning flashing)在天空上翻滚. 凉风(fresh breeze)和雨息(scent of rain)使她睡得更香了(sleep soundly).    

![08](https://ae03.alicdn.com/kf/Hbb619fe12f8040178a50fb840697f8e67.jpg)    

---

突然间, 一道闪电(a flash of lighting)坠落(crash), 撞击着大地发出了隆隆声. 女孩在惊吓中(in fright)跳了起来(leapt to her feet). 一切都被惊醒了. 在这惊恐中, 马儿踢蹬着后腿(hind legs)暴跳了起来(rearing up), 还在不停地哼着鼻子(snorting). 她迅速的抓住了(grabbed)一匹马的鬃毛(mane)并跳上了他的背.    

![09](https://ae02.alicdn.com/kf/H6e9c9689eca147b3836628bf788813c4u.jpg)    

---

一瞬间(in an instant), 马群像风一样飞奔而去(galloping away like the wind). 她向马群呼喊, 让他们停下来, 但是她的声音被雷声淹没了. 什么都无法让马群停下来. 她抱着马的脖子, 手指抓着马的鬃毛(twisted into his mane). 她抓的很紧很紧(clung on), 生怕跌入奔腾的马蹄之中(drumming hooves).    

![10](https://ae04.alicdn.com/kf/Hfc867615270846f6bfcad9fa659febf8j.jpg)    

---

马儿们被雷声和闪电驱使着(pursued by), 跑得越来越快. 他们就像一场棕色的洪水迅速地横扫着(swept)山谷. 恐惧驱使着他们不断的前行, 他们熟悉的牧场(grazing grounds)被远远地抛在了身后.    

最后, 暴风雨消失在了地平线. 疲倦的马儿们渐渐停了下来休息. 星星出现了, 月亮也在照耀着山丘, 女孩从未见过这样的场景. 她知道他们迷路了.    

![11](https://ae04.alicdn.com/kf/H7e96d30efbb24deb9162cd3d1d5c18b09.jpg)    

---

第二天早上, 她被一阵吵闹的马嘶声(neighing)吵醒了. 一个漂亮的斑点公马(stallion)在她面前来回的欢腾(prancing to and fro), 跺着他的脚(stamping his hooves), 摇晃他的鬃毛. 他比任何她曾经想象的马儿还要强壮, 自豪(proud), 以及英俊(handsome). 他告诉她, 他是漫步(roam)在这些山丘的所有野马的首领(leader). 他欢迎她与他们一起生活. 她很高兴, 而她所有的马儿也很高兴, 他们抬起了头, 高兴的嘶鸣着(neighed joyfully), 憧憬着与野马一样的自由生活.    

![12](https://ae04.alicdn.com/kf/Hffd9ceb9f3ac45eba5093e9c5867213aV.jpg)    

---

女孩的同胞们为搜寻她与消失的马儿找遍了所有的地方. 但是依然找不到他们.    

但是一年之后, 两名猎手骑进了野马生活的山丘. 当他们爬上了一个山丘并看向其他的山丘时, 他们看到了被野马首领引领的野马群. 在野马首领旁, 女孩正在引领一只小马(colt). 他们向她大喊着. 她也向他们挥了挥手, 但是野马首领却迅速驱使着她与所有的马儿走了.    

![13](https://ae03.alicdn.com/kf/H8a75f20f96d54ccb82e5df0ee78a6848d.jpg)    

---

猎手们骑着马飞奔(galloped)回了家, 并把他们所看到的场景告诉了大家. 于是, 男人们便骑上了他们最快的马儿, 又立刻出发去寻找女孩了.    

这是一场耗时的的追逐(long chase). 野马首领保卫着女孩和小马. 他围着他们一圈一圈的转, 以至于骑手们(riders)都无法靠近. 他们想要用绳索抓住他, 但是都被他避开了(dodged). 他的心中没有丝毫恐惧. 他眼如冷星(shone like cold stars), 鼻喷怒气, 蹄如闪电.    

骑手们都钦佩他的勇气(admire his courage). 他们或许永远也无法抓到女孩, 如果不是她骑的马被绊倒(stumbled)后把她摔了下来.    

![14](https://sc01.alicdn.com/kf/H13446d7c4c644f16af02d50bba40fdd6V.jpg)    

---

能重新见到她的父母她很高兴, 她的父母也认为她应该会很幸福. 但是不久之后, 他们发现她很伤心, 她非常想念小马和野马们.    

太阳落山之后的每天晚上, 人们都能听到野马首领在村庄附近的山丘上悲伤的嘶鸣, 呼唤着她回去.    

日子就这样过去了. 她的父母知道女孩很孤单. 她生病了, 医生也无能为力. 他们问她怎样才能让她重新好起来. "我喜欢与野马们一起奔跑," 她回答. "他们是我的同伴(relative). 如果你让我回到他们身边, 我今后会永远(evermore)开心的."    

![15](https://ae03.alicdn.com/kf/Hc91136a0680845268c03d971468c903cq.jpg)    

---

她的父母很爱她, 于是便同意了她与野马一起生活. 他们给了她一件漂亮的衣服, 并把村子里最好的马给了她.    

野马首领让他的野马下了山丘. 人们给他们披上了些好东西: 多彩的毯子以及带装饰的马鞍(saddles). 人们还在他们身上涂了图案, 在他们的鬃毛和尾巴上系了鹰羽以及彩带.    

作为回报(in return), 女孩把小马送给了她的父母. 所有人都很开心.    

![16](https://ae02.alicdn.com/kf/H8879e5d410ae4b858cc3fab7f183f943z.jpg)    

---

女孩又一次骑在了野马首领的旁边. 重新在一起的他们很自豪很幸福.    

但是她并没有忘记她的同胞. 每一年她都会回来, 而她也总是会为她的父母送来一只小马.    

![17](https://ae04.alicdn.com/kf/H2c0df17e307c45c485dadffb68b59050P.jpg)    

---

直到有一年, 她不再回来啦, 然后再也没有回来过. 但是当猎人们再一次见到野马群时, 他们看到, 在强壮的野马首领旁多了一位漂亮的母马(mare), 她的鬃毛和尾巴像纤细的云彩(wispy clouds)一样. 他们都说, 女孩最后一定是变成了一匹野马.    

如今, 我们仍然高兴的铭记(remember)我们与"马族人"之间的联系(relatives). 看到野马自由自在的驰骋, 这会给我们带来欢乐. 我们的思绪会与他们一起飞翔.
