### 我的狗狗布迪    

嘉伯莉·夏博内(述)  皮娅·林登鲍(画)    

亨利霍特出版公司    

![02.jpg](https://ae02.alicdn.com/kf/Hbf525015dda04d3c9dd2fa84b60c1cf1g.jpg)    

---

这是我的狗狗布迪, 她霸占了爸爸专用的沙发椅.    

![03](https://s3.ax1x.com/2020/12/10/rPyCnS.jpg)    

---

布迪是一只牛头梗(bullterrier), 她忠于自己的守卫(guard)工作, 她是世界上最棒的(brilliant)狗狗.    

![4](https://s3.ax1x.com/2020/12/10/rPyP0g.jpg)    

---

牛头梗既强壮勇敢, 又非常的凶猛(fierce), 窃贼(burglar)光是看到她就会两腿发软落荒而逃.    

![5](https://s3.ax1x.com/2020/12/10/rPypX8.jpg)    

---

遛狗是我放学后的任务, 但是她老是不下楼, 爸爸说她有点古怪(nutty), 但是我觉得她很本分.    

![6](https://s3.ax1x.com/2020/12/10/rPyAts.jpg)    

---

布迪喜欢与我 "背道而驰", 而我也只好让她抉择, 不过我们最终总是会在公园里结束旅程.    

![7](https://s3.ax1x.com/2020/12/10/rPykkj.jpg)    

---

公园里的其他狗狗都喜欢追球, 但是布迪更喜欢虫子和树叶.    

![8](https://s3.ax1x.com/2020/12/10/rPyZpq.jpg)    

---

布迪坐在石头上的样子好美啊, 活像一尊高贵的雕塑(royal statue). 你可以看到她机灵的小耳朵和小爪子, 还有她鼻子上的软毛. 她是我的狗狗, 我很骄傲.    

![9](https://s3.ax1x.com/2020/12/10/rPyEhn.jpg)    

---

有些时候布迪会自己出去冒险, 就在上周, 我找了她两个多小时.    

![10](https://s3.ax1x.com/2020/12/10/rPynXT.jpg)    

---

当布迪在杂货店(grocery store)门口等我的时候, 她喜欢观察来往的人与狗.    

![11](https://s3.ax1x.com/2020/12/10/rPymcV.jpg)    

---

小狗会让人她感到无聊,    

![12](https://s3.ax1x.com/2020/12/10/rPye10.jpg)    

---

但是她喜欢大狗.    

![13](https://s3.ax1x.com/2020/12/10/rPyMBF.jpg)    

---

她不喜欢被困在雨中(get caught in the rain), 她尤其是讨厌那些水坑(puddle), 因为她不想弄湿自己的脚脚.    

![14](https://s3.ax1x.com/2020/12/10/rPyKnU.jpg)    

---

如果可以, 她会绕着水坑走.    

![15](https://s3.ax1x.com/2020/12/10/rPyQ74.jpg)    

---

回到家后我想给她擦擦, 但是无论我怎么喊(yell)她, 她就是一动不动(budge).    

![16](https://s3.ax1x.com/2020/12/10/rPy8hR.jpg)    

---

布迪会在一个特定的时间吃饭, 而她只吃自己喜欢吃的东西.    

![17](https://s3.ax1x.com/2020/12/10/rPy1AJ.jpg)    

---

布迪从来不喜欢吸尘器(vacuum cleaner), 但是她的自控力(self    -control)让她不至于把它撕成碎片(ripping sth to pieces ).    

![18](https://s3.ax1x.com/2020/12/10/rPyJ91.jpg)    

---

在晚上的时候, 布迪总是很活泼. 为了看她的表演, 我总是匆匆地就洗完澡.    

![19](https://s3.ax1x.com/2020/12/10/rPy3N9.jpg)    

---

当布迪想要玩耍的时候, 她好可爱啊! 妈妈很好奇她为什么不在白天的时候出去玩, 我觉得嘛, 她应该是一个夜猫子(night person).    

![20](https://s3.ax1x.com/2020/12/10/rPyY1x.jpg)    

---

弟弟有时会被布迪绊倒, 妈妈因此很心烦(upset).    

![21](https://s3.ax1x.com/2020/12/10/rPyNjK.jpg)    

---

就像一个加农炮越过房间, 布迪总是能 "惊艳" 到我...    

![22](https://s3.ax1x.com/2020/12/10/rPytc6.jpg)    

---

哈哈, 布迪在沙发上跳傻狗舞呢! 我让爸爸赶紧拿摄像机来, 但是爸爸不理我, 而且楼上的邻居也总是抱怨布迪弄出来的声音. 哎, 为什么我周围的人都是些扫兴鬼(party pooper)啊!    

![23](https://s3.ax1x.com/2020/12/10/rPyanO.jpg)    

---

最终, 布迪玩够了, 要准备上 "床" 睡觉了...    

![24](https://s3.ax1x.com/2020/12/10/rPyBAH.jpg)    

---

我拍了拍她, 然后安然的进入了梦乡. 因为我知道, 有她的陪伴我很安全.    

![25](https://s3.ax1x.com/2020/12/10/rPywHe.jpg)    

---

像布迪这样鬼精灵(quirk)的狗狗总是让人无法抗拒(irresistible).    

当我准备合上这本书的时候, 我三岁的 "儿子" 立马跳了起来, 并求我又给他读了一遍.    

![27](https://s3.ax1x.com/2020/12/10/rPyDNd.jpg)
