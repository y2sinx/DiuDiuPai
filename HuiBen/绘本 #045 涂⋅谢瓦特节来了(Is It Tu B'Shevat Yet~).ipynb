### 涂⋅谢瓦特节来了    

Chris Barash | Alessandra Psacharopulo    

![sm](https://ae03.alicdn.com/kf/H33e5e468131d462ea160589c3f7e234eO.jpg)    

---

When we look out the window at trees dressed in browns(身披着棕色)    

![2-3](https://s.pc.qq.com/tousu/img/20210213/9193423_1613151640.jpg)    

---

and smile to imagine their leafy(叶茂的) green gowns(长袍)...    

Tu B'Shevat(涂⋅谢瓦特) is on its way(在来的路上).    

![4-5](https://s.pc.qq.com/tousu/img/20210213/3828436_1613151650.jpg)    

---

When our cousins in Israel(以色列) send pictures of bees    

![6-7](https://s.pc.qq.com/tousu/img/20210213/9020629_1613151657.jpg)    

---

as they buzz around blooms(盛开的花卉) on the pink almond(杏树) trees...    

Tu B'Shevat is on its way.    

![8-9](https://s.pc.qq.com/tousu/img/20210213/9696787_1613151639.jpg)    

---

When Gran(奶奶) hurries in from her trip to the store,    

![10-11](https://s.pc.qq.com/tousu/img/20210213/2526843_1613151642.jpg)    

---

with dates(枣子), pomegranates(石榴), and olives(橄榄) galore(大量地)...    

Tu B'Shevat is on its way.    

![12-13](https://s.pc.qq.com/tousu/img/20210213/1598687_1613151650.jpg)    

---

When Dad helps us make trees(树的造型), so crunchy(易碎的) and sweet,    

![14-15](https://s.pc.qq.com/tousu/img/20210213/6698250_1613151675.jpg)    

---

from all kinds of nuts, fruits, and seeds that we'll eat...    

Tu B'Shevat is on its way.    

![16-17](https://ae04.alicdn.com/kf/Hd65a3ae01abd424780e24ab3829f3a83Q.jpg)    

---

When we work at the park and help spread the news that we need to recycle(回收利用), reduce(减少能耗), and reuse(循环再用)...    

Tu B'Shevat is on its way.    

![18-19](https://sc02.alicdn.com/kf/H37d270000e094a6f9fe53ef690c8b9972.jpg)    

---

When Mom's back from the nursery(苗圃) with seedlings(秧苗) in tow(牵引车) and says we'll soon plant them and help them to grow...    

Tu B'Shevat is on its way.    

![20-21](https://s.pc.qq.com/tousu/img/20210213/9911802_1613151732.jpg)    

---

When our spades(铁锹) dig deep hollows as friends gather 'round(around),    

![22-23](https://ae02.alicdn.com/kf/H219029d63f994a7189dec7453bbb73aan.jpg)    

---

and everyone helps place our trees in the ground...    

![24-25](https://ae04.alicdn.com/kf/Hc49967123756479a9e4a2b4c4cd37e215.jpg)

When we reach out our hands as we happily say,

---

"Thank you, trees, for all that you give us each day!"    

![28-29](https://s.pc.qq.com/tousu/img/20210213/5958537_1613151790.jpg)
