### 孩子们的第一本圣经故事    

萨莉·劳埃德-琼斯    

[sm]()

---

天地初始, 世界一片荒芜(dark,still,empty). 随后, 上帝让世界充满了光(light), 随之而来的还有群山,河流,天空,海洋,以及动植物. 此外上帝还创造一男一女(made a man and a woman), 名叫亚当和夏娃(Adam&Eve).      

![6.png](https://s.pc.qq.com/tousu/img/20210207/8916365_1612676839.jpg)    

---

上帝为了惩罚变坏的(mean)人类, 降下了大洪水, 但是他提前让诺亚造了一艘大船(big boat), 因此拯救了所有的动物.    

![7.png](https://s.pc.qq.com/tousu/img/20210207/4674076_1612676867.jpg)    

---

上帝答应给亚伯拉罕和他的妻萨拉(Abraham&Sarah)数也数不尽的子孙, 然后这对老夫妻便有了儿子以撒(Isaac).    

![8.png](https://s.pc.qq.com/tousu/img/20210207/5864409_1612676860.jpg)    

---

巨人(Giant)哥利亚(Goliath)想要找人打架, 但是大家都很害怕, 最后一个叫大卫(David)的小男孩勇敢的站了出来, 因为他知道上帝是站在他这一边的(with God on his side).     

![13.png](https://s.pc.qq.com/tousu/img/20210207/9927191_1612676839.jpg)    

---

上帝的忠实粉丝丹尼尔(Daniel)成了国王的顶级助手(top helper), 但是那些心生妒忌(jealous)人却让国王立了一条不能向上帝祈祷的法律(make a law), 还把丹尼尔扔进关狮子的牢房(threw Daniel into the lions' den).     

![14.png](https://s.pc.qq.com/tousu/img/20210207/2264584_1612676873.jpg)    

---

在伯利恒(Bethlehem)的田野上, 天使告诉了牧羊人(shepherds)一个好消息: 上帝的儿子今夜会在马厩(stable)的马槽(manger)里诞生, 他将改变这个世界.    

![17.png](https://s.pc.qq.com/tousu/img/20210207/4302358_1612676842.jpg)    

---

捕鱼人彼得(Peter)捕了一晚上鱼却一无所获, 但是当耶稣(Jesus)说让他去捕鱼的时候, 他却捕了超多的鱼. 于是, 当耶稣说让彼得跟随他的脚步的时候, 彼得便跟随了耶稣的脚步.    

![20.png](https://ae04.alicdn.com/kf/H0221b6929a554d23b45bbd1a78de3af4R.jpg)    

---

有一天, 彼得跟他的朋友们在湖上划船时遇到了暴风雨, 他们看到了耶稣正在湖面上行走. 彼得问耶稣, 他可不可以也在湖面上走啊, 耶稣说当然可以.     

![21.png](https://ae03.alicdn.com/kf/H19decc627f0f4ea6bf7ed526984965ebF.jpg)    

---

耶稣只用五个面包和两条鱼(five loaves and two fish)就举办了一场盛大的野餐派对. 最后还多出来了12篮子的食物(12 basketfuls).    

![22.png](https://s.pc.qq.com/tousu/img/20210207/3152010_1612676877.jpg)    

---

耶稣可以让白开水变成美酒(turn water into wine), 可以让盲人看得见(make blind people see), 让一个瘸腿的人活蹦乱跳当然也不在话下.     

![23.png](https://s.pc.qq.com/tousu/img/20210207/6260199_1612676878.jpg)    

---

难怪 "Jesus" 常常用来表示惊讶的语气. Jesus!
