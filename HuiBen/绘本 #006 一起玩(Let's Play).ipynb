早上好! 今天我们该做些什么呢?  
我们可以去读一本书,去**采一些花**,或者去游泳,去**玩球**?  
我们也可以去爬一棵树,去**收集一些叶子**,或者去玩**捉迷藏**好啦.  
当然,我们还会去吃一些**奶酪**! 然后再玩**过家家**和打电话.  
直到最后——是时候跟大家说晚安啦!
