### 我好无聊啊    

Michael Ian Black | Debbie Ridpath Ohi    

西蒙与舒斯特出版公司    

![2.png](https://sc02.alicdn.com/kf/Ha7a59f0a63f14478a8f7615731c8cf0az.jpg)    

---

一个小女孩,    

![4.png](https://ae04.alicdn.com/kf/Hd958ea6180ff4f92b459afb4bbf1c5fbw.jpg)    

---

在各种抱怨: "我好无聊啊(bored)."    

![5.png](https://ae03.alicdn.com/kf/H0c953f85bea64d1a9f03c2591bb7c7d7U.jpg)    

---

突然, 她发现了一个土豆(potato). 她想着: "我该拿这个土豆做什么呢?" 然后, 她一手就把它送上了天.    

![6.png](https://ae03.alicdn.com/kf/H2611237514ce4d28b1280910d64f0c77G.jpg)    

---

哎哟(ow), 土豆掉了下来, 正好砸在了她的头上. 然后土豆也在抱怨着: "我好无聊啊."    

![7.png](https://ae04.alicdn.com/kf/H88daf28b6ad84ebc921dbf643009ac33W.jpg)    

---

小女孩问土豆想要做什么? 喜欢做什么? 土豆说不知道, 然后说自己喜欢火烈鸟(flamingo). 小女孩说这附近没有火烈鸟. 土豆说: "好吧, 这好令人失望啊, 我好无聊啊."    

![8.png](https://ae03.alicdn.com/kf/H1fc58cdf80914a48a257453046433714c.jpg)    

---

小女孩好奇: "一个土豆怎么可能会无聊呢?" 土豆说: "因为我必须跟一个无聊的孩子在一起玩(hang out). 所有的孩子都很无聊." 然后小女孩说: "你瞎说什么呢(what are you talking about)? 孩子都是很好玩的!" 土豆又说: "那你能证明嘛?"    

![9.png](https://sc01.alicdn.com/kf/H8e5603740da0477fbb73d61f03a319f4n.jpg)    

---

小女孩表演了翻跟头(turn cartwheel), 跳跃, 还有围着自己转呀转(spin), 一直转到自己头晕(dizzy)想吐了(throw up). 土豆对此表示很无聊.    

![10.png](https://sc04.alicdn.com/kf/Hfa4af7f7d2bf4fc5bf2cab1a78a0acc4e.jpg)    

---

小女孩又表演了套圈圈, 忍者踢腿(ninja kicks), 以及双手倒行(walk on hands). 土豆对此表示依然很无聊.    

![11.png](https://sc02.alicdn.com/kf/H7e21cc36ddc64b089c0cc467e20acea99.jpg)    

---

小女孩又说: "你知道吗, 孩子们可以想象一些东西!" 土豆说: "想什么?" "你看, 现在我是一位著名的芭蕾舞演员(ballerina)."    

![12.png](https://ae03.alicdn.com/kf/H1f6558d44e7b44a29020e84b1cd76da2l.jpg)    

---

"现在我是一名驯狮师(lion tamer), 在训练世界上最凶猛的(ferocious)狮子; 现在我是一位童话公主(fairy princess), 我拥有自己的城堡(castle), 还有自己的龙(dragon), 以及独角兽(unicorn)." 土豆对此表示依然很无聊, 无聊到它都开始打呼噜了(snoring).    

![13.png](https://ae02.alicdn.com/kf/Hfb4591eabfa64984a043c0d8aaf83d08V.jpg)    

---

孩子们可以荡秋千(swing), 可以跳跃(jump), 还可以飞翔(flying)...    

![14.png](https://sc02.alicdn.com/kf/H24853a9bfcca4f0a80bc722ad34f2548N.jpg)    

---

无论小女孩做什么, 土豆都表示很无聊.    

![15.png](https://ae04.alicdn.com/kf/H9ee91c4afac44112ba11da13e52732del.jpg)    

---

无奈, 小女孩最后把自己也伪装成了一个大土豆, 可是呢, 土豆对此表示依然很无聊.    

![16.png](https://ae03.alicdn.com/kf/H6f8c84232f0348b29d853ceedbc1ec6f1.jpg)    

---

我们可以做任何酷酷的事情(cool stuff), 拥有任何自己的想法, 以及做任何自己想做的事, 你怎么觉得这也很无聊呢! 我宁愿当一个小孩也不愿当一个土豆!    

![17.png](https://ae03.alicdn.com/kf/Hca566591881046a59e3f8424193b0e18x.jpg)    

---

小女孩不理土豆了, 头也不回的走了. 然后, 来了一只火烈鸟, 引起了土豆的兴趣.    

![18.png](https://ae02.alicdn.com/kf/Haa388687102d400a93825d99a3bfa9500.jpg)    

---

但是呢, 火烈鸟却又在嚷嚷着: "我好无聊啊!"    

![19.png](https://ae02.alicdn.com/kf/Hda372fabbad1497fa22ac5c8dfcf00c8p.jpg)
