### 青蛙是一个英雄    

马克斯·维尔修思    

---

![05](https://ae04.alicdn.com/kf/H8343aeaea6ae43f6a06bd31ce6c49832q.jpg)    

Dark clouds were gathering in the sky.    
乌云在天空中聚集.    

The sun disappeared behind the clouds.    
太阳消失在了乌云之后.    

---

![06](https://ae04.alicdn.com/kf/Hc072c7aa7f6b4b169817b8d036747edfd.jpg)    

"It's starting to rain," thought Frog happily.    
"开始下雨了," 青蛙开心的想着.    

The first drops were already falling on his bare skin.    
第一滴雨也落在了他干巴巴的皮肤上.    

Frog loved the rain.    
他喜欢下雨天.    

---

![07](https://ae02.alicdn.com/kf/Hadf9e2aa25d74e50ab8b98ded8d462a8a.jpg)    

He danced for joy as the raindrops fell thick and fast.    
雨下的越来越密, 越来越急, 青蛙开心的跳了起来.    

"It's raining, it's raining. My shorts are soaking wet!" he sang loudly.    
"下雨了, 下雨了, 我的短裤都湿透啦!" 他大声的唱着.    

---

![08](https://ae04.alicdn.com/kf/H3df0bd15897240a2adf341599f10a5b8b.jpg)    

The sky got darker and darker.    
天空变得越来越黑了.    

Now it was raining cats and dogs.    
雨也下得越来越大了.    

This was a bit too much even for Frog.    
即使是对一只青蛙, 这雨下得也太大了.    

He ran home dripping wet.    
然后, 他湿淋淋的跑回了家.    

---

![09](https://ae02.alicdn.com/kf/H294fbae40c9a4e09b4098ea11f392e42G.jpg)    

Frog made himself a nice hot cup of tea.    
青蛙给自己弄了一杯好喝的热茶.    

The raindrops pattered against the window but it was cosy inside.    
雨水拍打着窗户, 但是屋子里又温暖又舒适.    

After three days of rain however, Frog began to feel restless.    
然而, 雨下了三天之后, 青蛙开始有点不耐烦了.    

He wondered how Duck and Pig and Hare were.    
他想知道鸭子, 小猪和兔子在做什么.    

He hadn't seen them since the rain started.    
自从下雨后, 他就一直没有见过他们.    

---

![10](https://ae02.alicdn.com/kf/H03163978d5df4441af7e5e2198310d2aZ.jpg)    

On the fifth day, the river began to rise.    
到了第五天, 河水开始上涨了.    

It wasn't long before water came streaming into Frog's house.    
没过多久, 水就开始往青蛙的房子里漫了.    

At first, Frog thought it was funny but then he began to worry.    
起初, 青蛙还觉得好玩, 可是不一会儿他就开始担心了.    

---

![11](https://ae02.alicdn.com/kf/Haa7f4262330b4e92a0dffdf6ff501036L.jpg)    

He hurried over to Duck's house.    
他急匆匆的来到了鸭子家.    

It was flooded there as well.    
鸭子家也发大水了.    

"Where is all this water coming from?" asked Duck desperately.    
"这些水都是打哪儿来的啊?" 鸭子绝望的说.    

"The river has burst its banks," shouted Frog. "Let's go to Pig's house."    
"河水冲破了堤岸!" 青蛙大喊着. "让我们一起去小猪家吧!"    

---

![12](https://ae03.alicdn.com/kf/H1bbd5b8503df465f9b0f81d199bd2e4aK.jpg)    

Together they waded through the watery landscape.    
他们跋涉着穿过了这些"水地".    

---

![13](https://ae03.alicdn.com/kf/H35561e41216545a69a9e648a2a26ededC.jpg)    

Pig was leaning out of his attic window.    
小猪正在阁窗边斜出头来.    

"All my things are wet," he cried.    
"我所有的东西都湿透了!" 他大喊着说.    

---

![14](https://ae04.alicdn.com/kf/Hb02bf415ceee4342b1b264265655be18g.jpg)    

It was true.    
这是真的.    

Tables and chairs were floating around the room.    
桌子和椅子正在房子里到处漂呢.    

Everything was in a mess.    
所有的东西都乱成了一团.    

They couldn't stay there.    
他们无法待在这里.    

"Let's go and see Hare," suggested Frog.    
"让我们一起去兔子家看看吧," 青蛙提议道.    

---

![15](https://sc02.alicdn.com/kf/H8eb59f39ffca4dec97ad1cbad688a422H.jpg)    

Hare's house was on an island in the middle of the water.    
兔子的房子建在水中央的一座岛上.    

Hare stood at the door and waved on them.    
兔子站在门口向大家招手.    

"Come inside," he shouted. "It's dry in here."    
"快进来啊," 他大喊着. "我家里很干燥!"    

---

![16](https://ae03.alicdn.com/kf/Hfd3d5c333982498fa722059d4f09287ei.jpg)    

It was warm inside.    
兔子家很温暖.    

Gratefully, they dried themselves in front of the stove and told Hare how their houses had been flooded.    
心怀感激地, 他们在炉子前烘干了自己, 并把他们家是如何发大水的告诉了兔子.    

"You must all stay here," said Hare. "There's plenty of room and I've got plenty of food."    
"你们一定得留下了," 兔子说. "我家有很多房间, 也有很多食物."    

---

![17](https://ae02.alicdn.com/kf/H496db458a61e4be7bb3a90617682b4bbh.jpg)    

So they all sat down to a big pot of stew Hare had made.    
于是, 他们全都坐在了一大锅炖菜面前, 这是兔子做的.    

They were very hungry and they ate everything up.    
他们非常的饿, 很快就把所有东西吃完了.    

Then they settled down to a cosy evening, with the rain still pattering against the windowpanes.    
他们安顿了下来, 然后度过了一个舒适的夜晚, 而外面的雨水仍然在拍打着窗户.    

---

![18](https://ae03.alicdn.com/kf/Hd5765a0cdb8e45e7961476f41ed770ac9.jpg)    

They stayed as Hare's guests for days.    
他们在兔子家待了好多天, 作为客人.    

They were happy together, while outside it rained and rained.    
他们待在一起很开心, 而外面的雨仍然在不停地下啊下.    

---

![19](https://ae02.alicdn.com/kf/H97d325c42884411087ddeb861e238cfaM.jpg)    

Then, one day, they found they were down to their last loaf of bread.    
然后, 有一天, 他们发现他们只有最后一条面包了.    

"We have no more food left," declared Hare gravely.    
"我们没有更多剩余的食物了," 野兔严肃的宣布.    

"We'll die if we don't get help," said Duck.    
"我们会死的, 如果我们得不到帮助的话," 鸭子说.    

"I don't want to die," said Frog, "ever."    
"我不想死," 青蛙说, "一点都不想死."    

---

![20](https://ae02.alicdn.com/kf/H05c417f6a9a84198b38101b064368073z.jpg)    

The next day only the last crumbs of bread were left.    
在接下来的这一天, 他们把最后一条面包也吃掉了(只剩下最后一点儿面包屑了).    

They were all terribly hungry, but nobody knew what to do.    
他们都非常的饥饿, 但是没有人知道该怎么做才好.    

Outside, it had stopped raining but the water was still very high.    
外面, 雨已经停了, 但是水位仍然很高.    

---

![21](https://ae04.alicdn.com/kf/H47d531d06a3d4585bd6c661ed87fcd99x.jpg)    

"I know!" shouted Frog suddenly. "I'll swim across to those hills and fetch help."    
"我知道该怎么做了!" 青蛙突然大喊着. "我要游过这些山丘去寻找帮助."    

Hare looked concerned.    
兔子担心的看着他.    

"The current is very strong and it's such a long way," he said. "It's too dangerous."    
"水流这么的湍急, 并且路途这么的遥远," 他说. "这太危险了."    

"But I can manage it," cried Frog enthusiastically. "I'm the best swimmer of us all."    
"但是我可以做到的," 青蛙充满热情地说. "我是大家当中游泳游得最好的."    

They knew this was true.    
"大家都知道, 这是真的.    

---

![22](https://ae03.alicdn.com/kf/H00a437d5d6674e4aa52cd1bea369d919G.jpg)    

So Frog stepped bravely into the water.    
于是, 青蛙勇敢地迈进了水里.    

His friends watched nervously.    
他的朋友全都紧张地看着.    

Soon, he disappeared into the distance.    
不久, 青蛙便消失在了视线之中.    

---

![23](https://ae03.alicdn.com/kf/H997e57b217ae4bffad254afc16c5ac2eJ.jpg)    

The water was ice cold, but Frog didn't think about it.    
水是冰凉冰凉的, 但是青蛙根本顾不上这些.    

He thought of Duck and Hare and Pig who were hungry.    
因为, 他满脑子里想的都是鸭子,兔子和猪猪在忍受饥饿.    

---

![24](https://ae02.alicdn.com/kf/H60913572dde84294aaa3697a68e9e3dbs.jpg)    

The further Frog swam, the stronger the current became.    
青蛙游得越远, 水流就变得越湍急.    

Frog felt tired.    
青蛙很累了.    

He was hardly making any headway.    
他一点儿都无法前行了.    

Suddenly the current carried him away.    
突然, 水流把他卷走了.    

---

![25](https://sc01.alicdn.com/kf/H0373b0a11d6b42b28c477d0eea925e35e.jpg)    

Frog began to sink.    
青蛙开始下沉.    

"I'm just a frog that can't swim any more," thought Frog.    
"我是一只再也游不动的青蛙了," 青蛙想着.    

"I'll drown. I'm going to die and I'll never see my friends again."    
"我要淹死了. 我要死了, 我再也不会见到我的朋友们了."    

---

![26](https://ae03.alicdn.com/kf/H501efd29557f4b01add65d99f777575cL.jpg)    

Just then a familiar voice said, "Hello! What have we here?"    
就在这时, 一个熟悉的声音说道, "嗨! 看看是谁在这儿?"    

Two strong arms pulled Frog out of the water and into a boat.    
两只强壮的手臂把青蛙拉出了水面, 拖到了船上.    

It was Rat.    
这是老鼠.    

---

![27](https://ae02.alicdn.com/kf/H9a5e5222f7cb4bd59bb67836b3951ccdD.jpg)    

Frog told Rat all about the rain, the flood and the hunger, and how he had set out to get help.    
青蛙把这场雨里发生的所有事都告诉了老鼠: 什么发大水啦, 饥饿啦, 还有他怎么出发去寻求帮助啦.    

"Don't worry," said Rat.    
"不要担心," 老鼠说.    

"My boat is full of provisions for my travels. There's plenty of food here for everyone."    
"我的船上都是我为旅行准备的供给品. 这里有很多的食物, 足够我们每个人的."    

And he set sail for Hare's house, where the three friends were waiting for help to arrive.    
于是, 他们向着兔子家的方向驶去, 在那儿, 三个朋友正在等待帮助的到来.    

---

![28](https://ae02.alicdn.com/kf/H972143186e5c4551a705ab44a67de6a81.jpg)    

Pig, Duck and Hare cheered when they saw Frog return in a boat.    
小猪, 鸭子和兔子欢呼了起来, 当他们看到青蛙乘着小船回来的时候.    

But who was that with him?    
但是陪着他的是谁啊?    

---

![29](https://ae04.alicdn.com/kf/H9be3c5123c47450282fc22c50b6bbbe7Q.jpg)    

Of course, it was their good friend Rat!    
当然啦, 这是他们的好朋友老鼠!    

They could hardly believe their eyes.    
他们简直不敢相信自己的眼睛.    

---

![30](https://ae02.alicdn.com/kf/H3a5d557f73ea415c914378621671173b8.jpg)    

And Rat had so much food on board— bread, honey, jam, peanut    -butter, vegetables, potatoes and much more besides.    
老鼠的船上有太多的食物了—— 面包, 蜂蜜, 果酱, 花生酱, 蔬菜, 土豆, 除此之外还有很多.    

---

![31](https://ae02.alicdn.com/kf/H2e82c3808d444b61bc4334e732c79ed9L.jpg)    

"Rat, you've saved us," said Hare.    
"老鼠,你救了我们!" 兔子说.    

"No," said Rat, "you have Frog to thank for that. It was Frog who swam through the treacherous flood, risking his life to reach me."    
"不," 老鼠说, "你们应该感谢的是青蛙. 是他冒着生命危险, 游过了湍急的水流, 找到了我."    

They all looked at Frog.    
他们全都看向了青蛙.    

He was glowing with pride.    
他浑身都散发着自豪感.    

It wasn't exactly true, but still...    
尽管这并不完全是真的, 但是...    

---

![32](https://ae03.alicdn.com/kf/H0a3170e840944430b8430f3e927bda6bB.jpg)    

From then on things got better.    
在此之后, 事情就变得顺利多了.    

The friends celebrated their rescue, and Frog was the hero.    
朋友们都在庆祝他们的"被营救", 而青蛙是他们的英雄.    

The sun was shining again and the water was beginning to go down.    
太阳又开始出来照耀了, 大水也正在开始退去了.    

---

![33](https://ae02.alicdn.com/kf/H0adb567a558141ba933afd68e69bf5d0R.jpg)    

After a couple of days, the water had gone.    
又过了几天, 大水就完全退去了.    

Frog, Duck and Pig were able to return to their homes.    
青蛙, 鸭子和小猪也能够重返他们的家了.    

---

![34](https://sc01.alicdn.com/kf/H75a1d148e4414dc4b6672e440cae2513u.jpg)    

But everything was dirty and muddy.    
但是所有的东西看起来都脏兮兮的, 而且沾满了泥土.    

"No problem," said Rat, and with his help, they fixed things up just as they had been before.    
"这不是问题," 老鼠说. 并且有了他的帮助, 他们很快就把事情修复的跟原来一个样了.    

But things weren't quite the same as before.    
但是事情并未完全像之前一样.    

None of them would ever forget the terrible flood.    
因为他们永远也不会忘记这次糟糕的大洪水了.
