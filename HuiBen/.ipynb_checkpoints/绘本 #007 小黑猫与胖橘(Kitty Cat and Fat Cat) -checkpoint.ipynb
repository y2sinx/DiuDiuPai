### 故事大意  

小黑猫跑呀跑,来到了一堆红叶旁.  
突然,它看到了一只胖橘.  
胖橘的尾巴一上一下的,看起来挺好玩的!  
此时的胖橘正在盯着一只小老鼠.  
嘴里还在嘀咕着: "哈哈! 看你怎么逃出我的手掌心!"  

突然,小黑猫跳了起来,一把抓住了胖橘的尾巴.  
小黑猫说: "喵! 对不起哦!"  
胖橘说: "喵你个大头鬼啊! 吓死我啦!"  

小老鼠听到喵叫后吓得立马跑掉了.  
胖橘很生气,怒气冲冲的看着小黑猫.  
"喵你个二百五啊,就知道喵! 你没看见我在逮老鼠吗!" 胖橘说,  
"敢打扰我的好事,看我怎么揍你!"  

胖橘转身扑向了小黑猫.  
小黑猫被吓得立马跑了起来(往家的方向),胖橘则在其后紧紧地跟着.  
就在胖橘抓住小黑猫的刹那间,小黑猫成功的钻回了家.
