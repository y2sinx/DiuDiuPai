### 青蛙和鸟鸣    

马克斯·维尔修思    

![02.png](https://s.pc.qq.com/tousu/img/20210205/7160231_1612502514.jpg)    

---

一个美好的秋日, 小猪正在果园里(orchard)摘成熟的苹果(ripe apples)...    

![04.png](https://s.pc.qq.com/tousu/img/20210205/3806345_1612502535.jpg)    

---

...然后青蛙走了过来. 他看起来有些担心(worried).    

![05.png](https://s.pc.qq.com/tousu/img/20210205/2129606_1612502522.jpg)    

---

他说: 我发现了些什么.    

小猪问: 是什么?    

青蛙答道(replied): 跟我来, 我指给你看(show you).    

![06.png](https://s.pc.qq.com/tousu/img/20210205/6164655_1612502533.jpg)    

---

然后他们一起出发了(set off together). 小猪感觉有些紧张(nervous).    

![08.png](https://s.pc.qq.com/tousu/img/20210205/3128951_1612502517.jpg)    

---

他们来到了一处空旷地(clearing).

青蛙指着地面说: 看, 这只黑鹂(blackbird)有点不对劲(something wrong), 他一动不动的(moving).    

![09.png](https://s.pc.qq.com/tousu/img/20210205/8871889_1612502523.jpg)    

---

小猪说: 他在睡觉吧.    

就在这时, 鸭子也过来了.    

![11.png](https://s.pc.qq.com/tousu/img/20210205/1952203_1612502531.jpg)    

---

鸭子担心的问(asked with concern): 出什么事了, 发生意外了(accident)吗?    

青蛙说: 嘘(ssh), 他正在睡觉呢.    

但是鸭子却认为黑鹂看上去更像是生病了(look ill).    

![12.png](https://s.pc.qq.com/tousu/img/20210205/2875293_1612502522.jpg)    

---

就在那时, 兔子正在树林里散步. 他从老远就看到这儿好像发生了什么事(something was going on), 于是他也加入了他们.    

![13.png](https://s.pc.qq.com/tousu/img/20210205/1912268_1612502518.jpg)    

---

兔子跪在(knelt beside)黑鹂的旁边, 说: 他死了.    

青蛙说: 死了! 什么是死啊?    

![14.png](https://s.pc.qq.com/tousu/img/20210205/4131825_1612502531.jpg)    

---

兔子指了指蓝天说: 所有生物都会死去.    

青蛙不解的问: 我们也会死去吗?    

兔子并不确定, 然后补充道: 或许吧, 当我们老了的时候.    

![15.png](https://s.pc.qq.com/tousu/img/20210205/5666778_1612502522.jpg)    

---

兔子说: 我们得把他埋了(bury), 就埋在山脚下(the bottom of the hill)的那个地方吧.    

他们一起做了一个担架(stretcher), 然后把黑鹂抬到了草地上(meadow).    

![17.png](https://s.pc.qq.com/tousu/img/20210205/5258875_1612502531.jpg)    

---

他们在地上挖了一个深坑(deep hole).    

兔子说: 黑鹂的一生都在为我们歌唱, 现在他可以安息了(earned his rest).    

![19.png](https://s.pc.qq.com/tousu/img/20210205/7549660_1612502533.jpg)    

---

他们小心翼翼地把黑鹂(the dead bird)放到了坑里. 青蛙还在黑鹂的周围扔了些鲜花(flowers). 之后他们便用土(earth)把他盖住了.    

![20.png](https://s.pc.qq.com/tousu/img/20210205/6638343_1612502529.jpg)    

---

最后, 他们还在坑的上方放了一块漂亮的石头, 这看起来很安详(peaceful). 此刻这里没有一丝声响(sound)—— 更别提一声鸟鸣了(one note of birdsong).    

![21.png](https://s.pc.qq.com/tousu/img/20210205/2065559_1612502516.jpg)    

---

他们带着"些许感触"(be moved)静悄悄地走在回去的路上(on their way). 突然, 青蛙冲到了前面.    

他激动地大喊着: 我们一起来玩"抓猫猫(play catch)"的游戏吧. 小猪你来当"猫猫".    

![23.png](https://s.pc.qq.com/tousu/img/20210205/6836618_1612502535.jpg)    

---

他们玩耍着, 嬉笑着, 一直玩到了日落(sunset)才停了下来.    

![24.png](https://s.pc.qq.com/tousu/img/20210205/2628894_1612502519.jpg)    

---

青蛙说: 生命真的好奇妙啊!    

![25.png](https://s.pc.qq.com/tousu/img/20210205/6235010_1612502553.jpg)    

---

然后这群疲惫的小家伙们便开心的回家去了. 就在他们走过山脚的时候, 他们听到了一个声音. 一只黑鹂正站在一棵树上, 唱着一首欢快的歌—— 正如往常一样.    

![26.png](https://s.pc.qq.com/tousu/img/20210205/7736465_1612502520.jpg)
