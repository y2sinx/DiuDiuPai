在这本书中,你会遇到**帝企鹅**(king penguins),**帽带企鹅**(chinstrap penguins),还有一只孤单的**马可罗尼企鹅**(macaroni)... 正如**麻絮兄弟**所见到的那样,而麻絮兄弟是帝企鹅家族的.  

所有这些企鹅种类都可以在南极洲上发现,而南极洲是地球上一片最冷,最干,最荒凉的地方. 只有拥有一些特殊的技能(special qualities),才能在如此恶劣的环境中(harsh landscape)生存下来. 如果你能有幸到此处参观,你会发现,无论这些企鹅看起来多么的相似,但事实上它们每一位个体都具有独特的魅力.  

我们希望,通过了解这些非凡的动物,我们的后代也可以像我们那样去爱护它们.
