这一天,乌龟准备爬着穿过草原.    
他在不停的找着什么. 他往远处注视了一会儿,并将眼睛斜向了一棵小草.    
正在这时,他遇到了一只蜗牛.    

蜗牛问道: "你在找什么东西吗?"    
乌龟答道: "是的,蜗牛,我在找我的房子. 你有碰巧见过它吗?"

她爬到了乌龟的背上. 乌龟背着她继续前行了.    
他们找啊找,但是丝毫不见房子的踪迹.    
此时的太阳高高的挂在天空中.

麻雀说: "我没有,但是我也可以帮你找找!"

麻雀回应道: "好吧,那么我也跳到你的背上去吧."    
乌龟背着蜗牛和麻雀继续前行. 他们找啊找,但是丝毫不见房子的踪迹.

老鼠问道: "你们在找什么东西吗?"    
乌龟答道: "是的,老鼠,我们在找我的房子. 你有碰巧见过它吗?"

然后他也跳到了乌龟的背上.    
"等一下,我的花环掉了." 乌龟停了下来.    
老鼠伸长了胳膊够啊够,终于够到了他的花环.    

乌龟背着蜗牛,麻雀,瓢虫和老鼠继续前行着.    
他们找啊找,但是丝毫不见房子的踪迹.    
风吹着地面上的落叶,围着乌龟的脚打着旋儿.    
乌龟累了,叹息道: "啊! 你们好重啊!"
