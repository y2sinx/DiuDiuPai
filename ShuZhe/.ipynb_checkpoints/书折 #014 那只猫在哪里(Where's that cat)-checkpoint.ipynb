# Where's that cat?  

**Sam** Wilson / **Thea** Nicole de Klerk / **Chenél** Ferreira  

---

尼欧正在实验室里组装小汽车,而她的爸爸正在研究生物缩放机.  

![page4](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page4-scaled.jpg)  

---

尼欧好久没注意到自己的猫猫啦,她很好奇它跑到哪里去了.  

![page5](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page5-scaled.jpg)  

---

尼欧在大街上贴满了寻猫启示,但是她仍然找不到自己的猫猫.  

![page6](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page6-scaled.jpg)  

---

尼欧坐在厨房里,突然想到了什么: "猫猫会不会被爸爸的缩放机给缩小了呀!"  

![page7](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page7-scaled.jpg)  

---

尼欧拿着一个放大镜,在门前发现了两串小小猫的脚印. 她认为猫猫应该是跑到花园里去了. 但是花园那么大,猫猫那么小,她该怎么找才好呢?  

![page8](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page8-scaled.jpg)  

---

于是,她用爸爸的生物缩放机把自己也给缩小了.  

![page9](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page9-scaled.jpg)  

---

哈哈! 现在的尼欧变成了一个小小人了.  

![page10](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page10-scaled.jpg)  

---

小尼欧来到了花园,她的身高只有花儿一般大小. 当然,蝴蝶和蜈蚣在她面前也变成了庞然大物.  

![page11](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page11-scaled.jpg)  

---

小尼欧为了躲避大蜈蚣的进攻,跑到了厨房. 就在此时,她突然看到了什么.  

![page12](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page12-scaled.jpg)  

---

是的,她在这里看到了自己的猫猫. 可是现在的问题是: 她们怎样才能穿过花园回到实验室,变回她们原来的样子呢? 哈哈! 聪明的小尼欧当然有好主意啦!  

![page13](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page13-scaled.jpg)  

---

她们钻进了那辆刚刚组装好的小汽车,在花园里驰骋了起来.  

![page14](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page14-scaled.jpg)  

---

最终,她们不仅把自己变回了原来的样子,而且还把小汽车也给变成了正常的大小!  

![page15](https://bookdash.org/wp-content/uploads/2020/05/wheres_that_cat_en_20200515_page15-scaled.jpg)
