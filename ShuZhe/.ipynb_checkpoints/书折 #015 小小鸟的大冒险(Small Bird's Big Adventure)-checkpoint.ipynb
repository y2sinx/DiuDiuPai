# Small Bird's Big Adventure  

**Wesley** van Eeden `/` **Nick** Mulgrew `/` **Jennifer** Jacobs  

---

小小鸟是一只小鸟.  
他是如此的 "娇小玲珑".  

![04](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page4.jpg)  

---

小小鸟与巨人一起生活在一间漂亮的房子里. 小小鸟非常的喜欢巨人.  

巨人是一位善良的人. 她给了小小鸟很多的玩具.  

![05](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page5.jpg)  

---

当巨人不在小小鸟身边的时候,小小鸟感到非常的难过.  
于是有一天,小小鸟决定去寻找巨人.  

![06](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page6.jpg)  

---

小小鸟离开了自己的笼子.  

![07](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page7.jpg)  

---

一直往天上飞呀飞,直到他看到了整个世界.  

他想要去寻找巨人.  
他需要在哪里着陆呢?  

![08](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page8.jpg)  

---

小小鸟落在了一片巨大的森林中,森林里面满是树木,花草,以及宝藏.  

小小鸟喜欢与这些亮晶晶的珠宝玩耍! 但是巨人在哪儿呢?  

![09](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page9.jpg)  

---

小小鸟踏进了一片辽阔的雪之田野,这上面的雪又白又软.  

小小鸟喜欢在这凉凉的地面上翻滚. 但是巨人在哪儿呢?  

![10](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page10.jpg)  

---

小小鸟又飞上了天,但是这一次他比上一次飞的还高.  

他已经飞到了群星和星系之中!  
他看到了月亮! 还看到了彗星!  

但是巨人在哪儿呢?  

![11](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page11.jpg)  

---

在高高的天空上,小小鸟用他那敏锐的双眼,似乎看到了什么!  

他看到了什么呢?  

![12](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page12.jpg)  

---

小小鸟往下俯冲(dive),冲啊,冲啊!  

他会落在哪儿呢?  

![13](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page13.jpg)  

---

小小鸟正好落在了巨人的双手里.  

他尖声的说着: "我爱你,巨人!"  

![14](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page14.jpg)  

---

巨人把小小鸟推向了自己.  

对他说: "我也爱你!"  

![15](https://bookdash.org/wp-content/uploads/2017/02/small-birds-big-adventure_english_20161230_page15.jpg)
