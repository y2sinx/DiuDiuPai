"长颈鹿,长颈鹿,请你帮我找到我的笑声吧!"

![page5](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page5-scaled.jpg)    

---    

"河马,河马,请你帮我找到我的笑声吧!"    
河马说: "我这下面没有你的笑声啊."    

![page6](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page6-scaled.jpg)    

---    

"疣猪,疣猪,请你帮我找到我的笑声吧!"    
疣猪说: "我这里也没有你的笑声啊!"    

![page7](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page7-scaled.jpg)    

---    

"猴子,猴子,请你帮我找到我的笑声吧!"    
猴子问: "你是怎么把它弄丢的啊?"    

![page8](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page8-scaled.jpg)

"当我笑起来的时候,我的大牙就露出来了. 这把所有的动物都给吓坏了." 斑斑说, "然后我就难过了起来,再然后我的笑声就不见了."    

![page9](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page9-scaled.jpg)    

---    

"我找遍了所有地方都没有找到它!"    
猴子说: "但是,你找错了地方啊!"    

![page10](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page10-scaled.jpg)    

---    

随后猴子跳到了一棵树上,在上面取下了一片羽毛.    
然后,她开始在斑斑的身上到处挠痒痒.    

![page11](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page11-scaled.jpg)    

---    

一开始,斑斑还只是微微地笑,但没过多久他就开始哈哈大笑了.    
他笑着笑着,一直笑到捧着肚子在地上打滚了.    

![page12](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page12-scaled.jpg)    

---    

哈哈哈...    
其他所有的动物也都跟着笑了起来.    

![page13](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page13-scaled.jpg)    

---    

他们问: "你是在哪里找到了他的笑声的啊."    
猴子回答道: "他的笑声一直都在他的体内. 我只是让他开心了起来,然后笑声就从他的肚子里跑出来了."    

![page14](https://bookdash.org/wp-content/uploads/2018/01/the-lost-laugh_en_20181216_page14-scaled.jpg)    

---    

他们一直笑啊笑,直到他们的大牙也全都露出来了.    
"哈哈哈!" 鬣狗斑斑开心的说, "我再也不会把我的笑声了给弄丢了."
