In a village on the slopes of Mount Kenya, a little girl worked in the fields with her mother. She worried as she saw the big forests disappearing, but she knew the power of a tiny seed.    
这是一个关于"一位非常有名的非洲女性"(旺加里⋅马塔伊)的故事. 当她注意到自己国家的森林逐渐消失之后,她用一粒粒小小的种子重新把森林种了回来.

在东非肯尼亚山山坡上的一个小村庄里,这个小女孩正在同她的母亲在田里劳作. 她的名字叫旺加里(Wangari).  

---  

![5](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_2015020851.jpg)  

旺加里喜欢待在外面. 在自家的菜园里,她会用砍刀(machete)将土壤打碎. 她会将一粒粒微小的种子压入暖和的土里.  

---  

![6](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_2015020861.jpg)  

她最喜欢的时刻是在日落之后. 当天色太暗而看不到植物时,旺加里知道此时该回家了.  

在她回家的路途中,她将沿着狭窄的小径穿过田野和河流.  

---  

![7](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_2015020871.jpg)  

旺加里是一个聪明的孩子,总是迫不及待地想去上学. 但是她的父母希望她能留在家里帮他们干活.  

当她七岁的时候,她的哥哥说服了父母,把她送去了学校.  

---  

![8](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_2015020881.jpg)  

她喜欢学习知识!  

随着旺加里看了一本又一本的书,她学到了越来越多的知识.  

由于她在学校的表现非常出色,因此被邀请去了美国学习.  

旺加里非常兴奋! 她想要了解关于世界的更多知识.  

---  

![9](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_2015020891.jpg)  

在美国的大学里,旺加里学到了许多新东西. 她研究了植物及其生长. 她还记得自己是如何长大的: 在美丽的肯尼亚森之林荫中,与兄弟们一起玩耍.  

---  

![10](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208101.jpg)  

随着她学到的知识越来越多,她就越来越多的意识到自己对肯尼亚人民的爱意. 她希望他们能够过上幸福和自由的生活.  

随着她学到的知识越来越多,她就越来越多的想念自己在非洲的家.  

---  

![11](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208111.jpg)  

当她完成学业之后,她回到了肯尼亚. 但是她的国家变了. 巨大的农场横跨在(stretch across)土地之上.  

妇女们也没有了做饭的柴火. 人们很穷,孩子们很饿.  

---  

![12](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208121.jpg)  

旺加里知道自己该怎么办了. 她教会了妇女们如何用种子种树.  

这些妇女们卖掉了树木之后,用这笔钱照顾了自己的家人.  

妇女们很高兴. 在旺加里的帮助下,她们变得更加强大和拥有力量了.  

---  

![13](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208131.jpg)  

随着时间的推移,这些树木渐渐长成了森林,河流又开始流淌在肯尼亚大地之上. 很快,旺加里的故事就传遍了整个非洲.  

如今,数以百万计的树木都是用旺加里的种子长成的.  

---  

![14](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208141.jpg)  

之后,全世界的人们都注意到了旺加里的努力工作,并给了他一个非常有名的奖,这项奖就是鼎鼎有名的诺贝尔和平奖. 她是有史以来第一位获得该奖项的非洲女性.  

---  

![15](https://bookdash.org/wp-content/uploads/2015/02/a-tiny-seed_interior-spreads_20150208151.jpg)  

旺加里于 2011 年去世. 如今,当非洲人看到一棵美丽的树时,他们总会想起旺加里.
