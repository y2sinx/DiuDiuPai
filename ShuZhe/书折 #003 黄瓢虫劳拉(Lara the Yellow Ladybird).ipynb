---

Lara the ladybird was a special bug.    
瓢虫劳拉是一个特别的虫子.    

Unlike all her friends, she had bright yellow wings.    
不像她所有的朋友(那样拥有红色的翅膀),她的翅膀是明亮的黄色的.    

![05](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_05.jpg)    

[Lara sits on a flower. She watches the other ladybirds. They all have red wings. Her's are yellow.    
劳拉坐在花朵上. 看着其他的瓢虫们. 他们都有红色的翅膀. 而她的翅膀则是黄色的.]()    

---    

Everyone loved her yellow wings.    
每个人都喜欢她的黄翅膀.    

Each morning, Bibi Butterfly said hello.    
每天早上,蝴蝶比比总是会对她说哈喽.    

And Manto **Mantis** always waved.    
而**螳螂**曼托也会对她招招手.    

![06](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_06.jpg)    

[Lara's mother sends her off to school. On her way she waves to Manto Mantis and Bibi Butterfly.    
劳拉的妈妈送她去学校. 在她去学校的路上,她会跟螳螂曼托与蝴蝶比比打招呼.]()    

---    

Even Sesa, the **sulky** spider, was happy to see her.    
甚至是塞萨,那个**看起来凶凶的**蜘蛛,见到她也会很开心.    

At school, she played with lots of friends.    
在学校,她也有很多一起玩的朋友.    

![07](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_07.jpg)    

[Sesa Spider hangs from her nest. Lara arrives at school. Her friends are happy to see her.    
蜘蛛塞萨挂在自己的巢穴下. 劳拉来到了学校. 她的朋友们见到她都很开心.]()    

---    

But Lara wanted to be like the other ladybirds.    
但是劳拉希望自己可以跟其他的瓢虫一样.    

"I wish I had red wings like you, Mama," she cried.    
"我希望我的翅膀也能像你的那样是红色的," 她对妈妈呼求着.    

![08](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_08.jpg)    

[At home, Lara cries to her mother. She dreams of red wings.    
劳拉在家中哭着对她妈妈说,她希望自己的翅膀也是红色的.]()    

---    

So one day, to cheer her up, Lara's mother painted her wings bright red.    
于是有一天,为了让她高兴起来,劳拉的妈妈把她的翅膀染成了红色的.    

![09](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_09.jpg)    

[Lara stands on a chair in front of the mirror. Her mother paints are yellow wings using a tin of red paint.    
劳拉站在镜前的一把椅子上,她的妈妈正拿着一罐红颜料给她的翅膀涂色.]()    

---    

The next morning, nobody greeted Lara on her way to school.    
接下来的一天早上,在她去上学的路上,竟然没有人给她打招呼了.    

![10](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_10.jpg)    

[On the way to school, Lara waves to the other insects. They don't wave back.    
在去学校的路上,劳拉给其他的昆虫们打招呼,但是他们却没有回应劳拉.]()    

---    

And when she got there, none of her friends said hello.    
当她到了学校,竟然也是一个朋友都没有跟她打招呼的.    

![11](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_11.jpg)    

[Lara arrives at school. Her friends don't greet her.    
劳拉来到了学校,但是她的朋友们都没有像往常那样跟她打招呼.]()    

---    

Lara sat all alone. No one noticed her new red wings.    
劳拉一个人孤独的坐着. 没有任何人注意到她的新红翅膀.    

Until Miss Miya spotted her and said: "You've painted your lovely yellow wings!"    
直到米雅小姐看见了她,并跟她说话: "你把你那好看的黄色翅膀给染掉了啊!"    

![12](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_12.jpg)    

[Lara sits alone. She cries to her teacher.    
劳拉一个人孤独的坐着,向她的老师哭诉.]()    

---    

Lara's classmates were shocked.    
劳拉的同班同学们都震惊了.    

"Your wings are special!"    
"你的翅膀是独一无二的!"    

"So unique!" "So rare!"    
"是多么的特别!" "多么的稀有啊!"    

![13](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_13.jpg)    

[Miss Miya tells the children what Lara has done. They are shocked.    
米雅小姐把劳拉做了什么全都告诉了孩子们. 他们都非常吃惊.]()    

---    

"Lara," Miss Miya said, "your yellow wings are what make you you.    
"劳拉," 米雅小姐说, "是你的黄色翅膀使你成为了你啊."    

"Like Sipho's spot ... and Sally's legs."    
"就像西波的斑点 ... 以及莎莉的腿,那是你们独一无二的标志啊."    

![14](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_14.jpg)    

[Miss Miya tells Lara about Sipho's large spot and Sally's long legs.    
米雅小姐告诉了劳拉,有关 "西波的斑点" 和 "莎莉的腿" 的故事.]()    

---    

Back home, Lara took a long bath and **scrubbed** until her golden wings gleamed.    
劳拉回到家后,洗了一个很长时间的澡,把她那金闪闪的翅膀**擦洗**的都发光了.    

"I'll never paint my wings again!" she thought.    
"我再也不会给我的翅膀染色了!" 她这样想着.    

Except, maybe once or twice ...    
除非,或许可以有一两次 ...    

![15](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_15.jpg)    

[Lara lies in the bath and **scrubs** the red paint off her wings. Her wings are yellow again.    
劳拉躺在浴缸里,把她翅膀上的红颜料都**擦去了**. 她的翅膀又变成了黄色的.]()    

---    

To try a bit of purple ... or something nice.    
尝试一点紫色 ... 或者其它好看的颜色.    

But not for ever and just for fun.    
不是为了永远那样,而只是为了好玩.    

![16](https://bookdash.org/wp-content/uploads/2016/04/the-yellow-ladybird_english_20160409_Page_16.jpg)    

[Lara holds a paintbrush. Her wings are painted with colourful dots and stars and stripes.    
劳拉拿起一把刷子,在她的翅膀上画了一些五彩斑斓的圆点,以及一些小星星和条纹.]()
