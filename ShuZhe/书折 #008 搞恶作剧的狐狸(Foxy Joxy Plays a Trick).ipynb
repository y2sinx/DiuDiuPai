> 1. 摘要:  
Foxy Joxy, a sly fox, sells watermelons. One day he has a clever idea. The animals were not happy. What will they do to teach him a lesson?  
狐狸狐狐卖西瓜, 突然有了个坏想法. 动物被骗不开心, 看大家怎么教训他?

狐狸狐狐是个狡猾的家伙.  
他在卖又大又新鲜的西瓜.

他大喊着: "卖西瓜啦! 卖西瓜啦! 卖好吃的西瓜啦!"  
但是他标的价格太高了.  

---

![5](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page5.jpg)

兔子说: "你这西瓜的价格太高了啊!"

长颈鹿也这样说: "你要的价太高了啊!"

---

![6](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page6.jpg)

突然,狐狸想到了一个坏主意.

---

![7](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page7.jpg)

他一刀切开了一个大西瓜,挖掉了里面的果肉,然后嘎嘣几口就给吞掉了!

---

![8](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page8.jpg)

然后他往里面装满了水!

他拿起了针线,把西瓜皮又重新缝在了一起.

---

![9](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page9.jpg)

他用一个较低的价格把西瓜卖了出去.

很多动物都从狐狸这里买了西瓜.

---

![10](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page10.jpg)

"这个西瓜里面全都是水." 猴子抱怨道.

---

![11](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page11.jpg)

狐狐数着钱嘚瑟的笑着说: "哼! 这就是你们想要的便宜货."

---

![12](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page12.jpg)

大象怒气冲冲的说: "我们得好好教训下这只臭狐狸."

---

![13](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page13.jpg)

动物们凑在了一起,集体想出了一个计策.

"我想到了一个主意!" 猫头鹰说. 

---

![14](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page14.jpg)

他们取了一个空心的西瓜,然后把狐狸塞了进去,随后就拿它当足球戏弄了一番.

---

![15](https://bookdash.org/wp-content/uploads/2017/03/foxy-joxy-plays-a-trick_english_20170320_page15.jpg)

这下,狐狸两眼发起了晕,头顶转起了小星星!

自此以后他学到了教训,再也不敢欺骗别人了.
