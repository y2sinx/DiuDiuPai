***        

In the land of All-the-Beasts there was a GREAT HUNGER. Some of the animals who were so HUNGRY were        
在万兽之国,发生了大饥荒. 这些极度饥饿的动物分别是        

Tabby Tiger,Bruno Bear,Katy Crocodile,Robin Rat,Pinky Pig,Giddy Goat,Tommy Tortoise        
老虎⋅塔比,狗熊⋅布鲁诺,鳄鱼⋅凯蒂,老鼠⋅罗宾,小猪⋅粉粉,山羊⋅吉迪,乌龟⋅汤米        

and many more—more than you could ever count in a year.        
还有更多更多的动物,多的超乎你的想象,即使一年你都数不过来.        

They ran around the wood, here and there and everywhere, eating roots and twigs and any old scraps they could find. But still they were HUNGRY.        
他们在树林里到处跑啊跑,寻找可以吃的任何东西,如树根或小细枝等. 但是他们仍然很饿.        

One day they came to a Big Tree full of fruit. But they could not eat it, for they did not know what it was.        
有一天他们找到了一棵长满水果的大树. 但是他们并没有吃树上的水果,因为他们不知道那是什么树.        

![6](http://www.gutenberg.org/files/59994/59994-h/images/i_006.jpg)        
***        

They sat down in a circle round the tree, and said, "What can we do?"        
他们围着树绕成一圈坐了下来,说: "我们能做什么?"        

When they had thought a while, they said, "Let us send Robin Rat up the river to Leo, our King, and ask him what the fruit is and whether we may eat it."        
他们想了一会儿后说: "不如我们派老鼠罗宾到去河边找我们的狮子国王,问问他这是什么水果,是否可以吃."        

![7](http://www.gutenberg.org/files/59994/59994-h/images/i_007.jpg)        
***        

Robin Rat was young and spry. He scuttled up the tree and brought down one of its fruits to show King Leo.        
老鼠罗宾年轻敏捷. 他迅速的爬上树,取下了其中一个水果,准备把它拿到了狮子国王的面前.        

It was a DELICIOUS looking fruit!        
这看起来像是个美味的水果!        

It looked like an        
看起来像是个        

APPLE-ORANGE-PLUM-PEAR-BANANA        
苹果,橘子,李子,梨,香蕉的混合体.        

but it smelled like a        
闻起来也像一个        

BANANAPEARPLUMORANGEAPPLE.        
苹果,橘子,李子,梨,香蕉的混合体.        

![8](http://www.gutenberg.org/files/59994/59994-h/images/i_008.jpg)        
***        

Then Robin Rat scuttled down to the river bank and climbed into his little canoe.        
然后老鼠罗宾迅速的跑到河岸边,爬进了他的小独木舟.        

All the day and all the day he paddled and paddled and PADDLED up the river.        
他在河上划呀划呀划,划了整整一天.        

And the Great Red Sun dropped behind the trees.        
红色的大太阳在树林后落山了.        

![9](http://www.gutenberg.org/files/59994/59994-h/images/i_009.jpg)        
***        

Then he found King Leo on the bank, all ready to receive visitors. He was wearing his crown tipped on the back of his head because he felt happy. He smiled at Robin Rat as pleasant as you please, and asked him to stay to supper.        
然后他在河岸边发现了狮子国王,他正准备接待来访者. 他的头上戴着王冠,看起来很开心. 他对老鼠罗宾笑了笑,并要求他留下来吃晚餐.        

![11](http://www.gutenberg.org/files/59994/59994-h/images/i_011.jpg)        
***        

After supper they curled up and went to sleep. There was nothing else to do, you see. For this is the way it looked in the GREAT WOOD.        
晚饭过后,他们蜷着身子睡觉了. 因为没有其他事情可做了. 这就是大树林中的生活方式.        

![12](http://www.gutenberg.org/files/59994/59994-h/images/i_012.jpg)        
***        

In the morning King Leo said politely, "What can I do for you, my small friend?"        
早上,狮子国王对老鼠罗宾礼貌地说: "我的小朋友,我能为你做点什么?"        

![13](http://www.gutenberg.org/files/59994/59994-h/images/i_013.jpg)        
***        

Then Robin Rat answered, "Please tell us, King Leo, what is the name of this tree and whether we may eat the fruit of it. We are all SO HUNGRY!"        
然后老鼠罗宾回答: "狮子国王,请告诉我们,这棵树的名字是什么,以及我们是否可以吃它的果实. 我们都非常的饿!"        

King Leo looked at the fruit that was like an APPLEORANGEPEARPLUMBANANA and he sniffed at the fruit that was like a BANANAPLUMPEARORANGEAPPLE.        
狮子国王看了看并闻了闻那个水果混合体.        

Then he said, "It is a good fruit. You may eat it. The name of the tree is BOJABI."        
然后他说: "这是好果子. 你们可以吃它. 这树的名字叫波亚比."        

![14](http://www.gutenberg.org/files/59994/59994-h/images/i_014.jpg)        
***        

Then Robin Rat hung his cap over his right ear and climbed into his little canoe.        
然后老鼠罗宾将帽子戴在右耳边,爬进了他的小独木舟.        

All the day and all the day he paddled down the great river.        
他在那条大河上又划了一天.        

And all the way he was thinking how much he could eat of that DELICIOUS fruit.        
一路上,他一直在想自己可以吃多少这种美味的水果.        

And at night he came home.        
晚上他回到了家.        

![15](http://www.gutenberg.org/files/59994/59994-h/images/i_015.jpg)        
***        

All the Beasts were waiting for him on the shore. He came up, whisking his paddle this way and that way through the water, just to show how well he could do it.        
所有的野兽都在岸上等着他. 他站起来后,用各种各样的方式在水中划桨,仅仅是为了炫耀一番自己的能耐.        

"What is it, Robin Rat?" said All the Beasts. "Tell us the name!" they roared and howled and grunted and whined and shrieked and squealed, each in his own PARTICULAR voice.        
"这是什么水果,老鼠罗宾?" 百兽说. "快告诉我们这棵树的名字!" 他们咆哮着,嚎叫的,咕哝着,抱怨着,尖叫着,每个人都用自己独特的声音尖叫着.        

"Oh!" said Robin Rat. "I knew it a while ago, but now I have clean forgotten."        
"哦!" 老鼠罗宾说."我之前还记得的,但是现在我已经忘了."        

![16](http://www.gutenberg.org/files/59994/59994-h/images/i_016.jpg)        
***        

Then All the Beasts stepped into the water and upset Robin Rat's little canoe.        
然后,所有的野兽都跳进了水中,打翻了老鼠罗宾的小独木舟.        

They SPLASHED and they SPLUTTERED and they SP-L-ANKED Robin Rat.        
他们气急败坏的在水中惩罚着老鼠罗宾.        

Squeaksqueaksqueaksqueaksqueak!        
吱吱,嘎吱,嘎吱,嘎吱!        

Nobody heard a word more from him that day.        
那一天,老鼠罗宾(吓的)再也没敢说一句话.
