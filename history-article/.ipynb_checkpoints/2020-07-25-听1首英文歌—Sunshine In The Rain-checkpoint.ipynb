## 大意     

我觉得这首歌讲的是一个制杖青年对前任(大美丽)舔狗式的自嗨    
并且严重怀疑这哥们的脑子被苍蝇腿给碾压过    

这哥们说他是大雨天中出现的太阳,嗯嗯,估计是在下太阳雨    
这哥们说他去柏林的时候,大美丽跑去了伦敦,他去伦敦追大美丽,大美丽又跑去了纽约    
然后这哥们又去纽约追大美丽,这下大美丽又跑去了罗马    
这哥们还时不时的回味一下他跟大美丽过时的风花雪月以及被遗忘的良梦春宵    
殊不知,他跟大美丽之间的爱早就变成了电话中可有可无的问候    
呐呐,当舔狗真可怜啊    

这哥们现在还幻想着跟他前任可以心有灵犀呢    
想对大美丽抱怨着那些没有她陪的夜晚    
这哥们希望可以重新设定他跟大美丽的旅途    
盼望着终有一天可以一起住进他们温暖的小窝    

以下是这哥们对大美丽的灵魂拷问    
我的大美丽,你感觉到你的沙漠中下雨了么    
我的大美丽,你感觉到你的黑暗中出现光亮了么    
我的大美丽,虽然我不在你的身边,但是你能感觉到我对你的爱么    
我的大美丽,即使与你相隔千里,我也不曾一刻忘记你    
啊啊,我那美丽的大美丽    

这哥们一遍遍的扯他是大雨天中的太阳    
一遍遍的诉说他对大美丽的爱不曾改变    

哈哈,终于,这哥们神志不清了    
开始幻想宇宙中出现拯救他爱情的至高力量    
啊啊,宇宙中的至高力量,让我和我的大美丽再见一面吧    
啊啊,宇宙中的至高力量,让我妙语连珠变成行动的巨人吧    
啊啊,宇宙中的至高力量,就让我跟我的大美丽永远在一起吧    

然后这哥们开启了洗脑循环
