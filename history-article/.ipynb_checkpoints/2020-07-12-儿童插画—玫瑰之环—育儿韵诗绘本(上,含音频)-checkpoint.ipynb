# THE MAN IN THE MOON  

The Man in the Moon  
Came tumbling down,  
And asked his way to Norwich;  
![](http://www.gutenberg.org/files/20652/20652-h/images/moon1_l.gif)  

They told him south,  
And he burnt his mouth  
With eating cold pease-porridge.  
![](http://www.gutenberg.org/files/20652/20652-h/images/moon4_l.gif)  

从月亮上掉下来一个男人,他问去诺维奇的路在哪,别人告诉他在南边.  
然后他在吃冷豌豆粥的时候,烫到了嘴.  


# TO MARKET, TO MARKET  

To market, to market, to buy a fat Pig;  
Home again, home again, dancing a jig.  
![](http://www.gutenberg.org/files/20652/20652-h/images/market2_l.gif)  

To market, to market, to buy a fat Hog;  
Home again, home again, jiggety-jog.  
![](http://www.gutenberg.org/files/20652/20652-h/images/market1_l.jpg)

有个人去赶集,在集上买了一头大肥猪;  
买完猪后就跟猪一起蹦跶蹦跶的回家了.  


# THERE WAS A MAN  

There was a man, and he had nought,  
And robbers came to rob him;  
![](http://www.gutenberg.org/files/20652/20652-h/images/man2_l.gif)  

He crept up to the chimney-pot,  
And then they thought they had him;  
But he got down on t'other side,  
And then they could not find him;  
![](http://www.gutenberg.org/files/20652/20652-h/images/man3_l.jpg)  

He ran fourteen miles in fifteen days,  
And never looked behind him.  

有个一无所有的人,被一群劫匪抢劫;  
这个人爬到了烟囱上,然后偷偷的跑掉了;  


# THE LION AND THE UNICORN  


The Lion and the Unicorn  
Were fighting for the crown;  
![](http://www.gutenberg.org/files/20652/20652-h/images/lion2_l.gif)  

The Lion beat the unicorn  
All round about the town.  
![](http://www.gutenberg.org/files/20652/20652-h/images/lion3_l.gif)

Some gave them white bread,  
And some gave them brown;  
Some gave them plum-cake,  
And sent them out of town.  
![](http://www.gutenberg.org/files/20652/20652-h/images/lion4_l.jpg)  

狮子和独角兽争夺王冠,狮子胜了,然后他们开始在小镇里晃荡;
有些人给了他们白面包,有些人给了他们棕面包,还有些人给了他们李子蛋糕;
然后就把他们请出城了.
![](http://www.gutenberg.org/files/20652/20652-h/images/lion1_l.jpg)

And sat down beside her,  
And frightened Miss Muffet away.

Muffet小姐坐在一个小土堆上吃奶油点心的时候,突然跑出来一个大蜘蛛,把她吓跑了.  


# ORANGES AND LEMONS  

![](http://www.gutenberg.org/files/20652/20652-h/images/oranges1.gif)  

_Gay go up, and gay go down  
To ring the bells of London Town._  
_伦敦城突然响起了一阵欢乐的铃铛声._

Bull's eyes and targets  
Say the bells of St. Marg'ret's.  
公牛的眼睛和靶子说这是圣玛格丽特(教堂,城区等地点?下同)的铃铛.  

Brickbats and tiles,  
Say the bells of St. Giles'.  
碎砖和瓦片则说这是圣吉尔斯的铃铛.  

Pancakes and fritters,  
Say the bells of St. Peter's.  
煎饼和油条则说这是圣彼得的铃铛.  

Two sticks and an apple,  
Say the bells at Whitechapel.  
棍子和苹果则说这是怀特查佩尔的铃铛.  

Halfpence and farthings,  
Say the bells of St. Martin's.  
便士和法新则说这是圣马丁的铃铛.  

Oranges and Lemons,  
Say the bells of St. Clement's.  
橘子和柠檬则说这是圣克莱门特的铃铛.  

Old Father Baldpate,  
Say the slow bells at Aldgate.  
一个秃头的老先生则说这是阿尔德盖的铃铛.  

Pokers and tongs,  
Say the bells of St. John's.  
拨火棍和钳子则说这是圣约翰的铃铛.  

Kettles and pans,  
Say the bells of St. Ann's.  
水壶和平底锅则说这是圣安的铃铛.  

You owe me ten shillings,  
Say the bells at St. Helen's.  
圣海伦的铃铛则在说,你欠我十先令.

When will you pay me?  
Say the bells at Old Bailey.  
老贝利的铃铛则在说,你什么时候付给我钱?  

When I grow rich,  
Say the bells at Shoreditch.  
肖尔迪奇的铃铛则在说,当我变得有钱时.  

Pray when will that be?  
Say the bells of Stepney.  
斯蒂芬的铃铛则在说,你要让我盼到啥时候?

I am sure I don't know,  
Says the great bell of Bow.  
鲍尔的大铃铛则在说,我确实不知道.

_Here comes a candle to light you to bed,  
And here comes a chopper to chop off your head._  
_小宝贝,天黑了,快点睡觉吧,再不睡觉就会有人来砍你的头哦!_


# GOOSEY, GOOSEY GANDER  

Goosey, Goosey Gander,  
Where shall I wander?  
![](http://www.gutenberg.org/files/20652/20652-h/images/goose1_l.jpg)  

Upstairs, downstairs,  
And in my lady's chamber.  
![](http://www.gutenberg.org/files/20652/20652-h/images/goose2_l.gif)

There I met an old man  
That would not say his prayers:  
I took him by the left leg,  
And threw him downstairs.  
![](http://www.gutenberg.org/files/20652/20652-h/images/goose4_l.jpg)

一个大白鹅,在她女主人的房子里窜过来窜过去;
突然看到一个老男人,然后就揪着他的腿把他扔下了楼.
