Stages in the Transformations of a **Gnat/蚊**(Culex).      
![Frontispiece](http://www.gutenberg.org/files/16410/16410-h/images/frontispiece.jpg)    
A. Larva. (The head is directed downwards and the tail-**siphon/虹吸管** with **spiracle/气门** points upwards to the surface of the water.)    
B. **Pupal Cuticle/蛹皮** from which the **Imago/成虫** is emerging. (The pair of '**respiratory trumpets/呼吸喇叭**' on the **thorax/胸腔** of the **pupa/蛹** are **conspicuous/明显的**. The wings of the Imago are crumpled, and the **hind feet/后足** are not yet withdrawn.)    
C. Adult Gnat. Female.    
Magnified 5 times.
