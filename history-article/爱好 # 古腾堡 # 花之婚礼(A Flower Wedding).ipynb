A Flower Wedding        
Described By Two Wallflowers        

Decorated By Walter⋅Crane        

Cassell⋅&⋅Company⋅1905        
********************************        

**小天使敲风铃草**        
Yes, flower bells rang right merry that day,        
是的,那一天,喇叭花奏响了欢乐(的乐章),        
When there was a marriage of flowers, they say.        
他们说,这儿举行了一场花之婚礼.        
![**cherubs** rining **bluebells**](http://www.gutenberg.org/files/53619/53619-h/images/i-004.jpg)        
********************************        

**求婚**        
Young LAD'S LOVE had courted Miss Meadow⋅Sweet,        
年轻的小伙子向Meadow⋅Sweet小姐求了婚.        
And the two soon agreed at the Altar to meet.        
他们即将相约在圣坛举行婚礼.        
![The proposal](http://www.gutenberg.org/files/53619/53619-h/images/i-005.jpg)        
********************************        

**举行婚礼**        
A LILY white robe was worn by the Bride,        
新娘穿着白色的婚纱,        
And SWEET WILLIAM, the Groom, **drest** in red, at her side.        
新郎(威廉)则身穿红衣,伴在她的旁边.        
![the **processional**](http://www.gutenberg.org/files/53619/53619-h/images/i-008.jpg)        
********************************        

**三位侍女托着婚纱**        
Miss Violet, Primrose, and gay Marygold,        
Violet小姐,Primrose小姐,以及欢乐的Marygold小姐,        
With their Ladies' Fingers her train did **uphold**.        
用芊芊细手托着新娘的婚纱.        
![her train and three attendants](http://www.gutenberg.org/files/53619/53619-h/images/i-009.jpg)        
********************************        

**四个身穿蓝衣的伴娘**        
In Ladysmocks, Bridesmaids, Forget⋅me⋅not blue,        
伴娘们身穿蓝色勿忘我的罩衫,        
With their **sashes** all tied in Love⋅knot⋅true.        
腰间系着真爱结.        
![Four brridesmaids in blue](http://www.gutenberg.org/files/53619/53619-h/images/i-010.jpg)        
********************************        

**新娘的母亲和朋友**        
The Bride's Mother follows with loving Eyebright,        
忠诚的Eyebright伴着新娘的母亲,        
All in Winter Green and fine Furze **bedight**.        
她们都身穿冬绿色的衣服,佩戴着精致的装饰.        
![the brides's mother and friend](http://www.gutenberg.org/files/53619/53619-h/images/i-011.jpg)        
********************************        

**留着胡子的父亲**        
Whilst her father looked young, though with Old⋅man's⋅beard.        
新娘的父亲尽管留着老人胡,但是看起来仍然很年轻.        
![the father with a beard](http://www.gutenberg.org/files/53619/53619-h/images/i-012.jpg)        
********************************        

**父亲曾是一位花花公子**        
Was a Dande⋅lion in youth I have heard.        
我听说,他年轻时曾是一位花花公子(蒲公英).        
![the father as a **dandy**](http://www.gutenberg.org/files/53619/53619-h/images/i-013.jpg)        
********************************        

**牧师**        
The **troth** was **plighted** for **woe** or for **weal**,        
无论苦楚与幸福,执子之手便是偕老.        
And the lines **attested** by Solomon's Seal:        
该誓言将由所罗门的封印所检验.        
![The minister](http://www.gutenberg.org/files/53619/53619-h/images/i-016.jpg)        
********************************        

**新郎被绑**        
The Bachelor's Button was cast aside,        
自此,新郎将不再是单身汉了(单身汉的纽扣将被丢弃).        
![the groom is tied](http://www.gutenberg.org/files/53619/53619-h/images/i-017.jpg)        
********************************        

**女王与王冠,盾牌和权杖**        
And the **throng** that witnessed was London's Pride:        
见证婚礼的人群是伦敦城的骄傲:        
![queen with crown, shield and **scepter**](http://www.gutenberg.org/files/53619/53619-h/images/i-020.jpg)        
********************************        

**伟大的国王**        
There was Good King Henry, a tall Jonquil,        
这位是好王亨利,一个高个的Jonquil.        
![Large king](http://www.gutenberg.org/files/53619/53619-h/images/i-021.jpg)        
********************************        

**水仙**        
Like Narcissus himself by the waters  still;        
他的身姿如同静水中的水仙;        
![Narcissus](http://www.gutenberg.org/files/53619/53619-h/images/i-024.jpg)        
********************************        

**先生和女士跳舞**        
There were Lords & Ladies to grace the dance,        
这两位是跳舞的先生和女士,        
![Lord and lady dancing](http://www.gutenberg.org/files/53619/53619-h/images/i-025.jpg)        
********************************        

**喜欢迭迭香的女人**        
And Rose Mary, and—        
这位是玫瑰Mary,然后—        
![woman with fan of **rosemary**](http://www.gutenberg.org/files/53619/53619-h/images/i-028.jpg)        
********************************        

**身穿白色花裙的女人**        
Rose⋅la⋅France:        
这是玫瑰France:        
![woman in white flowered dress](http://www.gutenberg.org/files/53619/53619-h/images/i-029.jpg)        
********************************        

**带着金帽子,拿着金剑和权杖的男人**        
With his Golden Rod        
拿着金杖走来的这位        
![man with gold hat, gold sword and **sceptre**](http://www.gutenberg.org/files/53619/53619-h/images/i-032.jpg)        
********************************        

**苏丹和两名服务员**        
the Sweet **Sultan** came;        
是苏丹Sweet;        
![Sultan and two attendants](http://www.gutenberg.org/files/53619/53619-h/images/i-033.jpg)        
********************************        

**弯腰的老太太**        
Lastly, Creeping Jenny, an elderly dame.        
最后这位步履缓慢的老夫人是珍妮.        
![Old lady, bent over](http://www.gutenberg.org/files/53619/53619-h/images/i-036.jpg)        
********************************        

**两名拿着盘子的男人**        
To order the feast—there was Ling, and Hartstongue,        
为了提供盛宴—这儿有Ling和Hartstongue,        
![two men carrying platters](http://www.gutenberg.org/files/53619/53619-h/images/i-037.jpg)        
********************************        

**厨师,两名女佣和一只大鹅**        
And Goosefoot with Sage, the House⋅leek among        
还有带鼠尾草的Goosefoot和House⋅leek        
![Cook, two maids and an extremely large goose walking](http://www.gutenberg.org/files/53619/53619-h/images/i-040.jpg)        
********************************        

**端着豌豆的女人与另一个女人**        
Very Sweet Peas, & Good Cherry Pie,        
以及甜豆和樱桃派,        
Such a feast as an Alderman could not deny!        
恐怕总督都无法抗拒如此盛宴!        
![woman carrying bowl of peas with another woman](http://www.gutenberg.org/files/53619/53619-h/images/i-041.jpg)        
********************************        

**夫妇敬酒**        
In lovely King⋅Cups there was Chamomile Tea.        
在King⋅Cups中盛着甘菊茶.        
![a couple toasting](http://www.gutenberg.org/files/53619/53619-h/images/i-044.jpg)        
********************************        

**两个有钱的客人**        
And the fortune in gifts was a wonder to see!        
这些珍贵的婚礼礼物也是难得一见!        
A new Penny-Royal, A fine Golden Feather;        
这是皇家金币和金羽毛;        
![two rich guests](http://www.gutenberg.org/files/53619/53619-h/images/i-045.jpg)        
********************************        

**两匹马和一个男人站在前面**        
A pair of Horse-chestnuts,        
这是一对Horse-chestnuts,        
![Two horses with a man in front](http://www.gutenberg.org/files/53619/53619-h/images/i-048.jpg)        
********************************        

**一个粉色的天使背着雅各的梯子**        
a Jacob's Ladder,        
这是雅各的梯子,        
![A pink angel carrying two slocks of Jacob's Ladder](http://www.gutenberg.org/files/53619/53619-h/images/i-049.jpg)        
********************************        

**维纳斯捧着像镜子般的蓝花**        
Venus's⋅looking⋅glass,        
这是维纳斯的镜子,        
![Venus holding blue flower like a **looking-glass**](http://www.gutenberg.org/files/53619/53619-h/images/i-052.jpg)        
********************************        

**天使丘比特**        
a fine Arrow-head        
这是一个精致的箭头,        
Discovered long since in the river's bed;        
是很久以前在河床上发现的;        
![angel or cupid taking arrow from a neired](http://www.gutenberg.org/files/53619/53619-h/images/i-053.jpg)        
********************************        

**携带面料的女人**        
Garments of Flax,        
这是亚麻布,        
![woman carrying fabric](http://www.gutenberg.org/files/53619/53619-h/images/i-056.jpg)        
********************************        

**展示垫子的女人**        
and a Lady's Cushion;        
这是一个女士用靠垫;        
![a woman presenting a cushion](http://www.gutenberg.org/files/53619/53619-h/images/i-057.jpg)        
********************************        

**两位女士穿上软管**        
Hose⋅in⋅hose, Lady's slippers to put on,        
这是女士用的软管拖鞋,        
![two ladies puttin on hose](http://www.gutenberg.org/files/53619/53619-h/images/i-060.jpg)        
********************************        

**一位女士拿着金盘子和金酒杯,另一位女士在肩上端着水罐**        
Buttercups gold, and a Pitcher-plant,        
这是金色的Buttercups和一个Pitcher-plant,        
Nay, everything that a house could want.        
不但如此,凡是家之所需,应有尽有.        
![one lady with **platter** and golden **goblets**, another with **pitcher** on her shoulder](http://www.gutenberg.org/files/53619/53619-h/images/i-061.jpg)        
********************************        

**新婚夫妇乘着马车离开**        
In Venus's-fly-trap the pair drove away,        
新婚夫妇乘着维纳斯的飞车离开,        
![couple leaving in a trap or carriage](http://www.gutenberg.org/files/53619/53619-h/images/i-064.jpg)        
********************************        

**朋友们挥手告别**        
"Speedwell, and be happy," their friends gaily say;        
他们的朋友高兴的说着一路顺风;        
![friends waving goodbye](http://www.gutenberg.org/files/53619/53619-h/images/i-065.jpg)        
********************************        

**女佣发现新郎落下的箱子**        
But **alack**! what a **hubbub** when one chanced to find        
啊呀! 侍女突然发现,        
The Bride's only Box was there left behind!        
新郎唯一的行李箱被落下了!        
![maid finding the left behind chest](http://www.gutenberg.org/files/53619/53619-h/images/i-068.jpg)        
********************************        

**小丑与死亡共舞**        
The Wild-**thyme** they had, and the fuss that was made        
迭迭香和大惊小怪,        
Kept the guests in a **rout** thro' the Deadly night shade.        
使客人们在漆黑的夜晚里发生一阵骚乱.        
![Harlequin dancing with death](http://www.gutenberg.org/files/53619/53619-h/images/i-069.jpg)        
********************************        

**客人们纷纷离开**        
But the CLOCKS ticked **apace** to the ope of Daisy        
时光飞快的流逝,大家都走了.        
Snowflakes were fast falling when all said good-bye,        
雏菊花开了,雪花也落下了.        
![Guests taking leave](http://www.gutenberg.org/files/53619/53619-h/images/i-072.jpg)        
********************************        

**新郎站在一堆金币旁,一个手里拿着铲子,一个手里拿着秤**        
With regrets for "that box,"—yet they need not to **stint**,        
尽管他们对落下的那个箱子感到遗憾,但是他们不必节俭,        
For the Bridegroom was rich, he'd a post at the Mint.        
因为新郎是个有钱人,他在造币厂有一个隔间.        
![groom standing over pile of gold coins with a scoop in one hand and scales in the other](http://www.gutenberg.org/files/53619/53619-h/images/i-073.jpg)
