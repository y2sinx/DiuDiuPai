## The Owl & The Pussy-Cat  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_01.jpg)
The Owl and the Pussy-Cat went to sea in a beautiful pea-green boat.  
They took some honey, and plenty of money wrapped up in a five-pound note.  

猫头鹰和猫咪乘着一条豆绿色的美丽小船出海了.  
他们带了一些蜂蜜,以及被五磅纸币包裹着的很多钱.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_02.jpg)
The Owl looked up to the stars above,and sang to a small guitar,  
"O lovely Pussy! O Pussy, my love.  
What a beautiful Pussy you are!  
You are,you are! What a beautiful Pussy you are!"  

猫头鹰仰望着夜空的星星,并弹着小吉他唱道:  
"噢,可爱的猫咪!噢,猫咪,我的爱人.  
你是一只多么美丽的猫咪啊!  
你是,你是!你是最美丽的猫咪!"  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_03.jpg)
Pussy said to the Owl,  
"You elegant fowl! How charmingly sweet you sing!  
O let us be married, too long we have tarried.  
But what shall we do for a ring?"  

猫咪对猫头鹰说:  
"您这只优雅的小鸟!你唱的是如此的甜蜜!  
哦,让我们结婚吧!我们已经耽搁的太久了.  
但是,我们应该(到哪)去找一个戒指呢?"  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_04.jpg)
They sailed away for a year and a day, to the land where the Bong-tree grows.  
And there in a wood a Piggy-wig stood, with a ring at the end of his nose.  
His nose,his nose,with a ring at the end of his nose.  

他们航行了一年零一天,来到了奉树生长着的土地.  
在一个树林里,站着一位带着假发的猪,在他的鼻端挂着一个戒指.  
看他的鼻子,看他的鼻子,他的鼻端挂着一个戒指.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_05.jpg)
"Dear Pig, are you willing to sell your ring for one shilling?" Pussy said to the Pig.  
Said the Piggy, "I will."  

猫咪问:"亲爱的猪,我可以用一先令买你的戒指么?"  
猪说,"可以."  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_06.jpg)
So they took it away, and were married next day by the Turkey who lives on the hill.  
They dined on mince, and slices of quince which they ate with a runcible spoon.  

他们把戒指带走了,并在第二天完成了由居住在山上的火鸡主持的婚礼.  
他们用三齿勺吃了一些肉碎和榅桲(wēn po)片.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_07.jpg)
And hand in hand, on the edge of the sand, they danced by the light of the moon.  
The moon, the moon, they danced by the light of the moon.  
他们手牵着手,在月色的沙滩上,跳起了舞.  
在月色中,在月色中,他们跳起了舞.  

## The Duck & The Kangaroo.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_08.jpg)
Said the Duck to the Kangaroo,  
"Good gracious! how you hop over the fields and the water too, as if you never would stop!  

鸭子对袋鼠说:  
"你在田野和水洼边跳的好优美啊! 就好像你永远都不会停止似的!  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_09.jpg)
My life is a bore in this nasty pond, and I long to go out in the world beyond!  
I wish I could hop like you!"  

我这一生都在这个肮脏的小池塘里,我的生活好无聊啊,我好想去外面的世界看看呐!  
我真希望我能像你一样跳跃!"  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_10.jpg)
"Please give me a ride on your back!  
I would sit quite still, and say nothing but 'Quack'.
The whole of the long day through!  
"求求你,让我在你的背上搭个车吧!  
我会坐在你的背上一动不动,除了说一个'嘎'啥也不说.  
漫长的一天快要结束了(与下面关于月光的一句呼应??).

![](http://www.gutenberg.org/files/20113/20113-h/images/image_11.jpg)
And we'd go to the Dee, and the Jelly Bo Lee, over the land, and over the sea.  
Please take me a ride! O do!"  
Said the Duck to the Kangaroo.  

我们跨过陆地和海洋,去看"迪"和"杰利-波-利".  
求求你让我搭一下,求求你!"  
鸭子对袋鼠说.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_12.jpg)
Said the Kangaroo to the Duck,  
"This requires some little reflection;  
Perhaps on the whole it might bring me luck.  
And there seems but one objection, which is, if you'll let me speak so bold, your feet are unpleasantly wet and cold, and would probably give me the roo matiz."  
袋鼠对鸭子说:  
"我需要考虑一小下.  
或许在整体上这件事可能会给我带来好运,  
但是仍然有一个拒绝的理由,如果你允许我这么大胆地说,那就是,你的脚又湿又冷,可能会让我感到很不舒服."  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_13.jpg)
Said the Duck,  
"As I sate on the rocks, I have thought over that completely, and I bought four pairs of worsted socks which fit my web-feet neatly.  
鸭子说:  
"当我坐在岩石上时,我就已经完全考虑过这件事了.  
我买了四双精纺的袜子,正好适合我的脚.  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_14.jpg)
And to keep out the cold I've bought a cloak, and every day a cigar I'll smoke. 
All to follow my own dear true love of a Kangaroo!"  
为了防寒,我还买了一件斗篷,并且每天我都会抽一支雪茄.
其他所有的事都遵从我亲爱而真诚的袋鼠朋友!"  

![](http://www.gutenberg.org/files/20113/20113-h/images/image_15.jpg)
Said the Kangaroo,  
"I'm ready! 
All in the moonlight pale;  
But to balance me well, dear Duck, sit steady!  
And quite at the end of my tail!"  

袋鼠说:  
"我准备好了!   
世间的万物都披上了银白色的月光(与上面关于白天结束的一句呼应??).  
但是要平衡好我,亲爱的鸭子,请在我的尾巴端上坐好!"  

So away they went with a hop and a bound, and they hopped the hole world three times round;  
And who so happy, —O who, as the Duck and the Kangaroo?  

他们一蹦一跳的走了, 他们围着整个世界跳了三圈.  
还有谁能像鸭子和袋鼠一样那么开心呢?
