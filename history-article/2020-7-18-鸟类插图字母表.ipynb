**Title: The Illustrated Alphabet of Birds        
Author: Unknown        
Origin: Gutenberg**    
    
![](http://www.gutenberg.org/files/11073/11073-h/images/titlebird.jpg)    
    
****************************************************************    
    
![A](http://www.gutenberg.org/files/11073/11073-h/images/auk.jpg)    
    
A is an Auk,    
Of the Artic sea,    
He lives on the ice,    
Where the winds blow free.    
    
A是一只生活在北冰洋之上的海雀,        
那里的风整日肆无忌惮的刮着.    
    
****************************************************************    
    
![B](http://www.gutenberg.org/files/11073/11073-h/images/bluebird.jpg)    
    
B is a Blue Bird.    
In early spring,    
How sweet his songs    
Through the forest ring.    
    
B是一只蓝色的鸟.    
他们喜欢在早春的时候歌唱,这歌声会穿透森林.    
    
****************************************************************    
    
![C](http://www.gutenberg.org/files/11073/11073-h/images/condor.jpg)    
    
C is a Condor,    
On the Andes' height,    
He plumes his wings    
For a lofty flight.    
    
C是一只飞越安第斯山脉的神鹰,    
他扑棱着翅膀,只为寻找一个更高的地方.    
    
****************************************************************    
    
![D](http://www.gutenberg.org/files/11073/11073-h/images/duck.jpg)    
    
D is a Duck.    
Of the canvas back sort,    
To shoot at a flock    
Is considered fine sport.    
    
D是一种帆布背野鸭,很容易遭受人们的猎杀,    
因为这些人认为这是一场精彩的活动.    
    
****************************************************************    
    
![E](http://www.gutenberg.org/files/11073/11073-h/images/eagle.jpg)    
    
    
E's a Bald Eagle,    
So bold and so free;    
On the flag of our country    
He spans land and sea.    
    
E是一只自由且勇敢的秃头鹰,    
在我国的领土上,飞越陆地和海洋.    
    
****************************************************************    
    
![F](http://www.gutenberg.org/files/11073/11073-h/images/fishhawk.jpg)    
    
F is a Fish Hawk,    
Who lives on the shore:    
He catches his prey    
Mid the ocean's roar.    
    
F是一只生活在海岸边的鱼鹰,    
他在咆哮的海洋中,捕捉他的猎物.    
    
****************************************************************    
    
![G](http://www.gutenberg.org/files/11073/11073-h/images/goose.jpg)    
    
G is a Goose;    
His feathers we take    
And put them in sacking    
Our beds to make.    
    
G是一只鹅,    
他的羽毛经常被人们装进布袋,做成床.    
    
****************************************************************    
    
![H](http://www.gutenberg.org/files/11073/11073-h/images/hummingbird.jpg)    
    
H is a Humming Bird,    
Sporting mid flowers    
And brightly enjoying    
The sunny hours.    
    
H是一只在花丛中活动的蜂鸟,    
他们非常喜欢阳光明媚的时光.    
    
****************************************************************    
    
![I](http://www.gutenberg.org/files/11073/11073-h/images/ibis.jpg)    
    
I is an Ibis,    
Who wanders in bogs,    
And lives upon lizards,    
And fishes and frogs.    
    
I是一只在沼泽中漫步的朱鹭,    
他们以蜥蜴,鱼和青蛙为食.    
    
****************************************************************    
    
![J](http://www.gutenberg.org/files/11073/11073-h/images/jay.jpg)    
    
J is a Jay,    
With his blue and white coat,    
With a crest on his head,    
And a ring round his throat.    
    
J是一只松鸦,身披蓝白羽毛,    
他们的头顶有个鸟冠,喉咙附近有个小圈.    
    
****************************************************************    
    
![K](http://www.gutenberg.org/files/11073/11073-h/images/kingbird.jpg)    
    
K is a King Bird,    
Pugnacious and bold:    
A hero in fight,    
And a terrible scold.    
    
K是一只国王鸟,脾气暴躁而大胆,    
他是战斗中的英雄,也是喜欢招惹是非的坏怂(huài sǒng).    
    
****************************************************************    
    
![L](http://www.gutenberg.org/files/11073/11073-h/images/lark.jpg)    
    
L is a Lark,    
A sociable bird;    
His song in the meadow    
Is frequently heard.    
    
L是一只善于交际的百灵鸟,    
在草地上,可以经常听到他们的歌声.    
    
****************************************************************    
    
![M](http://www.gutenberg.org/files/11073/11073-h/images/magpie.jpg)    
    
M is a Magpie,    
He lives at the west,    
Steals and scolds and eats carrion;    
He's none of the best.    
    
M是一只住在西部的喜鹊,  
这种鸟不太友好,他喜欢招惹是非,偷东西,吃腐肉.
   
****************************************************************    
    
![N](http://www.gutenberg.org/files/11073/11073-h/images/nightheron.jpg)    
    
N is a Night Heron,    
Of fishes quite fond;    
He looks for them now,    
As he stands by the pond.    
    
N是一只喜欢鱼的夜鹭,    
他现在正站在池塘边找鱼吃.    
    
****************************************************************    
    
![O](http://www.gutenberg.org/files/11073/11073-h/images/owl.jpg)    
    
O is an Owl,    
Who hides through the day;    
And comes out at night,    
To seek for his prey.    
    
O是一只猫头鹰,他们会在白天的时候躲起来,    
只有在晚上的时候出来寻找他的猎物.    
    
****************************************************************    
    
![P](http://www.gutenberg.org/files/11073/11073-h/images/pigeon.jpg)    
    
P is a Pigeon    
So rapid in flight,    
That before you can shoot him    
He's gone out of sight.    
    
P是一只鸽子,可以飞的很快,  
所以在你准备开枪打他的时候,他已经跑的不见踪影.    
    
****************************************************************    
    
![Q](http://www.gutenberg.org/files/11073/11073-h/images/quail.jpg)    
    
Q is a Quail,    
Who hides in a tree,    
And whistles "Bob-White"    
With lively glee.    
    
Q是一只躲在树上的鹌鹑(ān chún),  
欢快的叫着"鲍勃⋅怀特",就像在吹口哨.    
    
****************************************************************    
    
![R](http://www.gutenberg.org/files/11073/11073-h/images/robin.jpg)    
    
R is the Robin,    
So kind and so good,    
Who covered with leaves,    
The poor babes in the wood.    
    
R是一只非常善良的知更鸟,    
经常被叶子覆盖,是树林中可怜的小东西.    
    
****************************************************************    
    
![S](http://www.gutenberg.org/files/11073/11073-h/images/swallow.jpg)    
    
S is the Swallow,    
She darts through the air    
To catch little insects,    
Her favorite fare.    
    
S是一只燕子,喜欢在空中猛冲,    
这是为了捕捉小昆虫,她们最喜欢的食物.    

****************************************************************    
    
![T](http://www.gutenberg.org/files/11073/11073-h/images/turkey.jpg)    
    
T is a Turkey,    
A fine dashing beau,    
By his fuming and strutting,    
His pride you may know.    
    
T是火鸡,一种非常华丽的鸟,    
从他们的怒气冲冲和昂首阔步,可以看出他们很骄傲.    
    
****************************************************************    
    
![U](http://www.gutenberg.org/files/11073/11073-h/images/upupa.jpg)    
    
U's the Upupa,    
Or Hoopoe. His crest    
He can raise or lower,    
As suits him best.    
    
U是戴胜鸟,他们的鸟冠可以随意的升起或落下,    
直到最适合他们.    
    
****************************************************************    
    
![V](http://www.gutenberg.org/files/11073/11073-h/images/vulture.jpg)    
    
V is a Vulture,    
Who feeds on the dead,    
When the dark battle-field,    
With corpses is spread.    
    
V是一只秃鹰,    
在尸横遍野的战后之场上,以死者为食,        

****************************************************************    

![W](http://www.gutenberg.org/files/11073/11073-h/images/woodpecker.jpg)    
    
W's a Woodpecker,    
Who with his long bill,    
Bores holes in a tree    
And of worms eats his fill.    
    
W是一只啄木鸟,他的鸟喙(niǎo huì)很长,    
他喜欢在树上钻孔,然后吃掉里面的蠕虫.    
    
****************************************************************    
    
![X](http://www.gutenberg.org/files/11073/11073-h/images/xanthoronus.jpg)    
    
X is Xanthoronus,    
Or Baltimore Bird.    
Oft in our orchards,    
His music is heard.    
    
X是巴尔的摩黄鹂,    
在我们的果园里可以经常听到他们的歌唱.    
    
****************************************************************    
    
![Y](http://www.gutenberg.org/files/11073/11073-h/images/yellowbird.jpg)    
    
Y is a Yellow Bird,    
With feathers so bright,    
Who sings all the day,    
And sleeps all the night.    
    
Y是一只金莺,他的羽毛非常的靓丽,    
他们在白天唱歌,晚上睡觉.    
    
****************************************************************    
    
![Z](http://www.gutenberg.org/files/11073/11073-h/images/zroost.jpg)    
    
Z stands for none    
of the feathered race.    
It must serve as a roost,    
Or lose the last place.    
    
Z不是指鸟而是指他们的巢穴,
否则他们将失去最后的栖身之所.
