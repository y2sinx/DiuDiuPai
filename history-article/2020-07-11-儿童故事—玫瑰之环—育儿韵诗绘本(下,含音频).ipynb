# GOOD KING ARTHUR  

When good king Arthur ruled this land,  
He was a goodly king;  
He stole three pecks of barley-meal,  
To make a bag-pudding.  

![](http://www.gutenberg.org/files/20652/20652-h/images/arthur1_l.gif)

A bag-pudding the king did make,  
And stuffed it well with plums;  
And in it put great lumps of fat,  
As big as my two thumbs.  

![](http://www.gutenberg.org/files/20652/20652-h/images/arthur2_l.jpg)

The king and queen did eat thereof,  
And noblemen beside;  

![](http://www.gutenberg.org/files/20652/20652-h/images/arthur3_l.jpg)

And what they could not eat that night,  
The queen next morning fried.  

![](http://www.gutenberg.org/files/20652/20652-h/images/arthur4_l.gif)  

当有个好国王亚瑟统治这片土地的时候,它偷了三大袋大麦粉,想用它做布丁.  
他做的这个布丁不仅塞上了李子,还塞了些大肥肉.  
国王和王后都吃了,而贵族们则在旁白站着.  
那晚他们吃剩下的,第二天早晨被王后炸着吃了.

# COCK-A-DOODLE-DOO  

Cock-a-doodle-doo!  
My dame has lost her shoe;  
My master's lost his fiddling-stick,  
And don't know what to do.  

![](http://www.gutenberg.org/files/20652/20652-h/images/cock1_l.gif)  

Cock-a-doodle-doo!  
What is my dame to do?  
Till master finds his fiddling-stick,  
She'll dance without her shoe.  

Cock-a-doodle-doo!  
My dame has lost her shoe,  
And master's found his fiddling-stick;  
Sing doodle-doodle-doo!  

![](http://www.gutenberg.org/files/20652/20652-h/images/cock2_l.jpg)

Cock-a-doodle-doo!  
My dame will dance with you,  
While master fiddles his fiddling-stick,  
For dame and doodle-doo.  

Cock-a-doodle-doo!  
Dame has lost her shoe;  
Gone to bed and scratched her head,  
And can't tell what to do.  

![](http://www.gutenberg.org/files/20652/20652-h/images/cock3_l.gif)
