> 1. 摘要:   
Crocodile, Tortoise, Penguin and Chicken find an egg. Is it their egg? 'Our egg,' say the proud Mum and Dad at the end. Discover whose egg it is.  
鳄鱼,乌龟,企鹅和鸡发现了一颗蛋. 这是他们的蛋吗? 最后一对自豪的家长说: "这是我们的蛋.". 让我们一起来发现这是谁的蛋吧!
