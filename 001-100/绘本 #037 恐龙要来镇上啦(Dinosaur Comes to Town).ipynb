### 恐龙要来镇上啦    

By Gene Darby    
Illustrated by Art Seiden    

Whitman Publishing Company    

---

![02](https://ae02.alicdn.com/kf/H069be728d4f3441fa535d51c5b6e8ab1f.jpg)    

---

远处传来了一只食肉(meat-eating)恐龙的沉重脚步声, 砰(thump)~ 砰~ 砰~    

![03](https://sc02.alicdn.com/kf/H40ea010850354b6d80baa174ba5b073cZ.jpg)    

---

兔子看到了那个大家伙, 然后挖了一条通向花栗鼠(chipmunk)家的路, 嘴里还在大喊着(bellow): "花栗鼠, 花栗鼠, 那边来了一只大恐龙, 我们赶快逃命去吧!"    

![04](https://ae03.alicdn.com/kf/Ha8ba204239344f4a844f36d5f51a91861.jpg)    

---

花栗鼠在前面跑着, 兔子在他身后跟着, 但是兔子却在出来的时候被洞口卡住了.    

![05](https://ae02.alicdn.com/kf/H3b0212e624b146f8ba13cd98e5eb007c2.jpg)    

---

无论是兔子先伸出脑袋还是先伸出尾巴, 花栗鼠都无法将兔子给拽出来.    

![06](https://ae04.alicdn.com/kf/H16677469cba4487a86d3c1e83e184d18f.jpg)    

---

突然摇晃着(waddle)来了一只海狸(beaver), 他也想帮着把兔子给拽出来, 但是他跟花栗鼠联手也没能做到.    

![07](https://ae03.alicdn.com/kf/H187cc2c4f7604a5e897ad7fc207adb5fZ.jpg)    

---

突然又来了一只臭鼬(skunk), 他们三人联手, 终于把兔子从洞中给拽了出来.    

![08](https://sc02.alicdn.com/kf/H689c301b98f34eee833e31853f9ec2fem.jpg)    

---

然后他们四人开始一起逃命...    

![09](https://sc02.alicdn.com/kf/H731380a517654617a9e359c586b9fe6fx.jpg)    

---

在他们逃跑的时候, 他们还遇到了一只小鹿和一只狐狸, 然后小鹿和狐狸也加入了他们逃跑的队伍. 快跑啊~ 快跑啊~ 吃肉的大恐龙要来了~    

![10](https://ae04.alicdn.com/kf/Ha5e9b000a8a84a73b53482587d41da79r.jpg)    

---

虽然他们六人都在拼命的奔跑, 但是脚步声却仍然在不断的逼近. 最后, 只听"砰"的一声, 就连大地都跟着颤抖了(the earth shook). 兔子连忙让大家躲了起来, 然后只见——    

![11](https://sc03.alicdn.com/kf/H02415bc6d4b6492ea47f9f221499f611O.jpg)    

---

大恐龙一一从他们身边走过了...    

![12](https://ae04.alicdn.com/kf/H0e481881a04e447cb186f216df9ef1cah.jpg)    

---

大恐龙继续走啊走, 在跨过了山丘之后, 来到了立着"限高牌"(limit sign)城市边缘, 但是限高牌并没有阻止大恐龙继续前行.    

![13](https://ae02.alicdn.com/kf/H5291ec3e5fa7431ba3f5b3eca0afe10dP.jpg)    

---

进了城的大恐龙一一走过了面包店(bakery), 杂货店, 洗衣房(laundry). 就连小汽车, 大巴士, 大卡车都为他停了下来.    

![14](https://sc04.alicdn.com/kf/H7910098427004d669cbd4e153ee594f2k.jpg)    

---

然后, 大恐龙又一一经过了市长家(mayor's house), 警察局, 消防局(fire department). 他沿着城市的主干道(main street)不停地走啊走, 一直走到了一个写着 "Drive In" 的地方才停了下来.    

![15](https://sc04.alicdn.com/kf/Haa1a0b18dfdb49a5abae2d852dc73ba0l.jpg)    

---

然后, 他一屁股坐了下来, 坐在了一根凳子(stool)上, 并向窗旁的女士点了六千万个汉堡(hamburgers). 最后, 他把这些汉堡全都吃掉了.    

![16](https://sc04.alicdn.com/kf/H6a9d2cbe5ccf42f3a2349a95decd3a29I.jpg)    

---

你确定这只大恐龙不是汉堡店请来的托儿~
