## The story of the Grand Canal.

Some 2,500 years ago, in ancient China, an `emperor`(皇帝) had a bright idea. If he joined together two rivers, `the Huai and the Yangtze`(淮河与长江), he could transport his `troops`(军队) quickly through his lands to conquer an enemy kingdom.  

So was built the first stage of a `canal`(运河) that would become one of `the greatest human-made wonders of the world`(世界最伟大的人造奇观). A thousand years later, various rivers and canal `sections`(航段) were combined into one great `waterway`(航道), shipping grain and other goods from Hangzhou in `the fertile south`(肥沃的南方) to Beijing in `the dry north`(干旱的北方).  

`Trade flourished and thriving cities`(贸易促生的城市) grew up along the canal's banks, as China entered a golden age. Life was not always easy, however. Floods and wars brought `disaster and hardship`(灾难和困苦), and eventually the canal fell into decline and neglect. But modern times `have breathed new life into the canal`(给运河注入了新的生命力). Sections have been repaired, and many of its beautiful landmarks `restored`(重建). You’ll see some of them on the pages of this book. Today the waterway stretches 1,115 miles (1,794 km)— the longest, as well as the oldest, canal in the world.  

So come with us and explore this ancient wonder. Our story follows the `fortunes`(命运) of the canal and the people that lived along it, through good times and bad. You’ll meet emperors and soldiers, merchants and mysterious travelers, and children just like you. You’ll find out what people wore, and what they ate, and how they spent their time, from `antiquity`(古代) to the present.

## Construction Begins, Yangzhou, -486.

The king's `advisors`(谋士) are deep in discussion about `military tactics`(军事策略).    

With sword in hand, `Fuchai`(夫差) is dreaming of conquering new lands.    

Fuchai’s companion, `Xi Shi`(西施), eats fruit `laid out by the maids`(由女佣布置的). She is a `spy`(间谍), sent by an `enemy state`(敌国) to distract the king.    

Maids use fans to keep the king cool and `swat away mosquitoes`(赶走蚊子).    

The workers tasked with carrying the soil away have `the most tiring job`(最累人的工作).    

![4.jpg](https://ae04.alicdn.com/kf/Hf29dbab659474327925e1630ed3199578.jpg)    

During `the Spring and Autumn Period`(春秋时期) of `the Eastern Zhou Dynasty`(东周), wars are common between the many small states of ancient China.

`King Fuchai of Wu`(吴国) orders the construction of a canal, the `Han Gou`(邗沟), to transport soldiers and equipment quickly to the north, where he wishes to expand his kingdom. 

It is `backbreaking work`(繁重的工作) for the thousands of peasants ordered to dig `the barren earth`(贫瘠的土地) in `the baking sun`(炎炎的烈日), using only basic tools.    

![5.jpg](https://sc04.alicdn.com/kf/Hc23bfe1614264029b89bb988b657b16dL.jpg)    

Find `the wild-haired shaman`(披头散发的萨满巫师) performing a ceremony to bless the canal's construction.    

Some workers have been captured trying to escape.    

By `swinging a heavy stone in the air, then dropping it quickly`(打夯), the men make `the bed of the canal`(河床) more solid.    

`Losing their tempers`(失去他们的耐心) in the heat, some workers have started to fight.    

Made with `iron`(铁) instead of the traditional `bronze`(青铜), the diggers' tools are strong.    

## A Palace on Water, Luoyang, 605.

All along the route, local people make offerings of food to please the emperor.  

A `mounted`(骑着马的) `guard of honor`(骑兵仪仗队) parades along the shore `in full battle regalia`(全副武装).  

It is exhausting work for the teams of `boat trackers`(纤夫) on the shore, who have to pull the vessels by rope.  

Pride of the fleet, `the battleship's catapults`(战舰石弩) can launch rocks at enemy ships.  

`Guard boats`(护卫艇) powered by `oarsmen`(桨手) `escort`(护送) the emperor's `barge`(驳船).  

![6.jpg](https://ae04.alicdn.com/kf/H4c8471e36fdf49eaa7d5f43e3747ed37b.jpg)  

The construction of the section of the Grand Canal between Yangzhou and Luoyang is finally complete.  

A `fanfare`(迎宾乐) `heralds`(为...欢呼) the arrival of an `imperial flotilla`(帝国舰队), which has come to inspect the new waterway.  

`Emperor Yang of Sui`(隋炀帝) commands the procession from the top deck of the `lavish`(耗资巨大的), dragon-shaped barge, which glides through the water like `a floating palace`(漂浮的宫殿).  

For the people of Luoyang, it is `a grand spectacle`(壮观的场面) `beyond their wildest imaginations`(超乎他们最狂野的想象).  

![7.jpg](https://ae02.alicdn.com/kf/H9350f56afdf7465cac97f9d0b663650dx.jpg)  

The emperor `is flanked by`(被...相伴) his `army officers`(军官). Fearful of his `short temper`(暴躁脾气), they `flatter him`(向他谄媚).  

The dragon is a symbol of the emperor's strength and power.  

A merchant has come `from afar`(来自远方) to trade with the emperor. Unfortunately, he's feeling a little `seasick`(晕船).  

The `phoenix`(凤凰) symbolizes the empress and brings good luck.  

`Empress Xiao`(萧皇后) sails in a smaller, phoenix-shaped barge, with her maids and `court ladies`(宫女).  

## Granary Attack! Liyang, 618.  

The roofs of `the assault wagons`(突击车) are covered with `cow hide`(牛皮) to withstand fire.  

The `lookout`(看守) in the mobile `watchtower`(瞭望塔) is hauled up by `pulley`(被滑轮拖着).  

Weapons are transported in connected wooden wagons, which also provide cover for tunnel diggers.  

The heavy `armor`(装甲) of the rebel army shines a brilliant `blue-black`(深蓝色) when polished.  

With its long spikes(刺), the `"Wolf's-Teeth Crusher" is a lethal weapon`(狼牙棒是一种致命武器).  

![8.jpg](https://ae04.alicdn.com/kf/Hb35e344ab66b4f0c8152606dd0ed0056G.jpg)  

Emperor Yang of Sui is dead, killed by his once-loyal troops, and China is in chaos, at the mercy of powerful `warlords`(军阀).  

The rebel soldiers march east to take control of the capital but soon `run out of rations`(口粮不足).  

Desperate, the hungry troops `lay siege to`(对...展开围攻) the huge `granary store`(粮仓) at Liyang on the Grand Canal, which is occupied by another rebel force, `the Wagang Army`(瓦岗军).  

The attack succeeds, but both armies pay a heavy price.  

After this, the Tang Army will conquer China.  

![9.jpg](https://ae03.alicdn.com/kf/Hcbe3cae081f1434f808011b35a2bbe4db.jpg)  

Wagang soldiers defending the `beacon tower`(烽火台) have lit a fire to alert friendly forces that help is needed.  

The Liyang granary is one of the biggest along the Grand Canal, with grain stores to feed 80,000 people a year.  

The Wagang defenders have begun digging an escape tunnel.  

Outnumbered, the soldiers throw rocks at any invaders their swords cannot reach.  

## A Military Coup, Kaifeng, 960.  

The Yellow River connects the growing city of `Kaifeng`(开封) to the Grand Canal.  

Can you spot the guard whose shield is painted with the face of a tiger?  

Only the strongest guards carry the `modao`(陌刀), a long, heavy sword with a wide steel `blade`(刀片) used to fight the enemy `cavalry`(骑兵).  

Red `tassels`(流苏) on a soldier's helmet indicate that he holds a superior rank.  

The men `huddle around the fire`(围聚在火堆前) to keep warm, singing songs and `toasting bread`(烤馍馍).  

![10.jpg](https://sc02.alicdn.com/kf/H5805a0aebdc244aca3e9ee2ed5e4ce04N.jpg)  

China is without a strong leader and has descended into `civil war`(内战).  

`Zhao Kuangyin`(赵匡胤), chief commander of the `imperial guard`(御林军), is a powerful and popular figure.  

On a winter's night, his loyal guards surround his `quarters`(营房) near Kaifeng, cheering their support.  

Over dinner, his officers `drop to their knees`(跪下来) and declare him emperor.  

Now `Emperor Taizu`(太祖皇), he founds the `Song Dynasty`(宋朝), which unites China, and brings peace and a golden age of culture and science.  

![11.jpg](https://ae03.alicdn.com/kf/H7157bc74846f4cf7a840b45b745ff6d9N.jpg)  

Zhao Kuangyin's secretary dresses him in the `imperial robe`(龙袍), a symbol of power.  

The officers toast the new emperor with wine served in `jade cups`(玉杯).  

The soldiers will `tuck into`(大快朵颐) a `hearty`(丰盛的) meal of `hotpot`(火锅), `dipping meat and vegetables into a simmering broth`(将肉和蔬菜放到热滚滚的汤里涮着吃).  

The `astrologer`(占星家) excitedly points to the night sky. He is persuading the men that the position of the planets indicates that Zhao Kuangyin is the rightful emperor.  

## A Paradise on Earth, Hangzhou, 1185.  

A young woman is `applying her makeup`(涂妆), using powders made from `crushed flowers`(花碎).  

At the `spice shop`(香料店), Arabian merchants gather to discuss business.  

The restaurant diners pay with a new invention: `paper money`(纸币).  

The street cleaner carries away trash in a `raft`(木筏).  

Local silk is famous across China for its `soft texture`(柔软的质地) and `beautiful sheen`(美丽的光泽).  

The market is always full of `gossip`(闲言碎语) about `the royal household`(皇室).  

![12.jpg](https://ae03.alicdn.com/kf/Hdf02f862aed34c03aba337bbbad61489k.jpg)  

At the southern end of the Grand Canal, and surrounded by `lush countryside`(郁郁葱葱的乡村), Hangzhou has long been famous for its beautiful gardens, bridges, and temples.  

Since the arrival of the canal, the city has grown prosperous, too.  

Along the busy `quayside`(码头周围) between the steep `Zhong'an`(众安) and `E'ya`(鹅鸭) bridges, shopkeepers `beckon`(用手召唤) `well-to-do`(富有的) passersby to browse their tempting displays, while `sharp-witted`(机敏的) merchants from `foreign shores`(外国) `strike deals with local traders`(与本地贸易商打交道).  

![13.jpg](https://ae03.alicdn.com/kf/H3fa6e3240d2842b6b2c3c5ac22329316J.jpg)  

Many travelers and `pilgrims`(香客) come to admire the pleasure gardens and visit the temples of `the West Lake`(西湖).  

The poet `Lin Sheng`(林升) writes verses on a shop wall.  

A `dragon boat`(龙舟) crew is practicing for the races on West Lake.  

Find the boy trying to light the `cauldron`(大锅) fire at the `dumpling stall`(卖饺子的摊位).  

`Pleasure boats`(游船) cruise the canal, `packed with`(坐满了) tourists eager to explore the city.  

## Night Fair, Maple Bridge, Suzhou, 760.  

These women are wearing the latest fashions from `India and Persia`(印度和波斯).  

Tears flow as a husband prepares to set sail for the capital, leaving his wife at home.  

`Zhang Ji`(张继), a famous poet, is inspired by the `moonlit scene`(月色) and composes the `lines`(诗句) to a beautiful poem in his head.  

A Japanese diplomat has arrived, on a mission to learn about Chinese culture.  

Can you find a `haughty official`(傲慢的官员), being helped along by his loyal servants?  

![14.jpg](https://sc04.alicdn.com/kf/Hd34511edb9854f1fbd81520904526afe9.jpg)  

It is almost midnight, but the busy port of Suzhou is still `teeming with life`(充满生机). 

In the moonlight, merchants and travelers `disembark from`(从...上岸) passenger boats under the old stone `Fengqiao`(Maple Bridge). 

Business is booming these days, and some bring goods from distant lands such as Persia and `Arabia`(阿拉伯) to trade.

Others are more interested in the `taverns and teahouses`(客栈和茶舍) of `the bustling night market`(熙熙攘攘的夜市).  

![15.jpg](https://ae04.alicdn.com/kf/Ha9f62e5b45984e7a900ce1c49b6405eaJ.jpg)    

Twice a day, monks strike the bell at this famous temple 108 times.  

Suzhou is a `flourishing weaving center`(昌盛的纺织中心). Weavers work together on beautiful `embroidery`(刺绣品).  

A sick passenger is carrried off the boat into the `medical ward`(病房).  

These `wrestlers`(摔跤员) are practicing for a competition.  

A Buddhist monk disembarks, with all his `worldly possessions`(身外之物) in a `bamboo carrier`(竹背筐).  

An Arabian merchant and a local trader discuss a business deal.  

## Stopping the Floods, Huaiyin, 1699.  

The `Gaojia Weir`(高家堰) was the first dam constructed to control the floods.    

The `Kangxi Emperor`(康熙皇帝) has studied the canal for a long time. This is his third tour.    

At the simple temple by the weir, villagers pray to the `deities`(神).    

The emperor is sheltered beneath a  `silk canopy`(丝绸雨篷).    

Branches, `reeds`(芦苇), and `wheat stalks`(麦草秸秆) are tied into bundles to make flood defenses.    

By carving the shape of the landscape on a `wooden plank`(木板), planners make a 3-D map.    

![16.jpg](https://ae04.alicdn.com/kf/Hf7fc7d1338234a10a97f096d312f155cp.jpg)    

The workers need to move fast.    

The rains have begun, and if the flood defenses aren’t in place soon, the Yellow River will burst its banks again.    

Over the years, the canal system has been neglected, and `silting upriver`(淤积的上游) has made the river change course.    

Now it flows dangerously close to `Huaiyin`(淮阴), where the Grand Canal and the Huai River meet, and causes floods that destroy homes and `livelihoods`(生计).    

The river must be tamed, and the Kangxi Emperor has taken command.    

He orders his workers to build a vast series of defenses to change its direction.    

It will be an `engineering masterpiece`(工程杰作).    

![17.jpg](https://ae04.alicdn.com/kf/H2a429fefdc2642daa2fd20c2f7e38cc8h.jpg)    

Can you spot this worker, throwing his hands up in the air in prayer for the rain to stop?    

Grain can no longer reach the capital because `the canal water is too high`(水位太高).    

Workers `haul`(用力拖) the `flood defenses`(防洪防御) into position to prevent the Yellow River from overflowing.    

The workers throw `straw capes`(斗笠) over their shoulders to keep dry in the rain.    

Canal diggers hammer wooden `posts`(木杆) into the ground to compact the earth.    

## A Ridge of Water, Nanwang, 1420.  

When it is finished, all travelers will `burn incense`(烧香) at the new temple for good luck.    

A sailor `winches down the mast of his cargo boat`(把货船的桅杆吊下来) as it prepares to `dock`(靠岸).    

Can you spot a baby, `glimpsing`(瞥) the outside world for the first time?    

A grandmother who is lost asks for directions.    

A street `vendor`(小贩) sells `seed pods of the lotus flower`(莲蓬), a tasty snack.    

A scholar buys `the "four jewels": a brush, ink, ink stone, and paper`(文房四宝).    

![18.jpg](https://sc03.alicdn.com/kf/H005e6cbd7da14d3bb60c291ae4945d11q.jpg)    

`Nanwang`(南旺) marks the canal's highest point.    

Until recently, the canal often ran dry and boats ran aground here. Not any more.    

Clever engineers have figured out a way to keep water levels steady by digging a canal to the `Wenhe River`(汶河), and now cargo boats can navigate this `tricky ridge`(难以跨越的水脊) all year round.    

`Builders`(建筑工人) are constructing a beautiful temple in celebration, dedicated to the `Dragon King`(龙王), a god of water and weather.    

![19.jpg](https://ae04.alicdn.com/kf/H459c85c8b88c487fa01cbad4576117b3n.jpg)    

Water from the river is stored in a `reservoir`(水库). The flow is controlled by a `sluice gate`(闸门).    

Balancing on an unsteady plank, a builder hauls `tiles`(瓦片) up to the roof by pulley.    

This builder `doesn't like heights`(不喜欢高), so he is having a rest.    

Workers are building a statue of the water god.    

A worker mixes up `glue`(建筑胶) made from animal bones to bind together the bricks.    

An official shows the new temple to `Song Li`(宋礼), the minister in charge of construction.    

## A Customs Pass, Linqing, 1600.  

In `the Ming era`(明代时期), opera is very popular and always gathers a crowd.    

The `customs office`(海关办事处) is the richest in China, collecting a quarter of all taxes.    

Find a traveler finishing his breakfast.    

A mother and daughter are discussing which color `cattail leaf fan`(蒲扇) to buy.    

The children are `dueling with leaf stalks`(玩一种拿叶杆拉勾勾的游戏). The loser is left with a broken stalk.    

Can you spot these old friends sharing a picnic by the canal?    

![20.jpg](https://ae03.alicdn.com/kf/Hd504211bcd384a3a9f6098f8ef19176fn.jpg)
