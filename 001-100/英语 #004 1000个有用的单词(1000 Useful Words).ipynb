shampoo: /ʃæmˈpuː/, 发音与"香波"相近, 指日常用的"洗发水"或"沐浴液"之类的东西.  

sunblock: /ˈsʌnblɒk/, 由 sun 和 block(阻挡) 这两个"子单词"构成, 指一个阻挡太阳(光线)的东西, 即"防晒霜". 

tissues: 此处是纸巾的意思. 如果与人体相关的话指"组织(器官)", 与工业相关的话指"纺织品". 

blowing my nose: 从字面上看是"吹吹自己的鼻子"的意思, 而其实是指"捏住鼻子,用气排出鼻涕"这一动作, 在北方方言中这一动作叫做"擤(xǐng)鼻涕".


![8-9.png](https://ae04.alicdn.com/kf/He1ab6c1ac49544c983811f7e49bf95d1l.jpg)

---

tights: 能否像"小学生组词一样的方式"理解英语单词的意思呢? 首先"tight(紧的)"是一个形容词, "tights"中的"+s"表明这个单词是一个"名词(按句子成分的需要)", 那么"tights"自然就是一个"紧紧的东西", 如"紧身衣","紧身裤","紧身袜子"等... 如果句子成分需要一个动词, 你能否将"紧"组词为"缩紧","收紧"呢?

buckle: 有些单词既可以当名词也可以当动词, 如"buckle"既可以指"扣子", 又可以指"系扣子"这一动作. 因此, 在记单词的时候, 我们可以弱化词性, 只记单词的一个名词含义就好啦?
