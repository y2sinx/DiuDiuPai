### 猪猪鲁安妮的完美家庭    

南希·卡尔森    

![02](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autod78a24258fd34a7aa97c4fddd267ba02)    

Louanne(鲁安妮) lived with her mother and her father. George lived next door with his mother and his father and his five sisters and his four brothers. There was never a dull moment(沉闷的时候) in their yard.    

![07](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto264da17c5d05423eb20c112c11731b25)    

"Don't you think ten kids make a perfect family?" Louanne asked her mom.    

"You and Daddy are the perfect family for me," said Mom.    

![08](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto07ba2138870640ea9d0611e1f08c7050)    

"Didn't you ever think about adopting(收养) some more kids?" Louanne asked her dad.    

"Nope," said Dad. "You're all I can handle."    

![09](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto237d250465de4259b8614356a50320a8)    

Then one night Mom said, "My office just called. I have to fly to Dallas(达拉斯) in the morning, and this is Daddy's fishing weekend. How would you like to spend the weekend at George's house?"    

"Would I ever(好哎,好哎)!" said Louanne.    

![10](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto1ecf85484df14849bf4c74b0c85ce35a)    

On Friday afternoon, Louanne packed two nice outfits(打包了两件漂亮的外套), a nightgown(一件睡衣), a good book, her toothbrush and toothpaste(牙刷和牙膏), a towel, and a large bag of jelly beans(软心豆粒糖).    

![11](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autob07e84c7e4e94707b85caca41a97e88b)    

"Off I go!" she told her parents.    

![12](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autof0a410ded199426ab705a9133bcfa566)    

"I'm afraid you'll have to share a bedroom with the girls," George's mother told Louanne.    

"Oh, that will be wonderful," said Louanne. "I've never shared a room before."    

![13](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autoaf6c2fb1bd454f09b7489c80ab38f73c)    

George took Louanne upstairs so she could leave her suitcase(手提箱) in her room. Then they went outside to play.    

![14](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto044765d19e5449c1816366969460c1a3)    

As they ran outside, Louanne slipped on a roller skate("踩"到了轮滑鞋) and fell down(跌倒了).    

![15](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto2b24bbd2f7734c9eb3004d6cd82b639c)    

"You have to watch your step(注意你的脚下) around here," said George. "Hal(哈尔) leaves his stuff all over the place."    

![16](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autoa7779acd3f8b455a923791943fe069d0)    

In the middle of a game of catch(接球游戏), Louanne felt her shirt getting wet.    

"Is it raining?" she asked.    

![17](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto9eb5a25b18b441e48043dd703ed723e8)    

"No, it's Tony," said George. "He got a new squirt gun(水枪) this morning. Just ignore him, and he'll go away."    

![18](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto51db4727703243cc923584218f2899a2)    

Suddenly a loud bell began to clang(一个很大的的铃声响了起来).    

![19](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autob77d3e814d4f45318a30eb1a5beca288)    

"Dinner!" yelled George, and Louanne was left behind in a cloud of dust(一团尘土).    

![20](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto8a2ae5f241184fbe923a878656786e24)    

By the time she made it to the table(当她成功"抵达"餐桌的时候), everyone else was already there.    

"Hey, Margaret(玛格利特), pass the corn(玉米) before you've taken it all!" Tony was yelling.    

"Tony," wailed Harry, "you took the last carrot!"    

![21](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autocf789660c81146cca7e1a5c40decdb06)    

"Pass the squash!" yelled Hal.    

"Please pass the squash," said his mother.    

"ME FIRST!" yelled Hal.    

Dinner at my house is never like this, thought Louanne.    

![22](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autof71ab87989a148fab44534eecd4a3f0f)    

Later Louanne went upstairs to unpack(打开包裹). Her clothes were strewn all over the room(她的衣服在房间里被扔的到处都是), and her bag of jelly beans was nearly empty.    

"It must have been Nancy," said George. "She gets into everything(她正在习惯所有的东西). She's too little to know better."    

![23](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autod4beb9ac68bc4b56b5683564866387ec)    

That night Louanne made an even worse discovery: Eleanor snored(埃莉诺晚上睡觉的时候会打呼噜).    

![24](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto151f451597434b4ab4c94c20edbfde41)    

The next morning Louanne didn't feel too rested(感觉没休息好). Maybe a cold shower will wake me up, she thought, but she was last in line for the bathroom.    

![25](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autoc6aa4604999b4f0a84d0939763bb046a)    

At breakfast George's mother announced a picnic(野餐) in the park. Oh, good, thought Louanne. Maybe George and I can sneak away(悄悄溜走) and play by ourselves.    

![26](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto25edd375a072436b8643f3eacafde12b)    

But George had to help keep an eye on(照看,密切注视) the little kids,    

![27](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autofde82df7fd79458493656631f2b88f7d)    

and the potato salad wasn't the way Louanne's mother made it, either.    

![28](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autocdf58333cbb94a54990d1d2923b741d4)    

That evening Louanne tried to find a quiet corner where she could read, but it was no use.    

![29](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto2b78e4b30ee948f0be7242adbf7b3ca7)    

"Doesn't all this noise all the time drive you crazy(让你抓狂)?" she asked George.    

George looked surprised. "I like it this way," he said.    

![30](https://590233ee4fbb3.cdn.sohucs.com/auto/1-auto4d9b0e4c32524897b65b6dce2dbdd0b3)    

Sunday morning Louanne packed her suitcase again. As soon as she saw her parents drive up, she rushed home(一见她爸妈开车回来,她立马就冲回了家).    

![31](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autoc9d6877b1a7e49a0ab4b04427467207b)    

"Boy(天哪), am I glad to see you!" said Louanne.    

"How was your weekend?" Mom asked. "Was it fun being part of a big family?"    

"Oh, a big family is just great— for George."    

![32](https://590233ee4fbb3.cdn.sohucs.com/auto/1-autof8e499ede9cc48aeb6c63269fbfe65c4)    

"But you and Daddy are the perfect family for me. Do we have any more jelly beans?"
