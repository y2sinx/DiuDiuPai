### 巫婆奶奶    

汤米·狄波拉    

西蒙与舒斯特出版公司    

---

![03_1.png](https://ae04.alicdn.com/kf/H9e289045c4bf467bb7735e51258a244c3.jpg)    

---

In a town in Calabria, a long time ago, there lived an old lady everyone called Strega Nona, which meant "Grandma Witch".    
很久以前, 在一个叫卡拉布利亚(Calabria)的小镇上, 住着一位老太太, 大家都管她叫"Strega Nona", 即"巫婆奶奶(Grandma Witch)".    

Although all the people in town talked about her in whispers, they all went to see her if they had troubles. Even the priest and the sisters of the convent went, because Strega Nona did have a magic touch.    
镇上的人虽然总是在小声地谈论她(talked about her in whispers), 但是一旦有了麻烦, 他们又全都屁颠屁颠地跑去找她. 甚至连修道院里的神父和修女(the priest and the sisters of the convent)也会去找她, 因为她真的会魔法(did have a magic).    

![04_1.png](https://ae04.alicdn.com/kf/H068f82c7e6e447b080fa6bbbd4ac4e35t.jpg)    

---

She could cure a headache, with oil and water and a hairpin.    
她可以用油, 水, 还有发夹(hairpin)治疗头痛.    

![05_2-1.png](https://sc03.alicdn.com/kf/Hae628caa59c14cf0985c27119c26b7a2G.jpg)    

---

She made special potions for the girls who wanted husbands.    
还可以为那些想要老公(husbands)的女孩子调制神奇的药水(special potions).    

![05_2-2.png](https://sc04.alicdn.com/kf/H3647264262c044f08b6567ca74621152V.jpg)    

---

And she was very good at getting rid of warts.    
而且她还对去疣(warts)非常在行.    

![05_2-3.png](https://sc03.alicdn.com/kf/Hc5a9ef4429314653bb8d3802ca9a9acfT.jpg)    

---

But Strega Nona was getting old, and she needed someone to help her keep her little house and garden, so she put up a sign in the town square.    
但是巫婆奶奶老了, 她需要有人帮忙打理(keep)她的小房子和花园, 所以她在镇广场(town square)贴了一张"招聘启事"(put up a sign).    

![05_1.png](https://sc04.alicdn.com/kf/H89563b812775470bab112c47afc8fbebh.jpg)    

---

And Big Anthony, who didn't pay attention, went to see her.    
安东尼并不在意那是什么工作(didn't pay attention)就跑去了巫婆奶奶家.    

"Anthony," said Strega Nona, "you must sweep the house and wash the dishes. You must weed the garden and pick the vegetables. You must feed the goat and milk her. And you must fetch the water. For this, I will give you three coins and a place to sleep and food to eat."    
"安东尼," 巫婆奶奶说, "你必须得打扫房子和洗盘子. 你必须得给菜园除草和摘菜(pick the vegetables). 你必须得喂羊和挤奶. 还有, 你必须得去打水. 做完这些工作, 我会给你三个硬币, 一个睡觉的地方, 还有可以吃的食物."    

"Oh, grazie," said Big Anthony.    
"哇, 谢谢(grazie)!" 安东尼说.    

"The one thing you must never do," said Strega Nona, "is touch the pasta pot. It is very valuable and I don't let anyone touch it!"    
"有件你绝对不能做的事," 巫婆奶奶说, "那就是千万不要碰我的面锅(pasta pot). 它非常的贵重, 我不允许任何人去碰它!"    

"Oh, si, yes," said Big Anthony.    
"哦, 好的," 安东尼说.    

![06_1.png](https://ae02.alicdn.com/kf/Ha98dfb19ca754da999c5b5312e148111y.jpg)    

---

And so the days went by. Big Anthony did his work and Strega Nona met with the people who came to see her for headaches and husbands and warts.    
日子就这样一天天的过去. 安东尼做着他的工作, 而巫婆奶奶也做着她的工作, 即给别人治头痛, 找老公, 去疣子.    

Big Anthony had a nice bed to sleep in next to the goat shed, and he had food to eat.    
安东尼靠着羊圈(goat shed)有一张舒适的床, 而且他还有吃的.    

![06_2-1.png](https://sc01.alicdn.com/kf/Hf54826a0392540c8a801fdc84e357d1eB.jpg)    

---

One evening when Big Anthony was milking the goat, he heard Strega Nona singing. Peeking in the window, he saw Strega Nona standing over the pasta pot.    
一天晚上, 当安东尼给山羊挤奶的时候, 他听到巫婆奶奶在唱歌. 从窗户里偷偷的瞄过去(peeking), 他看见巫婆奶奶正站在面锅前.    

![06_2-2.png](https://sc04.alicdn.com/kf/H640f823515a54a4cbfafed72183ad727E.jpg)    

---

She sang,    
Bubble, bubble, pasta pot,    
Boil me some pasta, nice and hot,    
I'm hungry and it's time to sup,    
Boil enough pasta to fill me up.    
她唱道:    
冒泡吧, 冒泡吧, 我的面锅,    
煮一些又好又热的意面给我吃吧,    
我饿了, 现在该是吃饭(sup)的时候啦,    
煮上满满的一大锅把我喂饱吧!    

And the pasta pot bubbled and boiled and was suddenly filled with steaming hot pasta.    
面锅冒着泡煮啊煮(bubbled and boiled), 突然间就塞满了热气腾腾的意面(steaming hot pasta).    

![07_2-1.png](https://sc01.alicdn.com/kf/H18d7d7bf215145e59f2e7d6bf4857b29L.jpg)    

---

Then Strega Nona sang,    
Enough, enough, pasta pot,    
I have my pasta, nice and hot,    
So simmer down my pot of clay,    
Until I'm hungry another day.    
之后巫婆奶奶唱道:    
够了, 够了, 我的面锅,    
我已经有了又好又热的意面,    
现在把火停下来吧(simmer down),    
一直停到改天我又饿了的时候.    

"How wonderful!" said Big Anthony. "That's a magic pot for sure!"    
"太神奇了!" 安东尼说. "那一定是一个有魔法的面锅."    

![07_2-2.png](https://ae03.alicdn.com/kf/H382bda2794444e02a6833956404bcdfbp.jpg)    

---

And Strega Nona called Big Anthony in for supper.    
然后巫婆奶奶喊来了安东尼吃饭.    

But too bad for Big Anthony, because he didn't get to see Strega Nona blow three kisses to the magic pasta pot.    
但是不幸的是, 安东尼并没看见巫婆奶奶给她神奇的面锅吹了三个吻.    

![07_1.png](https://ae03.alicdn.com/kf/H06c10ad3ca944a8291fbb9f11bb03589R.jpg)    

---

And this is what happened.    
于是便有了接下来的故事.    

The next day when Big Anthony went to the town square to fetch the water, he told everyone about the pasta pot.    
第二天当安东尼去镇广场打水的时候, 他把面锅的事告诉了所有人.    

And naturally everyone laughed at him, because it sounded to silly— a pot that cooked all by itself.    
当然, 所有的人都在嘲笑他, 因为这听起来太蠢了—— 一个会自己煮饭的锅.    

"You'd better go and confess to the priest, Big Anthony," they said. "Such a lie!"    
"你最好去找神父忏悔(confess)一下吧, 安东尼," 他们说. "你撒得这个谎太好笑啦!"    

And Big Anthony was angry and that wasn't a very good thing to be.    
安东尼很生气, 这可不是一件好事.    

"I'll show them!" he said to himself. "Someday I will get the Pasta Pot and make it cook! And then they'll be sorry."    
"我会证明给你们看的!" 他对自己说. "总有一天, 我会拿到面锅, 并让它自己煮面! 到时候你们就知道不该嘲笑我啦! "    

![08_2.png](https://ae02.alicdn.com/kf/H85bc81178c764529a2b657631c11b6b15.jpg)    

---

That day came sooner than even Big Anthony would have thought, because two days later Strega Nona said to Big Anthony, "Anthony, I must go over the mountain to the next town to see my friend, Strega Amelia. Sweep the house and weed the garden. Feed the goat and milk her and for your lunch, there are some bread and cheese in the cupboard. And remember, don't touch the pasta pot."    
那一天来的竟然(even)比安东尼想的还要早, 因为两天之后, 巫婆奶奶对安东尼说: "安东尼啊, 我要去山那边找我的朋友. 要记得打扫房子, 给菜园除草, 喂羊以及挤奶. 至于你的午饭, 你吃壁柜里的面包和奶酪就好啦! 还有要记住, 千万不要碰我的面锅."    

"Oh, yes— yes— Strega Nona," said Big Anthony. But inside he was thinking, My chance has come!    
安东尼虽然嘴上这么说着: "好的, 巫婆奶奶, 好的!" 但实际上他心里却在想着: "我的机会终于来了".    

![08_1.png](https://ae03.alicdn.com/kf/H4efd5d2e95e6414e8635a4f17e3d385bv.jpg)    

---

As soon as Strega Nona was out of sight, Big Anthony went inside, pulled the pasta pot off the shelf and put it on the floor.    
巫婆奶奶刚走出他的视线(out of sight), 安东尼就跑到房子里, 把面锅从架子(shelf)上取了下来, 然后放在了地面上.    

![09_2-1.png](https://ae03.alicdn.com/kf/H975442cb06ec4cb98a17b0c2779388b91.jpg)    

---

"Now, let's see if I can remember the words," said Big Anthony.    
安东尼说: "让我试试我记不记得那些话."    

And Big Anthony sang,    
Bubble, bubble, pasta pot,    
Boil me some pasta, nice and hot,    
I'm hungry and it's time to sup,    
Boil enough to fill me up.    
然后安东尼唱道:    
冒泡吧, 冒泡吧, 我的面锅,    
煮一些又好又热的意面给我吃吧,    
我饿了, 现在该是吃饭(sup)的时候啦,    
煮上满满的一大锅把我喂饱吧!    

And sure enough, the pot bubbled and boiled and began to fill up with pasta.    
果然(sure enough), 面锅开始冒泡和沸腾, 并开始塞满了意面.    

![09_2-2.png](https://sc04.alicdn.com/kf/Hade3ae0b15c0420d839d84296be0575fD.jpg)    

---

"Aha!" said Big Anthony, and he ran to the town square, jumped on the fountain and shouted, "Everyone get forks and plates and platters and bowls. Pasta for all at Strega Nona's house. Big Anthony has made the magic pasta pot work."    
"啊哈!" 安东尼跑到了镇广场, 跳到了喷泉(fountain)上, 然后大喊着, "大家快去拿叉子和碗盘儿(platters and bowls). 巫婆奶奶家有为所有人准备的意面. 安东尼让神奇的面锅煮饭啦."    

Of course everyone laughed, but ran home to get forks and plates and platters and bowls, and sure enough, when they got to Strega Nona's the pasta pot was so full it was beginning to overflow.    
虽然大家都在笑, 但是他们都回家去拿叉子和碗盘了, 当他们来到巫婆奶奶家时, 果然(sure enough), 面锅里的意面满的都开始往外溢了(overflow).    

![index-9_1.png](https://i.loli.net/2021/01/19/ObyxkaXrAsVhGop.png)    

---

Big Anthony was a hero!    
安东尼是一个英雄!    

He scooped out pasta and filled the plates and platters and bowls.    
他把意面舀出, 并盛在了每个人的盘子和碗里.    

There was more than enough for all the townspeople, including the priest and the sisters from the convent.    
这些面对整个小镇的人来说也太多了, 其中还包括修道院的神父和修女.    

And some people came back for two and three helpings, but the pot was never empty.    
有些人还回来盛了两三次, 但是面锅从来没有空过.    

![10_1.png](https://ae03.alicdn.com/kf/H108e46366b8140439e0aaf90f94b6085G.jpg)    

---

When all had had their fill, Big Anthony sang,    
Enough, enough, my pasta pot,    
I have my pasta nice and hot,    
So simmer down, my pot of clay,    
Until I'm hungry another day.    

当所有人都吃饱了的时候, 安东尼唱道:    
够了, 够了, 我的面锅,    
我已经有了又好又热的意面,    
现在把火停下来吧(simmer down),    
一直停到改天我又饿了的时候.    

But, alas, he did not blow the three kisses!    
但是, 不幸的是, 他并没有吹三个吻到里面.    

![11_2.png](https://sc01.alicdn.com/kf/H73d5f0f827b44928b5272218effc24b0W.jpg)    

---

He went outside and to the applause of the crowd, Big Anthony took a bow.    
安东尼跑到了鼓掌的人群当中, 然后鞠了一个躬.    

He was so busy listening to compliments from everyone that he didn't notice the pasta pot was still bubbling and boiling, until a sister from the convent said, "Oh, Big Anthony, look!"    
他正忙着听来自每个人的称赞(compliments), 以至于他都没有注意到面锅还在冒泡和沸腾呢, 直到一个修女对他说, "哎, 安东尼, 你快看呐!"    

And pasta was pouring out of the pot all over the floor of Strega Nona's house and was coming out the door!    
意面从面锅里倾泻了出来(pouring out), 在巫婆奶奶家的地面上爬的到处都是, 最后还从门口爬了出来.    

![11_1.png](https://ae04.alicdn.com/kf/H4c47eaf77d864ce38bb3344263c7f788N.jpg)    

---

Big Anthony rushed in and shouted the magic words again, but the pot kept bubbling.    
安东尼冲进了屋, 对面锅又喊了一遍那些神奇的话, 但是面锅还是在不停地冒泡.    

![12_2-1.png](https://ae02.alicdn.com/kf/H4e1a2aee7b584c52a4773882beddabf5z.jpg)    

---

He took the pot off the floor, but pasta kept pouring from it.    
他把面锅提了起来, 但是面还是在不停地从锅里倾泻出来.    

![12_2-2.png](https://ae04.alicdn.com/kf/H6eb5bb9594454d25a590b55dfe0239ceb.jpg)    

---

Big Anthony grabbed a cover and put it on the pot and sat on it.    
安东尼匆忙的拿了一个盖子(grabbed a cover), 盖在了上面, 然后自己还坐了上去.    

![12_1-1.png](https://ae02.alicdn.com/kf/Ha9e2ba69495848f880bc915b2807d9d6D.jpg)    

---

But the pasta raised the cover, and Big Anthony as well, and spilled on the floor of Strega Nona's house.    
但是面把盖子和安东尼都顶了起来, 不停地溢在(spilled on)巫婆奶奶家的地面上.    

![12_1-2.png](https://ae04.alicdn.com/kf/Hfc3835c4851a4de09a86e13223fd5c6ej.jpg)    

---

"Stop!" yelled Big Anthony.    
"停下来啊!" 安东尼大喊着.    

But the pasta did not stop and if someone hadn't grabbed poor Big Anthony, the pasta would have covered him up. The pasta had all but filled the little house.    
但是面并没有停下来, 而且, 要不是有人抓住了安东尼, 他早就被面给淹没了. 面塞满了巫婆奶奶家的房子.    

Out of the windows and through the doors came the pasta and the pot kept right on bubbling.    
面锅不停的冒着泡, 还从窗户和门口不停地往外跑.    

![13_2.png](https://ae02.alicdn.com/kf/H1f7adedbe448498f8e90e04ef3a797e0E.jpg)    

---

The townspeople began to worry.    
小镇的人(townspeople)开始担心了.    

"Do something, Big Anthony," they shouted.    
"快做些什么, 安东尼," 他们大喊着.    

Big Anthony sang the magic song again but without the three kisses it did no good!    
安东尼又唱了一遍魔法歌, 但是没有那三个吻, 这魔法歌一点都不管用!    

By this time the pasta was on its way down the road and all the people were running to keep ahead of it.    
这时面已经冲到了路上, 所有人都在拼命的跑在它的前头.    

![13_1.png](https://sc02.alicdn.com/kf/He59e50def366488d80196959fcb62877X.jpg)    

---

"We must protect our town from the pasta," shouted the mayor. "Get mattresses, tables, doors— anything to make a barricade."    
"我们必须保护我们的小镇免受意面的摧毁," 市长大喊着. "快去把床垫(mattresses), 桌子, 门, 以及所有能拿的东西全都拿过来, 我们建个路障(barricade)."    

But even that didn't work. The pot kept bubbling and the pasta kept coming!    
然而这并没有什么卵用, 面锅还是在不停的冒泡, 面还是在不停冒出来!    

![14_1.png](https://sc04.alicdn.com/kf/H7f3aed74bba94f56bb60c6e581bc28d8r.jpg)    

---

"We are lost," said the people, and the priest and the sisters of the convent began praying.    
小镇的人不知道如何是好, 而神父和修女只知道祈祷.    

"The pasta will cover our town," they cried.    
"这些面会把我们小镇埋起来的," 他们大喊着.    

And it certainly would have, had Strega Nona not come down the road, home from her visit.    
小镇确实会被埋起来的, 如果不是恰好这个时候巫婆奶奶从她的朋友家回来了.    

She didn't have to look twice to know what had happened.    
她一看(didn't have to look twice)就知道发生了什么.    

![15_1.png](https://ae04.alicdn.com/kf/Hf89a97b2f84c462d8a0a372398c36c8eL.jpg)    

---

She sang the magic song and blew the three kisses and with a sputter the pot stopped boiling and the pasta came to a halt.    
她唱了一遍魔法歌, 并吹了三个吻, 只听"啪"的一声(sputter), 面锅就停止冒泡了, 面也不再溢出来了.    

![16_2.png](https://ae03.alicdn.com/kf/He32caa9b6f2a4d73a6cb555b2300272cp.jpg)    

---

"Oh, grazia— thank you, thank you, Strega Nona," the people cried.    
"哦, 谢谢你, 巫婆奶奶," 人们喊着.    

![16_1-1.png](https://sc04.alicdn.com/kf/Ha0b839f9344b40aba2b1658f99de95ebD.jpg)    

---

But then they turned on poor Big Anthony.    
然后他们转向了可怜兮兮的(poor)安东尼.    

"String him up," the men of the town shouted.    
"把他吊起来," 小镇的男人们大喊着.    

![16_1-2.png](https://sc04.alicdn.com/kf/Hbeccef89ee0e463b82e40c7d77d8942bN.jpg)    

---

"Now, wait," said Strega Nona. "The punishment must fit the crime." And she took a fork from a lady standing nearby and held it out to Big Anthony.    
"慢着," 巫婆奶奶说. "惩罚必须与罪行相称." 她从旁边站着的一位女士手里拿了一把叉子, 然后把它递给了(held it out to)安东尼.    

"All right, Anthony, you wanted pasta from my magic pasta pot," Strega Nona said, "and I want to sleep in my little bed tonight. So start eating."    
"好吧, 安东尼, 既然你那么想吃我面锅里的面," 巫婆奶奶说, "今晚我睡在我的小床上, 而你呢, 就吃一晚上面吧."    

![17_1.png](https://ae03.alicdn.com/kf/Hcb1c8f866e274a47aceac55dcd6cd8b5e.jpg)    

---

And he did—    
然后, 他照办了——    

![18_1.png](https://ae04.alicdn.com/kf/Heb0f3c62b9f04d5c8f8e8613276f3202x.jpg)    

---

poor Big Anthony.    
哎, 可怜的安东尼!
