大约两百年以前, 有些被称为 "夏克尔" 或 "震颤派" 的宗教人士开始一起生活在一些大农场. 他们把彼此当成兄弟姐妹, 分享他们所拥有的一切. 他们自己种食物, 自己建房子, 自己造工具, 还自己做家具. 他们所创造的东西既简单又坚固, 既讲究又实用. 他们会在星期天或晚上的时候唱歌跳舞, 他们会围成一圈一圈的, 女性们往一个方向跳, 男性们往另一个方向跳. "简单的礼物" 大概创作于1848年, 至今依然流行, 但是听起来却略有不同了. 因为现在这首歌被唱的比较慢了, 而夏克尔人(Shaker)则喜欢伴着欢快的舞蹈唱这首歌.
