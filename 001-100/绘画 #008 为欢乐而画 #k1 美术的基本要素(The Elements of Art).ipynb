没有这七个基本要素(building blocks), 你将无法(wouldn't be able to)进行绘画.

---

线条(Lines)是空间中点的移动路径(path).

![f0026-01.jpg](https://s.pc.qq.com/tousu/img/20210301/7532874_1614584379.jpg)

dotted(点状的), thin(稀疏的), close(靠近的), thick(浓密的), bumpy(颠簸的), curved(弯曲的), wavy(波纹状的), jagged(锯齿状的).

线条可以用于描绘螺旋(spirals), 边框(borders), 花样(patterns), 并且显示纹理(texture)和方向.

![f0026-01a.jpg](https://s.pc.qq.com/tousu/img/20210301/2595324_1614584379.jpg)

---

形状(Shape)是二维的平坦的(flat), 仅包括高度和宽度(height and width).

![f0026-02.jpg](https://s.pc.qq.com/tousu/img/20210301/8780949_1614584379.jpg)

geometric(几何学图形的), organic(有机的).

---

构成(Form)是三维的, 包含高度, 宽度和深度(depth).

![f0026-03.jpg](https://s.pc.qq.com/tousu/img/20210301/5816843_1614584378.jpg)

---

颜色(Color)或色调(Hue)是眼睛所能感知到的来自物体或表面的光的反射(the reflection of light). 色轮(color wheel)解释了颜色是如何混合的, 并能帮助我们理解了颜色之间的关系. 

![f0026-04.jpg](https://s.pc.qq.com/tousu/img/20210301/6508659_1614584379.jpg)

---

纹理/质感(Texture)是表面的品质(quality)或物体的感觉(the feel of an object). 它可以是光滑的, 粗糙的, 颠簸的(bumpy), 坚硬的, 柔软的等等.

![f0027-01.jpg](https://s.pc.qq.com/tousu/img/20210301/5799753_1614584379.jpg)

smooth(光滑的), rough(粗糙的), sandy(沙质感的), soft(柔软的), knitted(针织感的), woven(编织感的), seedy(多籽的).

![f0027-01a.jpg](https://s.pc.qq.com/tousu/img/20210301/6981025_1614584379.jpg)

wood grain(木纹), scaly(有鳞的), 多草的(grassy), straight(直的/整齐的). 

---

空间(Space)是形状和元素之间的区域.

![f0027-02.jpg](https://s.pc.qq.com/tousu/img/20210301/4201858_1614584379.jpg)

---

值(Value)是从浅到深的色调范围(the range of tone). 有很多技巧(techniques)可以表现值的大小, 并且可以使用任何媒介(medium)表现. 值的大小可以传达(convey)对比度(highlight), 高光(highlight), 纹理, 以及构成.

![f0027-03.jpg](https://s.pc.qq.com/tousu/img/20210301/1398417_1614584379.jpg)

pencil(铅笔), pen and ink(钢笔和墨水), stippling(点刻法).

![f0027-03a.jpg](https://ae03.alicdn.com/kf/U6615569804664bba8c871af47849c7bbu.jpg)

tint and shade(色彩和阴影), watercolor(水彩).
