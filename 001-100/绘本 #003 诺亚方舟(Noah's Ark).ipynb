# Noah's Ark  

retold and illustrated by  
Lucy Cousins  

Candlewick Press  

---

![2](https://s3.ax1x.com/2020/12/07/DzUFu6.jpg)  

很久以前,这儿住着一个人,名叫诺亚.  
诺亚是一个好人,他相信上帝.  

---

![3](https://s3.ax1x.com/2020/12/07/DzUkDK.jpg)  

而在这个世界上,也有很多坏人.  
上帝想要惩罚他们,于是便对诺亚说:  
"我要发一场大洪水,把那些坏人们冲走.  
你去为你的家人以及所有的动物建造一艘逃生的方舟吧!"  

---

![4](https://s3.ax1x.com/2020/12/07/DzUPjx.jpg)  

诺亚为了建造方舟,同他的家人工作了很长时间(很多很多年).  
最终把它造好了.  

---

![5](https://s3.ax1x.com/2020/12/07/DzUCg1.jpg)  

诺亚同他的家人收集了很多的食物.  
之后,动物们两两成对的来了...  

---

![6](https://s3.ax1x.com/2020/12/07/DzUAHO.jpg)  

然后两两成双的进入了方舟.  

---

![7](https://s3.ax1x.com/2020/12/07/DzUnCd.jpg)  

当方舟满载的时候,诺亚感觉到天空下了一滴雨.  

---

![8](https://s3.ax1x.com/2020/12/07/DzUZUe.jpg)  

雨下啊下啊.  
一直下了四十天四十夜.  
世界完全被大水给淹没了.  

---

![9](https://s3.ax1x.com/2020/12/07/DzUe4H.jpg)  

终于,雨停了,太阳也出来了.  
诺亚派了一只鸽子去寻找干燥的陆地.  

---

![10](https://s3.ax1x.com/2020/12/07/DzUu8A.jpg)  

鸽子衔(xián)着一根橄榄枝回来了.  
"好哇(wa)!" 诺亚大喊着.  
"大水终于结束了."  

---

![11](https://s3.ax1x.com/2020/12/07/DzUKgI.jpg)  

但是又过了好多天,方舟才停到了干燥的陆地上.  

---

![12](https://s3.ax1x.com/2020/12/07/DzUMvt.jpg)  

最后,诺亚(一家人)同所有的动物都安全地走出了方舟,然后开始了在地球上新生活.  

---

![13](https://s3.ax1x.com/2020/12/07/DzUlKP.jpg)  

**露西⋅卡曾斯**给我们带来了一场视觉盛宴: 粉色的火烈鸟,绿色的沼泽鳄,灰色的大象,当然,还有诺亚这位圣经故事中有名的人物.  

美国某一书商这样评价本书: **这是对孩子们而言一本最好的诺亚方舟故事**.
