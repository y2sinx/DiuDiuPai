在"小脚丫之书",我们用书中的这些艺术和故事(来自生活的各个方面),以启迪孩子们的心灵和思想,以激发她们去深读(read deeper),去探求(search further),去探索她们创作的天赋(gifts).  

我们的灵感来源于许多不同的文化,而我们的焦点在于(focus on)激励孩子们的心灵自立,激励孩子们对求知的热情,以及激励孩子们包容其他的文化传统.  

我们的作家,艺术家,故事讲述者,自世界各地,她们深思熟虑,使我们的产品不仅结合了现在最好的东西,而且结合了过去最好的东西,以把我们的孩子培养成未来更好的管理者(caretaker).
