# Cat and Mouse in the Rain  

Tomek Bogacki  

---

![title](https://ae04.alicdn.com/kf/H14dcc9b6d46d4bbd89bf822159ea1332Q.jpg)  

---

![01](https://ae02.alicdn.com/kf/H7cb7744df5a7469e9541e04de4beb18ae.jpg)  

"You shouldn't go out on such a cloudy day. It might rain," the little mice said the their sister.  
"你不应该在这样一个多云的天气中出去. 很有可能会下雨的," 小老鼠的姐妹们对小老鼠说.  

But she didn't pay attention.  
但是她并没有听她们的话.  

She wanted to meet her friend in the green meadow.  
她想要去绿草地上见她的朋友.  

---

![02](https://sc02.alicdn.com/kf/H8f7bbc33616b4e35acf17e017a854536a.jpg)  

"Don't go out on such a cloudy day. It might rain," the little cats said to their brother.  
"你不应该在这样一个多云的天气中出去. 很有可能会下雨的," 小猫猫们对他们的猫弟弟说.  

But he didn't listen.  
但是他并没有听他们的话.  

He wanted to meet his friend, too.  
他想出去见他的朋友, 跟小老鼠一样.  

---

![03](https://sc01.alicdn.com/kf/H8c10cfb60da5423bbbc94fda5b648a90J.jpg)  

The little cat and the little mouse met in the green meadow as they did every day to play together.  
小猫猫和小老鼠在绿草地上见面了, 像往常那样, 打算在一起玩儿.  

"Let's do something exciting," said the little cat.  
"让我们一起去做点刺激的事吧," 小猫猫说.  

---

![04](https://ae04.alicdn.com/kf/H9f5d7eb1061c4a3888ac75b6e9ae3664u.jpg)  

"How can we do anything exciting? It is starting to rain," said the little mouse.  
"我们怎么能去做刺激的事啊? 天正在下雨呢," 小老鼠说.  

---

![05](https://ae02.alicdn.com/kf/H580530b09af74596ad56a6944b2eeba9A.jpg)  

"What a wonderful day!" said a voice above their heads.  
"多么美妙的一天啊!" 从他们的头上方传来这样一个声音.  

---

![06](https://ae03.alicdn.com/kf/H187ab5c113de42598c66a8f0fe8867adu.jpg)  

"Are you afraid of the rain?" asked the frog.  
"你们害怕这雨吗?" 青蛙问道.  

"No..." said the little mouse, uncertainly.  
"不害怕..." 小老鼠不确定的回答.  

"No..." said the little cat, doubtfully.  
"不害怕..." 小猫猫半信半疑的回答.  

---

![07](https://ae03.alicdn.com/kf/H3ed4d402931e4f99ad49ec3a24264dbcQ.jpg)  

"Follow me", said the frog, jumping right into a puddle.  
"跟着我来", 青蛙说着, 然后正好跳进了一个水坑.  

---

![08](https://ae02.alicdn.com/kf/H9a1127f335e24b26850ea5249ef06df4r.jpg)  

So the mouse jumped into the puddle, too, making a big splash.  
于是, 老鼠也跳了进去, 还溅起了一个大水花.  

---

![09](https://ae02.alicdn.com/kf/H9d711db1acf249ba991df6a0c5753757I.jpg)  

Then the cat jumped into the puddle, making a bigger splash.  
之后, 猫猫也跳了进去, 溅起了一个更大的水花.  

---

![10](https://ae03.alicdn.com/kf/H261f492c4c1d479e81d82145cfb51f25V.jpg)  

They jumped and splashed from puddle to puddle until they got all wet and their mothers called them home.  
他们从一个水坑跳到另一个水坑, 溅起了一朵朵的水花, 直到他们浑身都湿透了, 而且妈妈们喊他们回家了.  

---

![11](https://ae02.alicdn.com/kf/H79e5abf60251496098a14eee0dfa3a06e.jpg)  

"We played with a frog in the rain and had so much fun!" said the little mouse to her sisters.  
"我们跟青蛙在雨中玩得好开心啊!" 小老鼠对她的姐妹们说.  

"How could you have fun in the rain?" they asked.  
"你们怎么可能会在雨中也玩得很开心呢?" 小老鼠们问.  

"We had so much fun in the rain! We played with a frog!" said the little cat to his brothers.  
"我们跟青蛙在雨中玩得好开心啊!" 小猫猫对他的兄弟们说.  

"How could you play in the rain?" they asked.  
"你们是怎么在雨中玩耍的?" 小猫猫们问.  

---

![12](https://ae04.alicdn.com/kf/H18edbb72f50a47148e8915b835879fc2P.jpg)  

But on the next rainy day, all the little cats and all the little mice went out to play in the puddles with all the little frogs.  
但是, 在接下来的一个雨天, 所有的小猫猫们和所有的小老鼠们, 以及所有的小青蛙们全都出来踩水坑了.  

And they all had fun.  
并且, 他们全都玩得很开心.
