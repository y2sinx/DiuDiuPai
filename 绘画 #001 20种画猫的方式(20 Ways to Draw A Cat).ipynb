## 引言  

**Let’s learn how to draw animals together!**  
**让我们一起来学习画动物吧!**  

There are few things more fascinating and beautiful on Earth than animals.  
地球上很少有事物能够像动物那样迷人和美好.  

They come in all shapes and forms—from tiny birds to gigantic elephants, sweet deer to sly foxes, and striped zebras to spotted giraffes.  
它们具有不同的形状和种类——从小小的鸟儿到巨大的大象,从甜美的小鹿到狡猾的狐狸,从具有条状斑纹的斑马到具有圆形斑点的长颈鹿.  

You can pick your favorite animals from this book to draw, or you can think of animals that I haven’t included.  
你可以在本书中挑选你最喜欢的动物,或者考虑这本书中没有包含的动物,然后去画它们.  

There are forty-five different types of animals in this book, but that’s just a tiny number out of all the animals out there!  
这本书中包含 45 种不同类型的动物,这只占所有动物种类的一小部分.  

Next time you are at the zoo, keep an eye out for the unusual animals you haven’t heard of before.  
下一次你去动物园的时候,你可以特别留意一下那些不同寻常的的(你从未听说过的)动物.  

For example, have you ever heard of a **capybara**?  
比如,你曾听过**水豚**这种动物么?  

A capybara is the largest rodent in the world.  
水豚是世界上最大的啮齿类动物.  

That means it belongs to the same family as squirrels and bunnies, but it is as big as a dog!  
这意味着它与松鼠和兔子属于同一家族(科),但是它们的体型像狗一样大!  

Even the same types of animals come in so many varieties that it’s impossible to draw them just one way.  
即使是相同类型的动物它们亦有许多不同的变种,因此很难用一种方式把它们全部给画出来.  

At first glance, a family of zebras might look very similar, but if you look closely you will notice that they don’t have the same identical stripes.  
乍一看,斑马属的动物可能看起来非常相似,但是当你仔细看的时候,你就会注意到它们的斑纹其实并不相同.  

Think of yourself—you might look a little bit like the rest of your family, but you don’t look exactly like them.  
可以想下你自己——你可能跟你家中的某一人有点相似,但是你们长的并不是完全一样的.  

Every single animal looks special in its own way, so I made sure that each drawing is a little different from the last.  
每一个单独的动物个体看起来都不相同,因此本书中的每一幅画都与上一幅画有多多少少的不同.  

## 如何使用本书  

These drawings are all made up of a combination of lines, shapes, and patterns.  
这些图画全都是由线条,形状,和模式组成的.  

Look for squares, circles, straight lines, and **squiggly lines** to help break down the pictures, and then try copying the different animal drawings using those simple elements.  
(你可以在每一张图画中试着)找一些正方形,圆,直线和**曲线**把图画分解开来,然后试着用这些简单的元素去复制出这些不同的动物画.  

Draw the big shapes and lines first, and then add in the smaller details.  
你可以先画那些大的轮廓和线条,然后为其添加那些较小的细节.  

If you have **tracing paper**, you can trace the drawings directly, but don’t worry about getting them exactly the same.  
如果你有**临摹纸**,你可以直接临摹这些图画,并且不必担心你临摹出来的画与书中的画一模一样.  

When you think you have **gotten the hang of it**, try drawing your own!  
当你认为你已经**掌握其中的要领**之后,你就可以尝试着用你自己的方式去画了!  

Try to start out with a pencil and eraser so that you’re not afraid of making mistakes.  
你可以试着用铅笔和橡皮开始绘制,这样你就不必担心画错了.  

When you feel more comfortable with drawing, explore by using many different types of tools—pens, colored pencils, markers, or even paints.  
当你感觉你对这些画更熟悉的时候,你就可以探索其他不同种类的工具了——如钢笔,彩色铅笔,马克笔,甚至可以直接使用颜料去画.  

Color in the existing drawings and then color in your own drawings.  
你可以先给书中的这些图画上色,然后再给你自己的图画上色.  

If you are feeling especially adventurous, you can cut out paper shapes and paste them on the blank pages.  
如果你喜欢冒险,你可以抠出书中的这些形状,然后将它们粘在一些空白的纸上.  

It’s always fun to try as many things as possible to decide what you like the best.  
这其实是很好玩的,因为只有你尝试了各种不同的可能之后,你才能决定自己真正喜欢的是什么.  

See how many different pictures you can come up with of your favorite animals.  
看看你可以为你最爱的动物们画出多少张不同的图画.  

How many ways can you draw a turtle shell? You can make it a circle, square, or irregular shape.  
你能用多少种方式画一个龟壳呢? 你可以把它画成圆形的,方形的,以及不规则形状的.  

Think about what you can draw inside the shell.  
想想你可以在壳内画些什么.  

You can fill it with stripes, little spots, big spots, squares of different sizes, or just fill it in with **a solid color**.  
你可以使用条纹,小斑点,大斑点,大小不同的正方形去填充它,或者只使用一个**纯色块**去填充它.  

You can also decide to draw nothing inside the shell if you think it looks best that way, since sometimes less is more.  
你也可以决定壳内什么都不画,如果你认为这样很好的话,因为有些时候 "越简单,越完美."  

Don’t forget to show your friends and family when you’re done.  
当你画完一张画的时候,不要忘记拿给你的朋友或者家人看看.  

Or even better yet, ask them if they want to make some drawings of their favorite animals, too!  
如果可以这样那就更好了: 问问它们也想不想画一些自己喜欢的动物!
