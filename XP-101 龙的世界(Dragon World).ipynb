#### 术语表(Glossary)

1. Celestial(天上的): Relating to the sky or outer space.

1. Chimu(尺木): A lump on the head of a Chinese dragon that allows them to fly(龙无尺木,不能升天).

1. Constriction(收缩): A method used by snakes to wrap around their prey(缠着猎物) and crush them(将它们粉碎,勒死).

1. Hypnotize(对...进行催眠): The ability to control and influence minds.

1. Immortality(不朽): The ability to be able to live forever.

1. Invincible(无敌的): Impossible to harm or defeat.

1. Iridescent(色彩斑斓的): Possessing many colors that seem to shift and change.

1. Putrid(腐烂的): Something that is decaying or rotting(这两个单词都是腐烂的意思).

1. Shapeshifting(易形): The ability to change shape and take on a new appearance.

1. Yeouiju(意允珠): A magic orb(魔球) that grants power(能够赋予力量的).
