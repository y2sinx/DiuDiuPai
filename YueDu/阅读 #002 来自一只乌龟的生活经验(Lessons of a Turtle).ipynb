桑迪·金格拉斯    

安德鲁斯·麦克米尔出版公司    

---

![03](https://s.pc.qq.com/tousu/img/20210129/8178928_1611933532.jpg)

**Lessons of a Turtle** is a wise and warm(暖人的) little book about slowing down(放慢自己的人生脚步) and finding your own path(寻找自己的人生道路) in life. It's a celebration of simplicity(简单), a reminder(提醒物) that the little things in life really are the big things.
