### 初识100只动物    

多林·金德斯利出版社(DK)    

![2.png](https://sc03.alicdn.com/kf/H1e8899138f504aaaba4cd3e0b473d570E.jpg)    

这是一本识记动物英文名的小单词书, 共包含十三个主题, 其第一个主题"宠之最爱"如上图所示. 除了你肯定叫得上名来的阿猫阿狗以及兔子和金鱼外, 这里还有些你常见却不知其名的动物, 如长尾小鹦鹉(parakeets), 仓鼠(hamster), 豚鼠(guinea pig)等.    

![4.png](https://ae02.alicdn.com/kf/Heacb4a764e844e84b5d79c08f6cc99d3E.jpg)    

其它十二个主题分别是: 农场动物(Farm animals), 大型动物(Big beasts), 小型动物(Small and tiny), 奔跑者和跳跃者(Runners and jumpers), 会飞的动物(Up in the air), 海洋动物(In the sea), 喜欢待在树上的动物(In the trees), 黑白相间的动物(Black and white), 色彩明亮的动物(Bright colors), 拥有条纹和斑点的动物(Stripes and spots), 拥有鳞片和甲壳的动物(Scales and shells), 以及可爱的动物宝宝们(Baby animals)等.    

![7.png](https://ae02.alicdn.com/kf/H05cace2b62354df78a830f02fa1645a11.jpg)    

你知道"fly"这个单词除了"飞"还有苍蝇的意思嘛? 你知道"dragonfly"其实并不是某种"飞龙"而只是一只人畜无害的蜻蜓嘛? 我相信, 当你看完本书而再遇到这100个"英文单词"时, 你的脑海里一定会在第一时间内反应出"它本该的样子", 而不再仅仅是一个干巴巴的单词了. 对哒, 这就是形象记忆!    

![10.png](https://ae03.alicdn.com/kf/H9a38528dc1fc4bd6b5480162d74dfca2P.jpg)    

作为一位"脸盲+单词盲"的无辜患者, 你可曾向这颗蓝色星球发出过这样的质问: "你们俩明明长得一个样子, 为什么会有两个名字啊?" 看看这四只豹子(海豹不是豹子)你就知道我在说啥啦: jaguar(美洲虎/豹), cheetah(猎豹), panther(黑豹), leopard(花豹). 好吧, 请原谅我的无知! 在我眼里, 我都把它们统称为"急支糖浆".    

![11.png](https://ae02.alicdn.com/kf/H465df64c733640d5b58640ad2315e2b5N.jpg)    

你知道嘛, 记英语中的动物单词是要分雌雄的, 比如: "cock"是指公鸡, 而"hen"是指母鸡. 因此我特意查了一下孔雀(豌豆鸡)的雌雄, 从"peacock"到"peahen"这操作没有错.    

![14.png](https://ae03.alicdn.com/kf/Hd6a7f14ef5d34458a1618a4677859f78c.jpg)    

是哒, 记英语中的动物单词不仅要分雌雄, 还要分成年的未成年的呐! 此处请记住, "kid"除了指人类小孩外还指山羊幼崽(小山羊), 否则, 当别人说自家山羊(she-goat)生了一个"kid"的时候你可能会怀疑人生.
