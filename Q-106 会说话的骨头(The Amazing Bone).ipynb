# 会说话的骨头    

威廉·斯蒂格    

![102](https://s.pc.qq.com/tousu/img/20210508/7097326_1620405779.jpg)    

放学后, 猪猪珀尔(Pearl)并没有直接回家(going straight home), 而是在四处闲逛(dawdled).    

![105](https://s.pc.qq.com/tousu/img/20210508/3924697_1620405786.jpg)    

面包房(bakery), parsnip(欧洲防风草), pumpernickel(裸麦粗面包), crullers(油炸小煎饼), suger dust(糖粉).    

![106](https://s.pc.qq.com/tousu/img/20210508/4862323_1620405818.jpg)    

乡下的老头们(old gaffers)在扔叮当作响的马蹄铁(pitched their ringing horseshoes), 以及吐烟草汁(spat tobacco juice).    

![107](https://s.pc.qq.com/tousu/img/20210508/3090429_1620405917.jpg)    

珀尔坐在草地上(grass ground)欣赏着明媚的春光(bright spring). 她的浅色连衣裙(light dress)感觉就像花瓣(petals)一样.    

![108](https://s.pc.qq.com/tousu/img/20210508/5734755_1620405824.jpg)    

就在这个地方, 珀尔发现了一根会说话的骨头.    

![109](https://s.pc.qq.com/tousu/img/20210508/6207940_1620405827.jpg)    

骨头不仅会说西语,德语,波兰语, 还会模仿(imitate)很多很多的声音.    

I didn't make the world(这个世界不是我创造的,所以我也不知道自己为什么会说话).    

![110](https://s.pc.qq.com/tousu/img/20210508/3081919_1620405804.jpg)    

rheumatism(风湿病), flabbergast(使...大吃一惊).    

![111](https://s.pc.qq.com/tousu/img/20210508/3530299_1620405797.jpg)    

Who should... but...: 谁会... 原来是... 用于强调某件事发生的突然.    

Who should rush out from(从...跑出了) in back of a boulder(大石头的后面) and spoil everything(破坏所有的事情) but three highway robbers(拦路强盗) with pistols and daggers(手枪和匕首).    

![112](https://s.pc.qq.com/tousu/img/20210508/5012637_1620405881.jpg)    

骨头发出了像蛇一样的嘶嘶声(hiss like a snake)以及像狮子一样的咆哮声(roar like a lion), 然后就把强盗们吓跑了.    

![114](https://s.pc.qq.com/tousu/img/20210508/4394305_1620405825.jpg)    

barred their path(挡住了他们的去路), a sprig of lilac in his lapel(他的西服翻领上夹着一支丁香花).    

![115](https://s.pc.qq.com/tousu/img/20210508/1751292_1620405839.jpg)    

plump(胖乎乎的), tender(肉质鲜嫩的), main course at dinner(晚餐主菜), seize sb in a tight embrace(在一个紧紧的拥抱之中抓住某人).    

![116-2](https://s.pc.qq.com/tousu/img/20210508/2951316_1620405914.jpg)    

ravenous crocodile(贪婪的鳄鱼), dote on fresh fox chops(喜欢吃新鲜的狐狸骨头).    

![116](https://s.pc.qq.com/tousu/img/20210508/3387953_1620405776.jpg)    

这只狡猾的狐狸(the wily fox)并不像那些强盗们一样容易受到欺骗(was not as easily duped as the robbers).    

![117-2](https://s.pc.qq.com/tousu/img/20210508/7982713_1620405776.jpg)    

狐狸推推搡搡地赶着珀尔前往他隐蔽的家(hideaway).    

![117](https://s.pc.qq.com/tousu/img/20210508/4142566_1620405776.jpg)    

disgusted with himself for being so soft-hearted(厌恶自己如此的心慈手软), have no shame(厚颜无耻).    

I didn't make the world(这样的世界又不是我造成的,所以我饿了就要吃掉你).    

The bone commenced to(开始) revile the fox(对狐狸展开无情的谩骂). You coward(胆小鬼)! You worm(蠕虫,小人), you odoriferous wretch(臭烘烘的可怜虫)!    

![118](https://s.pc.qq.com/tousu/img/20210508/3426133_1620405607.jpg)    

在接下来的路上, 大家都很安静.    

![119](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/fe38f33c-7d77-4cac-9122-a1d4f6170aec.JPG)    

珀尔坐在地面上(sat on the floor), 盯着墙看(stared at the walls).    

![120](https://s.pc.qq.com/tousu/img/20210508/9038784_1620406085.jpg)    

骨头告诉珀尔, 狐狸在磨刀霍霍向猪猪.    

![121-2](https://s.pc.qq.com/tousu/img/20210508/7049173_1620406077.jpg)    

骨头告诉珀尔, 狐狸在将大火烧得旺旺.    

![121](https://s.pc.qq.com/tousu/img/20210508/4407828_1620406072.jpg)    

珀尔听到锁孔在转动, 吓得一句话也说不出来, 甚至都不敢往门口多看一眼.    

![122](https://s.pc.qq.com/tousu/img/20210508/7900454_1620406080.jpg)    

It's nothing personal(并不是针对你自己,我对所有的猪猪都这样).    

![123-2](https://s.pc.qq.com/tousu/img/20210508/7891117_1620406073.jpg)    

听到骨头在念叨些什么, 狐狸愣住了(standing stock-still).    

![123](https://s.pc.qq.com/tousu/img/20210508/3194890_1620406079.jpg)    

难以置信, 狐狸竟然变成了兔子一般大小!    

![124-2](https://s.pc.qq.com/tousu/img/20210508/6409618_1620406069.jpg)    

哈哈哈~ 狐狸又变成了老鼠一般大小!    

![124](https://s.pc.qq.com/tousu/img/20210508/3825560_1620406070.jpg)

![125](https://s.pc.qq.com/tousu/img/20210508/9713329_1620406075.jpg)

![126](https://s.pc.qq.com/tousu/img/20210508/1519324_1620406086.jpg)
