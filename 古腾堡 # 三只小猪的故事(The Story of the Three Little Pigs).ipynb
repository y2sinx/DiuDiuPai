**Once upon a time** there was an old Sow with three little Pigs, and as she had not enough to keep them, she sent them out to seek their fortune.        

从前有一只老母猪,生了三只小猪,小猪渐渐长大后,老母猪决定让他们独自去寻找属于自己的人生.        

![imgplate-1](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-1.jpg)        
********************************

The first that went off met a Man with a bundle of straw, and said to him, "Please, Man, give me that straw to build me a house"; which the Man did, and the little Pig built a house with it.        

第一只离开的小猪遇到了一个人,并从他那里要了一捆稻草,然后用这捆稻草建了一栋自己的房子.   

![img002](http://www.gutenberg.org/files/18155/18155-h/images/img002.jpg)                
![img003](http://www.gutenberg.org/files/18155/18155-h/images/img003.jpg)        
********************************
        
Presently came along a Wolf, and knocked at the door, and said, "Little Pig, little Pig, let me come in."    

To which the Pig answered, "No, no, by the hair of my chinny **chin/下巴** chin."        

"Then I'll **huff/吹气** and I'll **puff/吹气**, and I'll blow your house in!" said the Wolf. So he huffed and he puffed, and he blew his house in, and ate up the little Pig.        

不久,来了一只狼,敲了敲门,说道:"小猪,小猪,让我进去." 小猪对此回答:"不,不,以我下巴上的毛起誓,我不会放你进来的." 然后,狼一用力就把小猪的房子推倒了,然后就吃了这只小猪.   

![imgplate-2](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-2.jpg)        
********************************
        
The second Pig met a Man with a bundle of **furze/荆豆**, and said, "Please, Man, give me that furze to build a house"; which the Man did, and the Pig built his house.    

Then along came the Wolf and said, "Little Pig, little Pig, let me come in."    

"No, no, by the hair of my chinny chin chin."         

第二只猪也碰到了一个男人,并从他那里要来一捆荆豆,然后用这捆荆豆建造了自己的房子. 然后狼又来拜访这只小猪了.        

![img004](http://www.gutenberg.org/files/18155/18155-h/images/img004.jpg)        
********************************

"Then I'll puff and I'll huff, and I'll blow your house in!" So he huffed and he puffed, and he puffed and he huffed, and at last he blew the house down, and ate up the second little Pig.        

由于这只小猪的房子依然不牢固,所以这只小猪也被狼吃了.        

![img005](http://www.gutenberg.org/files/18155/18155-h/images/img005.jpg)        
********************************

The third little Pig met a Man with a load of **bricks/砖头**, and said, "Please, Man, give me those bricks to build a house with"; so the Man gave him the bricks, and he built his house with them.   

第三只小猪遇到了一个满载砖头的男人,并从他那里要了一些砖头,然后用那些砖头建造了自己的房子.        

![imgplate-3](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-3.jpg)        
********************************

So the Wolf came, as he did to the other little Pigs, and said, "Little Pig, little Pig, let me come in."    

"No, no, by the hair of my chinny chin chin."    

"Then I'll huff and I'll puff, and I'll blow your house in."        

Well, he huffed and he puffed, and he huffed and he puffed, and he puffed and he huffed; but he could not get the house down.    

狼同样来到了这只小猪的房子前,想要故伎重演,但是由于这只小猪的房子过于牢固,他失败了.        

![img006](http://www.gutenberg.org/files/18155/18155-h/images/img006.jpg)        
********************************        

When he found that he could not, with all his huffing and puffing, blow the house down, he said, "Little Pig, I know where there is a nice field of **turnips/萝卜**."            

"Where?" said the little Pig.        

"Oh, in Mr. Smith's **home-field/自家菜园**; and if you will be ready to-morrow morning, I will call for you, and we will go together and get some for dinner."        

"Very well," said the little Pig, "I will be ready. What time do you mean to go?"        

"Oh, at six o'clock."          

当他发现自己无法把小猪的房子推倒后,他告诉小猪,史密斯先生家的地里有一片好吃的萝卜. 如果小猪明早起得来,可以跟他一起去拔萝卜,然后一起回来吃饭. 小猪同意了,他们约定在早上六点钟汇合.        
********************************

Well, the little Pig got up at five, and got the turnips and was home again before six.         

这只小猪五点就起床了,早在六点之前就拿到了萝卜并回到了家中.   

![img007](http://www.gutenberg.org/files/18155/18155-h/images/img007.jpg)        
![imgplate-4](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-4.jpg)             
********************************

When the Wolf came he said, "Little Pig, are you ready?"        
"Ready!" said the little Pig, "I have been and come back again, and got a nice pot-full for dinner."        

六点钟的时候狼来了,问小猪准备好了吗. 小猪说准备个毛线啊,我早就把萝卜拿回来了,现在已经吃的饱饱的了."

![imgplate-5](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-5.jpg)        
********************************

The Wolf felt very angry at this, but thought that he would be up to the little Pig somehow or other; so he said, "Little Pig, I know where there is a nice apple-tree."        

"Where?" said the Pig.        

"Down at Merry-garden," replied the Wolf; "and if you will not deceive me I will come for you, at five o'clock tomorrow, and we will go together and get some apples."  

狼虽然非常生气,但是吃不到小猪不甘心啊,于是他又跟小猪约定明早五点钟去玛丽花园摘苹果.        
********************************

Well, the little Pig woke at four the next morning, and **bustled up/忙碌起来**, and went off for the apples, hoping to get back before the Wolf came; but he had farther to go, and had to climb the tree, so that just as he was coming down from it, he saw the Wolf coming, which, as you may suppose, frightened him very much.         

虽然小猪四点就起床去摘苹果了,但是由于路途遥远,还要爬树,耽搁了很长时间, 所以正当小猪准备从苹果树上爬下来的时候,狼也来到了玛丽花园, 这可把小猪吓坏了.          

![img008](http://www.gutenberg.org/files/18155/18155-h/images/img008.jpg)        
![img009](http://www.gutenberg.org/files/18155/18155-h/images/img009.jpg)        
********************************
        
When the Wolf came up he said, "Little Pig, what! are you here before me? Are they nice apples?"        

"Yes, very," said the little Pig; "I will throw you down one." And he threw it so far that, while the Wolf was gone to pick it up, the little Pig jumped down and ran home.        

然后,狼问小猪树上的苹果好不好吃,小猪说非常好吃,然后他故意往远处给狼扔了一个苹果. 就在狼捡这个苹果的时候,小猪爬下了树并跑回了家.       

![imgplate-6](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-6.jpg)         
![img010](http://www.gutenberg.org/files/18155/18155-h/images/img010.jpg)        
********************************

The next day the Wolf came again, and said to the little Pig, "Little Pig, there is a Fair in the Town this afternoon: will you go?"        

"Oh, yes," said the Pig, "I will go; what time shall you be ready?"        

"At three," said the Wolf.        

接下来的一天,狼又来到小猪的房子前,问小猪下午去不去镇上的集市. 然后跟小猪约定好了三点钟去.        
********************************        

So the little Pig went off before the time, as usual, and got to the Fair, and bought a butter **churn/搅乳器**, and was on his way home with it when he saw the Wolf coming.  

虽然这只小猪又像往常一样提前去了集市,买了一个黄油搅拌器,但是在他回家的路上,他又碰到了狼.

![img011](http://www.gutenberg.org/files/18155/18155-h/images/img011.jpg)        
********************************

Then he could not tell what to do. So he got into the churn to hide, and in doing so turned it round, and it began to roll, and rolled down the hill with the Pig inside it, which frightened the Wolf so much that he ran home without going to the Fair.        

小猪藏进了搅拌器,然后从山坡上滚了下来,直接把狼吓得跑回了家. 小猪又逃过了一劫.        

![imgplate-7](http://www.gutenberg.org/files/18155/18155-h/images/imgplate-7.jpg)        
![img012](http://www.gutenberg.org/files/18155/18155-h/images/img012.jpg)        
********************************

He went to the little Pig's house, and told him how frightened he had been by a great round thing which came down the hill past him.        

Then the little Pig said, "Hah! I frightened you, did I? I had been to the Fair and bought a butter churn, and when I saw you I got into it, and rolled down the hill."        

狼又来到小猪的房子前,说他刚才去集市的路上有个东西从山坡上滚了下来,可把他吓坏了. 小猪嘲笑了他一番,说那是他假扮的.
****************************************************************     

Then the Wolf was very angry indeed, and declared he would eat up the little Pig, and that he would get down the chimney after him.        

狼感觉受到了侮辱,变得非常生气,扬言要顺着小猪家的烟囱爬到房子里, 然后立马吃掉这只小猪.        

![img013](http://www.gutenberg.org/files/18155/18155-h/images/img013.jpg)        
********************************

When the little Pig saw what he was about, he hung on the pot full of water, and made up a **blazing/炽热的** fire, and, just as the Wolf was coming down, took off the cover of the pot, and in fell the Wolf. And the little Pig put on the cover again in an instant, boiled him up, and ate him for supper. And lived happy ever after.        

出乎狼的意外,狼掉进了小猪的陷阱,然后小猪把狼给煮了,狡猾的大灰狼最终成了小猪的晚餐. 从此以后,小猪过上了幸福快乐的生活.
