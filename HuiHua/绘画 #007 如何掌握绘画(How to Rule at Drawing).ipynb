# 如何掌握绘画    

——素描和涂鸦的50个技巧    

雷切尔·哈勒尔    

克罗尼克出版社

### 6) 涂鸦:    

Don't underestimate your doodles.    
不要低估你的涂鸦(doodles).    

The best ideas often come to you while you are doing something else.    
最好的想法往往来源于你做一些其他事的时候.    

![6.png](https://i.loli.net/2021/01/26/iP8sT7moRDASqjM.png)    

### 9) 改变对象的风格:    

Draw the same person five times, each in a different style.    
画同一个人五次, 每次都换一种不同的风格.    

![9.png](https://i.loli.net/2021/01/26/iLjsPK31vOQIyaY.png)    

### 10) 模仿艺术家的作品:    

Whether you're copying one of their pieces exactly or drawing something the way you imagine they would draw it, you'll find that your art takes on a style of its own.    
无论是你完全复制还是按你的想象(他们可能会这样画)去画一幅艺术家的作品, 你都会发现这幅作品拥有其自己的风格.    

Embrace the similarities—and the differences!    
既要接纳其相似性(similarities), 也要包容其所不同(differences).    

![维纳斯的诞生, 桑德罗·波提切利](https://i.loli.net/2021/01/26/z56w2Nhs43ygpxV.png)    

### 30) 以对象为准:    

Draw what the object actually looks like, not what you assume it looks like.    
你应该画这个对象实际的样子, 而不是你自己所假想的那个样子.    

![30.png](https://i.loli.net/2021/01/26/OtA7DaKJTUR4nZL.png)    

### 47) 对象的负空间:    

Pick an object and look closely at your chosen object as a dark silhouette against the background.    
挑选一个对象并仔细观察, 然后将其作为一个相对于背景的黑色轮廓(dark silhouette).    

Draw what you see around the object.    
画出这个对象周围你所看到的东西.    

Sit back and admire how the object interacts with its environment in a new and interesting way because you drew its negative space instead of the object itself.    
坐下来欣赏这个对象是怎样以一种'全新'和有趣的方式与其环境进行交互的, 因为你所画出的是对象的负空间(negative space)而不是对象本身.    

![47.png](https://i.loli.net/2021/01/26/Bm65FWJlzAb7HKI.png)    

### 50) 不要放弃:    

Don't stop drawing, even if you think it looks horrible.    
不要停止画画, 即使你认为它看起来很糟糕.
