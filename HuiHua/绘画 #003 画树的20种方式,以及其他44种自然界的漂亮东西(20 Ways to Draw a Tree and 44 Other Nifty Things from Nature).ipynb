### 引言    

**20 WAYS TO DRAW A TREE** is a fun and `interactive book` designed to help you explore a wide variety of approaches to drawing.    
**画树的20种方式**是一本有趣的`交互式书籍`, 旨在助你探索许多不同的绘画方式.    

It encourages `experimentation with media` and explores different ways of looking and seeing.    
本书鼓励尝试`介质试验`(使用各种材料作画), 以探索不同的绘画视角.    

Inside you will find plenty of examples to delight and inspire, and space to add your own creations.    
在本书中, 你会发现大量有趣且有启发性的例子, 以及一些空白, 用于添加你自己的创作.    

Nature is a rich `source of material` for challenging and exciting things to draw.    
自然, 为我们绘制挑战性和激动性的事物, 提供了一个丰富的`素材源`.    

In this book, forty-five themes from nature have been chosen, with each drawn in a wide variety of styles.    
在本书中, 我们选择了45个源于自然的主题, 并且每一主题下的绘画都使用了很多不同的风格.    

For each theme, the `subject matter` remains the same, but the approach to drawing it differs.    
对于每一个主题, `主题所表现的内容`是相同的, 但是其绘画方式是不同的.    

You will see twenty ways to draw each topic: Perhaps you can come up with twenty more?    
对于每一个主题, 你都可以看到20种不同的绘画方式: 或许你能想出其他更多的方式?    

Drawing, in `its most basic form`, suggests the use of paper and pencil to capture or record a subject.    
绘画, `其最基本的形式`, 就是使用纸和笔去捕捉或记录一个"表现对象".    

We may be familiar with a certain way of drawing, but there are many ways we can "draw" that make use of a much broader range of materials, media, and techniques than we normally use.    
我们可能只对某一种绘画方式特别熟悉, 但是世界上有很多的绘画方式可供选择, 比如使用更广泛的材料, 介质, 以及我们平常并不会用到的那些技巧.    

A `fine-line pen` will give clean, `crisp`, and controlled lines.    
一支`细线条笔`会给出干净, `清晰`, 且可控的线条.    

A `dip pen` and ink will give a looser line that varies in weight.    
`蘸水笔`和墨水则会使线条松散, 轻重产生变化.    

`Painterly brushstrokes` give energy and suggest movement.    
而`颜料刷的笔触`会带来活力与运动感.    

Paint spread on `acetate` and then `scratched into` before being pressed onto paper gives texture and depth.    
在将颜料压入纸上之前, 先在`醋酸纤维纸`上进行分散和`涂抹`, 可以获得纹理和深度.    

An edge of card dipped in paint and used as a printing tool gives random and unique marks.    
将蘸上颜料的卡片边缘用作绘画工具, 会产生"杂乱"和独特的痕迹.    

Each of these methods is effectively a way of "drawing" or recording the subject matter— but the results are dramatically different.    
这里的每一种绘画方式都是有效的主题记录(或主题绘画)手段, 但是其最终所呈现的效果却是截然不同的.    

The way we "see" a subject also affects our drawing.    
我们"看待"一个表现对象的视角也会影响我们的绘画.    

Our focus may vary widely.    
我们的关注点可以在很大的范围内变化.    

Perhaps we are interested in shape, texture, line quality, or color?    
或许我们对形状, 质地, 线条质量或颜色感兴趣?    

Maybe our `focus` is on the `decorative elements` of the object?    
或许我们的`兴趣点`是对象的`装饰元素`?    

When we draw, we are able to use our `personal responses` to the subject matter to highlight, enhance, ignore, or omit certain aspects that we find particularly interesting or, alternatively, unappealing.    
当我们画画的时候, 我们可以根据自己的`个人反应`来强调和加强某些我们觉得有趣的方面, 或者忽视和省略那些乏味的地方.    

For example, when we draw a tree, are we interested in the outline shape?    
比如, 当我们画一棵树的时候, 我们是对树的轮廓形状感兴趣?    

Or is it the texture of the bark and leaves?    
还是对树皮和树叶的纹理感兴趣?    

Perhaps we are focusing on the repetition of the branches?    
或许我们的关注点是枝条的重复?    

Each time our focus changes, our `drawn response` changes, too.    
每次我们的焦点改变, 我们的`绘画反应`也会跟着改变.    

###  如何使用本书    

There are twenty drawings for each theme shown in this book.    
本书中的每一主题下都有20幅绘画.    

Some are very simple; some are more complex.    
有些非常简单, 而有些则比较复杂.    

Some are `realistic` while others `are reduced to` their most basic geometric elements.    
有些`栩栩如生`, 而有些则被`简化成了`其最基本的几何元素.    

Some drawings are `rendered` accurately; others ignore `scale and proportion`.    
有些画`被渲染得`很精确, 而有些则被忽略了`尺度和比例`.    

Some approaches are bold and `angular` while in contrast, others are soft and `delicate`.    
有些画是大胆而`有棱角的`, 而另一些则是柔和而`精美的`.    

Look at the images and select those that interest you.    
翻翻这些绘画, 然后挑选出那些你感兴趣的.    

What is it about them that pulls you in?    
想想是这些画的哪一点吸引了你?    

Can you draw something similar?    
你能画出一些与此类似的绘画吗?    

Can you improve on the image?    
你能改进这些绘画方式吗?    

Some pages have been `left blank` for you to fill; others have spaces for you to draw between the images.    
那些被`留白的页面`是为了让你自己填充, 当然你也可以填充那些画与画之间的空白.    

Maybe the trees can be turned into a forest?    
或许有一天, 这些树可以变成森林?    

Maybe the shells can treble in number?    
或许这些贝壳的数量可以增加两倍?    

Perhaps the snowflakes will become a blizzard!    
或许这些雪花将会变成一场暴风雪!    

There are no right or wrong ways to draw.    
画画的方式没有正确与错误之分.    

Just as we all have our own `handwriting`, everyone has their own `hand with artwork`— a unique way of `making our marks`.    
就像我们每个人都有自己独特的`字迹`一样, 每个人都有自己的`绘迹`, 这是`我们留有自己独特印记`的一种方式.    

Try not to be `too self-conscious` about the results.    
试着别对结果`太在意`.    

Some drawings will work, and others will be less successful.    
有些画看起来不错, 而另一些就不那么如意了.    

Enjoy the process and embrace happy accidents.    
享受这个过程并拥抱那些意外的快乐.    

Most of all, have fun, be playful, and `try things out`.    
最重要的是, `尝试一下`, 并且玩得开心.    

There are many, many wonderful ways to draw a tree!    
总之, 这儿有很多很多绝妙的方式去画一棵树!

### 关于艺术家    

**Eloise Renouf** graduated with a degree in `printed textile design` from `Manchester Metropolitan University, UK`.    
**埃洛伊塞·雷诺夫**毕业于`英国曼彻斯特大都会大学`, 获得`印花面料设计`学位.    

She worked as a `fashion print designer` in studios in both London and New York before establishing her own `stationery company` in 2000.    
在2000年创立自己的`文具公司`之前, 她一直在位于伦敦和纽约的工作室担任`时装印花设计师`.    

She now designs and sells her own range of `limited-edition prints` and `fabric accessories`, whilst also undertaking `commission and licensing work` for `homewares`, stationery, and illustration.    
目前, 她在设计和销售自己的一系列`限量版印刷品`和`织物配件`, 同时也承担着`家居用品`,文具,插图的`委任和授权工作`.    

She lives and works in Nottingham, England, with her partner and three children.    
目前, 她与配偶及三个孩子共同生活在英格兰诺丁汉.
