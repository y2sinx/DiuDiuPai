## 画郁金香的20种方式    
——以及其他44种非常漂亮的花卉    

丽莎·康登(著)    

### 引言:    

Like other things in nature, most flowers have a `geometric quality` and a `naturally balanced composition`, so rendering them can be a really enjoyable and inspiring experience.    
像自然界中的其他东西一样, 大多数花卉都有其`几何特性`与`自然和谐的构成`, 因此, 对花卉的绘制工作其实是一种真正令人愉悦和鼓舞人心的体验.    

I am particularly fond of drawing them in a more abstract (less realistic) way and then `clustering them together into patterns`.    
我特别喜欢用一种更抽象(减弱其实际特征)的方式来绘制它们, 然后`将它们聚集在一起构成图案`.    

**20 Ways to Draw a Tulip** shows examples of twenty different ways to draw forty-five different flowers from around the world.    
**画郁金香的20种方式**展示了20种不同的绘制方式, 以分别绘制45种自世界各地的花卉.    

It may be important to rethink your `conception` of what it means to "draw" for this book.    
为了阅读本书, 重新思考一下你对"绘画"的`观念`可能很重要.    

Traditionally, we think of drawing as taking pencil or pen to paper and `rendering an exact likeness of something`, with perfect shading and line work.    
传统上, 我们认为绘画是用铅笔或钢笔在纸上, 用完美的阴影和线条`画出一个完全相似的东西`.    

But, for the purposes of this book, let's think of drawing as simple `line drawings`— distinct straight and curved lines and dots placed against a `plain background`.    
但是, 为了阅读本书, 让我们把画画想象成简单的`线条绘制`(线描)—— 在一个`简单的背景`上放置不同的直线, 曲线, 以及点.    

I drew all of these flowers with `fine-tipped black pens` on white paper.    
所有这些花卉均是我使用`细尖的黑色钢笔`在白纸上绘制的.

### 如何使用这本书:    

When I am trying to draw something with lines, I like to think about `the basic forms`, which are basically circles and lines.    
当我尝试着用线条来画某个东西的时候, 我喜欢思考`基本形的问题`, 即这个东西是由哪些基本的圆和线条构成.    

When `examining a drawing` of a flower, you might ask yourself, "How many lines and circles do I see here?"    
当你`审视一幅画`的时候, 你可能会问自己: "我在这上面看到了多少根线和多少个圆?"    

You might `break down` the elements of a flower.    
你可能会将一朵花的元素`拆解开`.    

"What shape are the `petals`?    
"这些花瓣的形状是什么?    

Is the flower `symmetrical`?    
这朵花是对称的吗?    

How can I render the `stamen on the interior of the flower`?    
我怎样才能画好`这朵花内部的雄蕊`呢?    

Do I want the `stem` to appear straight or `crooked`?    
我希望`花杆`看起来是直一些好还是`弯`一些好?    

Are the leaves perfectly uniform or slightly `askew`?"    
所有花的叶子是统一的还是有些歪斜的?"    

Remember that there are so many `quirks` in nature, and including those quirks in your drawings makes them much more interesting!    
自然界有众多的`古怪点`, 记住: 如果你在的你的绘画中保留了这些古怪点, 这可能会使你的画作看起来更有趣!

### 关于艺术家:    

Fine artist and illustrator **Lisa Congdon** is best known for her colorful and detailed paintings and drawings.    
**丽莎·康登**这位优秀的插画师因其色彩艳丽和富于细节的画作而为人所知.    

She enjoys `hand lettering` and pattern design, and keeps a popular daily blog about her work and life called `"Today is Going to be Awesome"`.    
她喜欢`手写字体`和图案设计, 她维护着自己的博客`"今天将会很棒"`, 这是一个热门的每日博客, 上面记录了她的工作与生活.    

Lisa's clients include Quarry Books, Chronicle Books, Simon & Schuster, and the Museum of Modern Art, among others.    
丽莎的客户包括: "采石场"出版社, "编年史"出版社, 西蒙和舒适特出版社, 以及现代艺术博物馆等.    

She lives and works in Oakland, California.    
她目前生活和工作在加利福尼亚的奥克兰.    

To see more of her work, visit Lisa's website `www.lisacongdon.com`.    
若想了解她的更多作品, 可访问她的网站: [www.lisacongdon.com](www.lisacongdon.com).
