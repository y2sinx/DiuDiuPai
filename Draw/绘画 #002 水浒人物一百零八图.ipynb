### 后记精要.  

本书后记中的汉字是按古书方式排列的(从左到右,从上大小),看起来特别的别扭,现在简单的说下其内容,你也别费那个劲看了. 顺便一提,我本以为后记中的**景仰**二字写错了,其实不然,**敬仰**亦可写作**景仰**!  

作者说他对水浒的兴趣来源于小时候的那种**带图画的火柴盒子**,当初自己好不容易收藏的**一百〇八将**却被他老爸以"不学无术"为借口给抄走了.  

作者后来经历了祖国的"文化大浩劫",只能偷偷摸摸的用别人扔掉的宣纸画自己心中的英雄好汉. 作者还将自己画好的人物装成了小册子给朋友们看,朋友们都说好. 后来这些英雄好汉也被批斗了,作者只好将它们束之高阁了.  

浩劫过后,作者应杨柳青画社之约,于是便准备将这些梁山好汉们给召集起来. 作者的绘画风格参照了明代画家陈老莲的画作<<水浒叶子>>,以及京剧中的人物形象等. 他以"一躲二学"的方法,重点绘制了那些下层人物的风貌. "一躲"是指在人物造型上力求与他人不同,"二学"则是指吸取他人在人物神态和线条组织等方面的技巧.  

这一百零八图中亦有二十余幅是由作者的儿子(戴红杰)所绘.  

当然,后记中还说了一堆其他噼里啪啦的话,我看着实在是难受,也就没继续看下去了. 若你有那个兴趣,请点击文末"阅读原文"后自行查看,此处谨提醒"请使用竖屏显示器浏览".
