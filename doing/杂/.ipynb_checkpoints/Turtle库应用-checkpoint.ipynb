## 亭桥

```python
%%disabled
from turtle import *
from ddpy.brpic import *
from ddpy.brmath import *

# 总宽与总高 
W, H= 200, 200

#大孔
_HoleX = 0.34 #0.36  
_HoleTheta = 250 #大孔半径>150

HoleX = _HoleX * W  

#坡
_SlopeY = 0.4
_SlopeTheta = 90 

SlopeX = (W - HoleX)/2
SlopeY = _SlopeY * H

# 柱子
_PillarX = 0.24 
_PillarY = .16
_PillarDx = 5
_PillarDy = 5

PillarX = _PillarX * W   # 两个柱子之间的距离(左柱左-右柱右)
PillarY = _PillarY * H   # 柱子的高度

# 亭子
_PavilionX = 0.36
_PavilionY = 0.38         
_DownY = 0.125 # 下弧的高占总高的比
_MidY = 0.6  # 注意 _UpY+_MidY>1
_UpY =  0.6 # X0的大小与(_MidY+_UpY-1)成正比
_UpTheta = 12
_MidTheta = 22
_DownTheta = 5

PavilionX = _PavilionX * W 
PavilionY = _PavilionY * H 
DownY = _DownY * PavilionY
MidY = _MidY * PavilionY
UpY =  _UpY * PavilionY  
X0 = PavilionX*(MidY-PavilionY+UpY)/(2*MidY)
Y0 = PavilionY-UpY

# 因亭总高采用比例而不是定值,所有画布高度可能有剩余
RestY = H - SlopeY - PillarY - PavilionY
HW = (H - RestY)/W 

########## 配置字典 ##########
SizeDict = {'PenColor':'grey51',
            'BgColor': 'Azure',
            'BridgeColor': 'SandyBrown',
            'PavilionColor': 'Gold1',
            'PillarColor':'red1',
            
            '_HoleX': _HoleX,
            '_HoleTheta': _HoleTheta,
            'HoleX': HoleX,
            
            '_SlopeY': _SlopeY,
            '_SlopeTheta': _SlopeTheta,
            'SlopeX': SlopeX,
            'SlopeY': SlopeY,
            
            '_PillarX': _PillarX,
            '_PillarY': _PillarY,
            '_PillarDx': _PillarDx,
            '_PillarDy':_PillarDy,
            'PillarX': PillarX,
            'PillarY': PillarY,
            
            
            '_PavilionX': _PavilionX,
            '_PavilionY': _PavilionY,
            '_DownY': _DownY,
            '_MidY': _MidY,
            '_UpY': _UpY,
            '_UpTheta': _UpTheta,
            '_MidTheta': _MidTheta,
            '_DownTheta': _DownTheta,
            'PavilionX': PavilionX,
            'PavilionY': PavilionY,
            'DownY': DownY,
            'MidY': MidY,
            'UpY': UpY,            
            'X0': X0,
            'Y0': Y0,
            
            'RestY': RestY,
            'HW': HW,
            }

pprint(SizeDict)

########## 辅助函数 #########
    
def skip2point(x=0,y=0,visible=False):
    """从当前点跳到另一点"""
    if visible==True: 
        goto(x,y)
    else:
        penup()
        goto(x,y)
        pendown()

def delta2point(dx=0,dy=0,visible=True):
    """从当前点步进dx,dy"""
    if visible==True: 
        goto(xcor()+dx,ycor()+dy)
    else:
        penup()
        goto(xcor()+dx,ycor()+dy)
        pendown()

def pavilion_arc(dx=100,dy=100,theta=20):
    """为等腰三角形的两斜边画弧.
    xy是右底角的坐标; s是三角形的半底边; h是三角形的高;theta是弧的圆心角;"""
    seth(0) # 函数内部的起始方向
    heading1 = 180-arctand(dy/dx)-theta/2
    heading2 =  90+arctand(dx/dy)+theta/2
    r = np.sqrt(dx**2+dy**2)/2 / sind(theta/2)

    seth(0)  #右弧
    left(heading1)
    circle(r,theta)
    
    seth(0)  #左弧
    right(heading2)
    circle(r,theta)
    
def slope_arc(dx=100,dy=100,theta=32,anti=True):
    """从当前点的0方向, 绘制右桥坡(左桥坡)"""
    seth(0)  # 函数内部的起始方向
    r = np.sqrt(dx**2+dy**2)/2 / sind(theta/2)
    
    heading = arctand(dy/dx)+theta/2
    anti_heading =  180-heading
    
    heading,r = (anti_heading,r) if anti==True else (heading,-r)
    
    left(heading)
    circle(r,theta)

def hole_arc(HoleX,theta):
    """从左往右绘制大孔"""
    seth(theta/2)
    r = HoleX/2 / sind(theta/2)
    circle(-r,theta)

def bridge(): 
    """桥"""
    color(PenColor,BridgeColor) 
    
    #起始状态
    skip2point() #原点
    seth(0) #向左
        
    # 绘制左坡
    seth(0)
    skip2point(-HoleX/2) #孔左
    begin_fill()
    delta2point(-SlopeX) #左坡下
    slope_arc(SlopeX,SlopeY,_SlopeTheta,False) #左坡上
    skip2point(-HoleX/2) #孔左

    # 绘制大孔
    hole_arc(HoleX,theta=_HoleTheta) 

    # 绘制右坡
    seth(0)
    skip2point(HoleX/2) #孔右 
    delta2point(SlopeX) #右坡下
    slope_arc(SlopeX,SlopeY,_SlopeTheta,True) #右坡上

    # 绘制桥面
    delta2point(-HoleX)
    end_fill()
    
    # 回到原点
    skip2point()
    seth(0)
    
    # 清除孔内多余的填充色  
    color(PenColor,BgColor)
    skip2point(-HoleX/2)
    begin_fill()
    hole_arc(HoleX,theta=_HoleTheta) 
    skip2point()
    seth(0)
    end_fill()

########## 柱子 ##########
def pillar(x=PillarX/2,y=SlopeY,dx=_PillarDx,dy=_PillarDy): 
    """(x,y)为柱子的底心坐标; dx:为柱子的直径;
    dy:在柱子高度之上附加的高度(为填补亭边缘的弯曲)"""
    color(PenColor,PillarColor) 
    
    seth(0)
    
    begin_fill()
    skip2point(x,y) #底心
    delta2point(dx/2,0) #底右
    delta2point(0,PillarY+dy) #顶右
    delta2point(-dx,0)  #顶左
    delta2point(0,-PillarY-dy) #底左
    delta2point(dx/2,0) #底心
    end_fill()

########## 亭子 ##########
def pavilion():

    color(PenColor,PavilionColor) 
    skip2point(0,SlopeY+PillarY) #中转点:亭底的中心

    begin_fill()
    # 上弧
    skip2point(X0,SlopeY+PillarY+Y0)
    pavilion_arc(X0,UpY,_UpTheta)
    skip2point(0,SlopeY+PillarY) #中转点
    # 中弧
    skip2point(PavilionX/2,SlopeY+PillarY)
    pavilion_arc(PavilionX/2,MidY,_MidTheta)
    skip2point(0,SlopeY+PillarY) # 中转点

    # 下弧
    skip2point(PavilionX/2,SlopeY+PillarY)
    pavilion_arc(PavilionX/2,DownY,_DownTheta)
    skip2point(0,SlopeY+PillarY) # 中转点
    end_fill()

    # 消除多余的填充帖
    color(PenColor,BgColor)
    begin_fill()
    skip2point(PavilionX/2,SlopeY+PillarY)
    pavilion_arc(PavilionX/2,DownY,_DownTheta)
    skip2point(0,SlopeY+PillarY) # 中转点
    end_fill()

def main():    
    setup(1000,1000,0,0)
    pensize(1)
    pencolor(PenColor)
    sc=getscreen()
    sc.bgcolor(BgColor) #背景色
    
    bridge() #桥
    pillar(-PillarX/2,SlopeY)  #右柱
    pillar(PillarX/2,SlopeY)#左柱
    pavilion()
    pprint(SizeDict)
    done()
if __name__ == "__main__":
    main()
```

## 桥亭原始版本0

```python
%%disabled
from turtle import *
from math import *

G = (sqrt(5)-1)/2

def goto_point(x=0,y=0,show_line=True):
    """从当前点直线前进到另一点,笔的默认状态为落下"""
    if show_line==True: goto(x,y)
    else:
        penup()
        goto(x,y)
        pendown()

def locating_point(tag=0,size=10,color='red'):
    """定位点"""
    dot(size,color)
    print("locating point {}: {}".format(tag,position()))

def sind(degree):
    return sin(deg2rad(degree))

def cosd(degree):
    """当参数为角度时直接转化为弧度进行计算"""
    return cos(deg2rad(degree))

def tand(degree):
    return tan(deg2rad(degree))

def skip(fun):
    penup()
    fun
    pendown()
    
def forward_skip(dis):
    penup()
    forward(dis)
    pendown()
def home_skip():
    penup()
    home()
    pendown()
    
def coord(r=300):
    # 坐标系
    pencolor('red')
    
    tracer(False)
    forward(300)
    backward(300)
    left(90)
    forward(300)
    backward(300)
    tracer(True)
    right(90)
    pencolor('black')

def pillar(x,y,dx,dy):
    #第一根柱子
    goto_point(0,0,False)
    
    color('black','red')
    begin_fill()
    goto_point(x-dx,y,False)
    goto_point(x+dx,y)
    goto_point(x+dx,y+dy)
    goto_point(x-dx,y+dy)
    goto_point(x-dx,y)
    end_fill()
    
    goto_point(0,0,False)

    
def ddpy(R=100,times=2,alpha=180,theta=45,dx=2):
    """
    R:大孔半径
    times:大孔半径与小孔半径的比
    alpha: 大小桥孔的圆心角
    theta: 亭子的底角
    dx: 柱子的半径"""
    
    r = R/times # 小孔半径
    if theta == None: theta = rad2deg(atan(1-2*G**2)) #theta=13.283时,长纵比等于1
    
    
    ## d表示横,h表示纵
    d = 2*R*sind(alpha/2) # 大石的长
    h = d*G # 大石的高
    
    d2= (h-2*r*sind(alpha/2))/2 # 大小孔的间距
    d3= h # 柱间距等于大石的高 
    
    h2= d3*G # 柱子的高
    h3= d/2*tand(theta) #亭子的高      
                     
    x = d3/2                  # 右柱的下心x=柱间距的一半
    y = h                     # 右柱的下心y等于大石的高
    
    coord()
    color('black','gray')    
    begin_fill()
    
    ########## 桥 ##########
    # >1 
    goto_point(d/2,0,show_line=False)
    # >2
    forward(d2)
    # >3
    left(alpha/2)
    circle(-r,alpha)
    # >4
    seth(0)  
    forward(d2)
    # >5
    left(90)
    circle(h,90)
    # >6
    forward(d)
    # >7
    circle(h,90)
    # >8
    seth(0)
    forward(d2)
    # >9
    seth(alpha/2)
    circle(-r,alpha)
    # >10
    seth(0)
    forward(d2)
    # >1
    seth(alpha/2)
    circle(-R,alpha)
    end_fill()
    # >0
    home_skip()
    
    ########## 亭子 ##########
    

    ## 柱子左
    pillar(-x,y,dx,h2) 
    ## 柱子右
    pillar(x,y,dx,h2) 
    
    ## 亭顶三角形
    ## 假设右弧的右半径,与三角形的底边垂直,且相交于右下角点
    
    ## 右弧: 
    color('black','yellow')
    begin_fill()
    r3=h3/2/sind(theta)**2
    
    goto_point(0,h+h2,False) # 三角形的底边中心
    forward(d/2)             # 三角形的右下角点
    seth(180)
    circle(-r3,2*theta)
    
    ##左弧
    seth(0)
    goto_point(0,h+h2,False) # 三角形的底边中心
    backward(d/2)             # 三角形的左下角点
    seth(0)
    circle(r3,2*theta)      
    end_fill()
    
    ## 回到初始位置
    goto_point(0,0,0)
    seth(0)
    
    ## LOGO长纵比
    print("长纵比: {}".format((d/2+h)/(h+h2+h2)))   

def main():
    setup(1000,700,0,0)
    pensize(2)
    ddpy()
    mainloop()
    
if __name__ == "__main__":
    main()
```

## 亭桥原始版本2

```python
%%disabled 2
from turtle import *  
from ddpy.brmath import *
from ddpy.brpic import *

Ratio_r_R = 0
Ratio_h_d = 2
Ratio_PilarSpace_d = 0.6
Ratio_PilarHeight_PilarSpace = .6
Ration_DeltaPillarHeight = 1/20

Ratio_MidArcHeight_Rest = 1/2
Ratio_DownArcHeight_Rest = 1/10
Ratio_Increase_d = 3/10

Ratio_Xc_d = 2/10
Ratio_DeltaYc = 0

# 配色 
# RGB颜色参考: http://tool.oschina.net/commons?type=3
PenColor = 'grey51'
BgColor = 'white'
StoneColor = 'Snow3'

PavilionColor = 'Gold1'
PillarColor = 'red1'

BigAngle=52


##根据click2xy()而变
# Xc,Yc=50.0,217.0

def pillar(x=0,y=0,dx=10,dy=100): #red1
    """(x,y)为柱子左下角的坐标;dx,dy为柱子的宽和高"""
    color(PenColor,PillarColor) 
    begin_fill()
    go2point(x,y) 
    go2delta(dx,0,1)
    go2delta(0,dy,1)
    go2delta(-dx,0,1)
    go2delta(0,-dy,1)
    end_fill()

def arc2(s=100,height=100,theta=20):
        """为等腰三角形的两斜边画弧.
        xy是右底角的坐标; s是三角形的半底边; h是三角形的高;theta是弧的圆心角;"""
        
        heading1 = 180-arctand(height/s)-theta/2
        heading2 =  90+arctand(s/height)+theta/2
        r = np.sqrt(s**2+height**2)/2 / sind(theta/2)
        
        seth(0)  #右弧
        left(heading1)
        circle(r,theta)
        seth(0)  #左弧
        right(heading2)
        circle(r,theta)

def click2xy(): 
    """获取屏幕上点的坐标"""
    def getPos(x,y): 
        print("Xc,Yc={},{}".format(x,y)) #只能打印而不能return
    getscreen().onclick(getPos) #或onscreenclick(getPos)  
    done()

def logo_ddpy(R=50,alpha=300,theta=20,dx=3):
    """
    R:大孔半径
    Ratio_r_R:大孔半径与小孔半径的比
    alpha: 大小桥孔的圆心角
    theta: 亭子的底角
    dx: 柱子的直径
    d表示横向距离,h表示纵向距离
    """
    
    d = 2*R*sind(alpha/2)  #石长
    P = d*(1+Ratio_Increase_d)# 亭长
    h = Ratio_h_d * d  #石高
    r = Ratio_r_R * R  #小孔半径
    hole_space = (h-2*r*sind(alpha/2))/2  #孔间距(大小孔边界之间的距离)

    pillar_space= Ratio_PilarSpace_d * P   #柱间距
    pillar_height= Ratio_PilarHeight_PilarSpace * pillar_space  #柱子的高
    
    pavilion_height = h + pillar_height #亭底的高度      
                     
    
    ###### 桥 ######
    color(PenColor,StoneColor)   #snow4, 
    seth(0)
    go2point(d/2)
    
    begin_fill()
    forward(hole_space)
    left(alpha/2)
    circle(-r,alpha)
    seth(0)  
    forward(hole_space)
    left(90)
    circle(h,90)
    forward(d)
    circle(h,90)
    seth(0)
    forward(hole_space)
    seth(alpha/2)
    circle(-r,alpha)
    seth(0)
    forward(hole_space)
    seth(alpha/2)
    circle(-R,alpha)
    end_fill()
    
    ###### 亭子 ######
    color(PenColor,PavilionColor) #Gold
    
    rest = tand(BigAngle)*(h+d/2)-pavilion_height
    height_mid = Ratio_MidArcHeight_Rest * d
    height_down = Ratio_DownArcHeight_Rest * d
    xc = Ratio_Xc_d * d
    yc = (1-Ratio_Xc_d)*height_mid
    height_up=rest-yc
    
    go2point(x=xc,y=(1+Ratio_DeltaYc)*yc+pavilion_height)
    begin_fill()
    arc2(s=xc, height=height_up, theta=theta/2)  #上弧
    
    go2point(P/2,pavilion_height)
    arc2(s=P/2, height=height_mid, theta=theta) #中弧

    go2point(P/2,pavilion_height)
    arc2(s=P/2, height=height_down, theta=theta/5) #下弧
    go2point(P/2,pavilion_height)
    end_fill()

    ## 清除不必要的填充色
    color(PenColor,BgColor) #背景色
    begin_fill() 
    go2point(P/2,pavilion_height) # 消除下弧之下的颜色
    arc2(s=P/2, height=height_down, theta=theta/5) 
    go2point(P/2,pavilion_height,)
    end_fill()

    
    ###### 柱子 ##l####
    pillar(pillar_space/2,h,dx,(1+Ration_DeltaPillarHeight)*pillar_height)  #右
    pillar(-pillar_space/2,h,dx,(1+Ration_DeltaPillarHeight)*pillar_height)  #左
    
    return rest

def main():
    setup(1000,700,0,0)
    sc=getscreen()
    sc.bgcolor(BgColor) #背景色
    pensize(1)
    hideturtle()
    d = logo_ddpy()
    click2xy()
    print("d: ",d) #长纵比=亭尖高度/2/d
    done()
    
if __name__ == "__main__":
    main()
```

## 亭桥原始版本1

```python
%%disabled 1
from turtle import *  
from ddpy.brmath import *
from ddpy.brpic import *

G = 0.6180339887498949 # (sqrt(5)-1)/2


    
def pillar(x,y,dx,dy):
    #第一根柱子
    go2point(0,0,False)
    
    color('black','red')
    begin_fill()
    go2point(x-dx,y,False)
    go2point(x+dx,y)
    go2point(x+dx,y+dy)
    go2point(x-dx,y+dy)
    go2point(x-dx,y)
    end_fill()
    
    go2point(0,0,False)
    
def arc(x=100,y=0,s=100,h=100,theta=15):
        """为等腰三角形的两斜边画弧"""
        heading1 = 180-arctand(h/s)-theta/2
        heading2 =  90+arctand(s/h)+theta/2
        r = np.sqrt(s*s+h*h)/2 / sind(theta/2)
                
        seth(0)
        left(heading1)
        circle(r,theta)
        
        seth(0)
        right(heading2)
        circle(r,theta)
    
def logo_ddpy(R=70,times=5,alpha=180,theta=15,dx=4):
    """
    R:大孔半径
    times:大孔半径与小孔半径的比
    alpha: 大小桥孔的圆心角
    theta: 亭子的底角
    dx: 柱子的半径"""
    
    r = R/times # 小孔半径
    if theta == None: theta = rad2deg(arctand(1-2*G**2)) #theta=13.283时,长纵比等于1
    
    
    ## d表示横,h表示纵
    d = 2*R*sind(alpha/2) # 大石的长
    h = d*G # 大石的高
    
    d2= (h-2*r*sind(alpha/2))/2 # 大小孔的间距
    d3= h # 柱间距等于大石的高 
    
    h2= d3*G # 柱子的高
    h3= d/2*tand(theta) #亭子的高      
                     
    x = d3/2                  # 右柱的下心x=柱间距的一半
    y = h                     # 右柱的下心y等于大石的高
    
#     coord()
    color('black','gray')    
    begin_fill()
    
    ########## 桥 ##########
    # >1 
    go2point(d/2,0,False)
    # >2
    forward(d2)
    # >3
    left(alpha/2)
    circle(-r,alpha)
    # >4
    seth(0)  
    forward(d2)
    # >5
    left(90)
    circle(h,90)
    # >6
    forward(d)
    # >7
    circle(h,90)
    # >8
    seth(0)
    forward(d2)
    # >9
    seth(alpha/2)
    circle(-r,alpha)
    # >10
    seth(0)
    forward(d2)
    # >1
    seth(alpha/2)
    circle(-R,alpha)
    end_fill()
    # >0
#     go2point()
    
    ########## 亭子 ##########
    
    ## 亭顶三角形
    ## 假设右弧的右半径,与三角形的底边垂直,且相交于右下角点
    
    color('black','yellow')
    ytz = h+h2
    go2point(50,ytz+30,0)

    begin_fill()

    arc(s=50,  h=50, theta=20)

    go2point(100,ytz,0)
    arc(s=100, h=50, theta=20)

    go2point(100,ytz,0) # 右角的坐标
    arc(s=100, h=10, theta=5)

    go2point(100,ytz,0)
    end_fill()

    color('black','white')

    begin_fill() #清除不必要的填充色
    go2point(100,ytz,0) # 右角的坐标
    arc(s=100, h=10, theta=5)
    go2point(100,ytz,0)

    end_fill()


    
    ## 回到初始位置
    go2point(0,0,0)
    seth(0)
    
    go2point(0,0,0)
    ## 柱子左
    pillar(-x,y+4,dx,h2) 
    ## 柱子右
    pillar(x,y+4,dx,h2) 
    
    ## LOGO长纵比
    print("长纵比: {}".format((d/2+h)/(h+h2)))   

def main():
    setup(1000,700,0,0)
    hideturtle()
    pensize(5)
    try: logo_ddpy()
    except Exception as e:
        print(e)
    done()
    
if __name__ == "__main__":
    main()
```

## 斐波那契螺旋线(黄金螺旋线)

```python
%%disabled
from turtle import *
#import cmath
speed(10)
pensize(5)
 
#判断较小整数的函数
def little(a,b):
    if a<b:
        return a
    else:
        return b
#初始点
#矩形
x0 = 250
y0 = 150
r= little(2*x0,2*y0)
 
#定义绘制、填充1/2正方形及扇形的函数
def square_sector(r,color_couple):
    color("black",color_couple[0])
    begin_fill()
    #顺时针画1/2正方形
    #for i in range(2):
    fd(r)
    rt(90)
    fd(r)
    end_fill()
    #顺时针画剩余1/2正方形及1/4圆弧
    color("black",color_couple[1])
    begin_fill()
    for i in range(2):
        rt(90)
        fd(r)
    rt(90)
    circle(-r,90)
    end_fill()
 
 
#黄金比例
#golden_ratio = (cmath.sqrt(5)-1)/2#can't convert complex to int
golden_ratio = 0.618
 
#5组填充色——[color_rect,color_arc]#1/2矩形，扇形
color_ls = [["#8F653B","#EDCB67"],["#B58539","#7D6A98"],\
            ["#FAC374","#56ACA9"],["#D87D84","#DA8F3C"],["#769A43","#CF4749"]]
 
pu()
goto(x0,y0)#矩形的右上顶点,默认初始角度0
pd()
seth(-90)
#循环绘制、填充矩形及扇形
for i in range(5):
    r = 300*(golden_ratio**i)#边长
    #rt(90)#"正方形圆弧图块整体"右转90度绘制
    square_sector(r,color_ls[i])

hideturtle()
done()
```

## 随机树

```python
%%disabled
from turtle import *
import random as rd
from time import sleep

def tree(branchLen):
    if branchLen > 3:
        if 8 <= branchLen <= 12:
            if rd.randint(0, 2) == 0:
                color('snow')
            else:
                color('lightcoral')
            pensize(branchLen / 3)
        elif branchLen < 8:
            if rd.randint(0, 1) == 0:
                color('snow')
            else:
                color('lightcoral')
            pensize(branchLen / 2)
        else:
            color('sienna')
            pensize(branchLen / 10)
 
        forward(branchLen)
        a = 1.5 * rd.random()
        right(20*a)
        b = 1.5 * rd.random()
        tree(branchLen-10*b)
        
        left(40*a)
        tree(branchLen-10*b)
        
        right(20*a)
        up()
        backward(branchLen)
        down()
 
 
def petal(m):  # 树下花瓣
    for i in range(m):
        a = 200 - 400 * rd.random()
        b = 10 - 20 * rd.random()
        up()
        forward(b)
        left(90)
        forward(a)
        down()
        color("lightcoral")
        circle(1)
        up()
        backward(a)
        right(90)
        backward(b)
 
 
def main():
    setup(600,600,0,0)
    screensize(bg='wheat')
    
    tracer(5, 0)
    left(90)
    up()
    backward(150)
    down()
    color('sienna')
    tree(60)
    petal(100)
main()
done()
```

## 表

```python
%%disabled
from turtle import *
from datetime import datetime

def jump(distanz, winkel=0):
    penup()
    right(winkel)
    forward(distanz)
    left(winkel)
    pendown()

def hand(laenge, spitze):
    """绘制指针;
    laeng: 指针的大线段长度(被表盘中心分隔的小线段长度为0.15);
    spitze:指针的三角形边长"""
    fd(laenge*1.15) 
    rt(90)
    fd(spitze/2.0)
    lt(120)
    fd(spitze)
    lt(120)
    fd(spitze)
    lt(120)
    fd(spitze/2.0)

def make_hand_shape(name, laenge, spitze):
    reset()
    jump(-laenge*0.15)
    begin_poly()  # 当前点为多边形的起点
    hand(laenge, spitze)
    end_poly()  #  当前点为多边形的终点
    hand_form = get_poly() #获取多变形中所有的点
    register_shape(name, hand_form) # 在海龟的形状列表(TurtleScreen's shapelist)中添加指针形状

def clockface(radius):
    """设置表盘的12分刻度线(时刻度线)与60分刻度点(秒刻度)"""
    reset()
    pensize(7)
    for i in range(60):
        jump(radius)
        if i % 5 == 0:
            fd(25)  # 12分刻度线的长度为
            jump(-radius-25)
        else:
            dot(3)  # 60分刻度点的大小
            jump(-radius)
        rt(6)

def setup():
    global second_hand, minute_hand, hour_hand, writer
    mode("logo")
    make_hand_shape("second_hand", 125, 25)
    make_hand_shape("minute_hand",  130, 25)
    make_hand_shape("hour_hand", 90, 25)
    clockface(160)
    # 秒针海龟
    second_hand = Turtle()
    second_hand.shape("second_hand")
    second_hand.color("blue3", "blue3")
    # 分针海龟
    minute_hand = Turtle()
    minute_hand.shape("minute_hand")
    minute_hand.color("green3", "green3")
    # 时针海龟
    hour_hand = Turtle()
    hour_hand.shape("hour_hand")
    hour_hand.color("red3", "red3")
    for hand in second_hand, minute_hand, hour_hand:
        hand.resizemode("user")
        hand.shapesize(1, 1, 3)
        hand.speed(0)
    ht() # 隐藏海龟?
    writer = Turtle()
    #writer.mode("logo")
    writer.ht() #
    writer.pu()
    writer.bk(85)

def wochentag(t):
    wochentag = ["Monday", "Tuesday", "Wednesday",
        "Thursday", "Friday", "Saturday", "Sunday"]
    return wochentag[t.weekday()]  # 将数字转化为字符串

def datum(z):
    monat = ["Jan.", "Feb.", "Mar.", "Apr.", "May", "June",
             "July", "Aug.", "Sep.", "Oct.", "Nov.", "Dec."]
    j = z.year
    m = monat[z.month - 1]
    t = z.day
    return "{} {} {}".format(m,t,j)

def tick():
    t = datetime.today()
    sekunde = t.second + t.microsecond*0.000001
    minute = t.minute + sekunde/60.0
    stunde = t.hour + minute/60.0
    try:
        tracer(False)  # Terminator can occur here
        writer.clear()
        writer.home()
        writer.forward(65)
        writer.write(wochentag(t),align="center", font=("Courier", 14, "bold"))
        writer.back(150)
        writer.write(datum(t),align="center", font=("Courier", 14, "bold"))
        writer.forward(85)
        tracer(True)
        second_hand.setheading(6*sekunde)  # 秒针和分针的角度都是6的倍数
        minute_hand.setheading(6*minute)  
        hour_hand.setheading(30*stunde)  # 时针的角度是360/12=30的倍数
        tracer(True)
        ontimer(tick, 1000) # 每隔1s运行一次tick(),秒针表现为走走停停的效果;如果想要表现秒针连续走的效果,可以将其设置为100
    except Terminator:
        pass  # turtledemo user pressed STOP

def main():
    tracer(False)
    setup()
    tracer(True)
    tick()
    return "EVENTLOOP"

if __name__ == "__main__":
    mode("logo")
    msg = main()
    print(msg)
    mainloop()
```

## 阴阳鱼

```python
%%disabled
from turtle import *

def yin(radius, color1, color2):
    width(3)
    color("black", color1)
    begin_fill()
    circle(radius/2., 180) # 阴阳鱼的第一个半圆
    circle(radius, 180)  # 阴阳鱼的第二个半圆
    left(180)
    circle(-radius/2., 180) # 阴阳鱼的第三个半圆
    end_fill()
    left(90)
    up()
    forward(radius*0.35) # 小白圆的上顶点:(0.5-0.15)*radius
    right(90)
    down()
    color(color1, color2)
    begin_fill()
    circle(radius*0.15)
    end_fill()
    left(90)
    up()
    backward(radius*0.35) # 回到原点
    down()
    left(90) # 在原点调整为初始方向的逆方向,可使阴阳鱼共用一段程序

def main():
    reset()
    yin(200, "black", "white")
    yin(200, "white", "black")
    ht()
    return "Done!"

if __name__ == '__main__':
    main()
    mainloop()
```
