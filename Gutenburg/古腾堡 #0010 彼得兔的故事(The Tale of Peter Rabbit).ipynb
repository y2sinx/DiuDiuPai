**O**nce upon a time there were four little rabbits, and their names were—Flopsy, Mopsy, Cotton-tail and Peter.  
从前有四只小兔子,他们的名字分别是—Flopsy, Mopsy, Cotton-tail 和 Peter(彼得).

They lived with their mother in a **sand-bank**, underneath the root of a very **big fir-tree**.  
他们和妈妈一起住在一棵**大杉树**下的树根**沙堆**中.

![08](http://www.gutenberg.org/files/14838/14838-h/images/peter08.jpg)

---

"Now, my dears," said old Mrs. Rabbit one morning, "You may go into the fields or down the lane, but don't go into Mr. McGregor's garden. Your father had an accident there; he was put in a pie by Mrs. McGregor."  
"现在,孩子们," 有一天早上兔妈妈说, "你们可以去田野或小路边玩了,但是绝对不要去麦格雷戈先生的菜园啊. 你们的父亲就是在那里出事的,他已经被麦格雷戈夫人做成馅饼了."

![11](http://www.gutenberg.org/files/14838/14838-h/images/peter11.jpg)

---

Now run along and don't get into **mischief**. I am going out."  
现在,去玩吧,孩子们,好好听话,不要**调皮捣蛋**. 我要出门去买东西了."

![12](http://www.gutenberg.org/files/14838/14838-h/images/peter12.jpg)

---

Then old Mrs. Rabbit took a basket and her umbrella and went through the wood to the baker's. She bought a loaf of brown bread and five **currant buns**.  
之后,兔妈妈拿着她的篮子和雨伞,去树林对面的面包店了. 她买了一条棕面包和五个**醋栗面包**.

![15](http://www.gutenberg.org/files/14838/14838-h/images/peter15.jpg)

---

Flopsy, Mopsy and Cotton-tail who were good little bunnies went down the lane together to gather blackberries.  
Flopsy, Mopsy 和 Cotton-tail 都是听妈妈话的乖兔子,他们一起去小路边摘黑莓了.

![16](http://www.gutenberg.org/files/14838/14838-h/images/peter16.jpg)

---

But Peter who was very naughty, ran straight away to Mr. McGregor's garden and squeezed under the gate!  
但是彼得兔非常顽皮,他直接跑向了麦格雷戈先生的菜园,并在门口挤了进去!

![19](http://www.gutenberg.org/files/14838/14838-h/images/peter19.jpg)

---

First he ate some **lettuces** and some French beans. And then He Ate Some **Radishes**.  
首先,他吃了一些**生菜**和一些法国豆,然后他吃了一些**小萝卜**.

![20](http://www.gutenberg.org/files/14838/14838-h/images/peter20.jpg)

---

And then, feeling rather sick, he went to look for some **parsley**.  
之后,他感觉非常恶心,然后想去找些**香芹**吃.

![23](http://www.gutenberg.org/files/14838/14838-h/images/peter23.jpg)

---

But round the end of a cucumber frame, whom should he meet but Mr. McGregor!  
但是在香芹在黄瓜架的尽头,要想去那儿一定会碰到麦格雷戈先生!

![24](http://www.gutenberg.org/files/14838/14838-h/images/peter24.jpg)

---

Mr. McGregor was on his hands and knees planting out young cabbages, but he jumped up and ran after Peter, waving a **rake** and calling out "Stop thief!"  
麦格雷戈先生跪在地里种着嫩白菜,突然,他发现了彼得,然后他立马站了起来,然后开始追彼得,并挥舞着**耙子**,嘴里不停地喊着 "停下,你这个小偷!".

![27](http://www.gutenberg.org/files/14838/14838-h/images/peter27.jpg)

---

Peter was most dreadfully frightened; he rushed all over the garden, for he had forgotten the way back to the gate.  
彼得害怕极了,他在菜园里到处乱撞,因为他忘记了回大门的路.

He lost one shoe among the cabbages, and the other amongst the potatoes.  
他在白菜地里丢了第一只鞋,然后土豆地里丢了另一只鞋.

![28](http://www.gutenberg.org/files/14838/14838-h/images/peter28.jpg)

---

After losing them, he ran on four legs and went faster  
鞋丢了之后,他用四条腿逃跑的更快了.

So that I think he might have got away altogether if he had not unfortunately run into a **gooseberry** net.  
因此,我认为,如果不是他不小心闯进了**醋栗**网,他很可能就已经成功的逃跑了.

And got caught by the large buttons on his jacket.  
但是,他外套上的大纽扣被醋栗网勾住了,因此他没能逃脱.

It was a blue jacket with brass buttons, quite new.  
这些黄铜纽扣是他身上正穿着的那件崭新的蓝色外套上的.

![31](http://www.gutenberg.org/files/14838/14838-h/images/peter31.jpg)

---

Peter gave himself up for lost and **shed big tears**;  
彼得找不到出去的路了,他非常难过,**哭了起来**.

But his sobs were overheard by some friendly sparrows.  
但是他的啜泣声被一些友好的麻雀们听见了.

Who flew to him in great excitement and **implored** him to **exert himself**.  
他们兴奋的飞向他,并**鼓励**他要**尽自己最大的努力**.

![32](http://www.gutenberg.org/files/14838/14838-h/images/peter32.jpg)

---

Mr. McGregor came up with a **sieve** which he intended to pop on the top of Peter, but Peter **wriggled** out just in time. Leaving his jacket behind him.  
这时麦格雷戈先生拿来了一个**筛子**,打算用他扣住彼得,但是彼得**扭动了一下身体**,成功逃脱了. 但是他的夹克却被丢在了自己身后.

![35](http://www.gutenberg.org/files/14838/14838-h/images/peter35.jpg)

---

He rushed into the **tool-shed** and jumped into a can. It would have been a beautiful thing to hide in, if it had not had so much water in it.  
他冲进**工具棚**,然后跳进了一个喷水壶. 如果喷水壶里面没有太多水的话,那一定是个绝佳的藏身之处.

![36](http://www.gutenberg.org/files/14838/14838-h/images/peter36.jpg)

---

Mr. McGregor was quite sure that Peter was somewhere in the tool-shed, perhaps hidden underneath a **flower-pot**.  
麦格雷戈先生非常确定彼得就躲在工具棚里的某个地方,也许就藏在某个**花盆**的下面.

He began to turn them over carefully, looking under each.  
他开始小心翼翼地翻花盆,一个一个的查看.

Presently Peter sneezed "Kertyschoo!"  
突然,彼得打了个喷嚏 "阿..球...!"

![39](http://www.gutenberg.org/files/14838/14838-h/images/peter39.jpg)

---

Mr. McGregor was after him in no time, and tried to put his foot upon Peter, who jumped out of a window, **upsetting** three plants.  
麦格雷戈先生立马就发现了他,并试图抓住他,但是彼得迅速的跳出了窗户,还**打翻了**三个花盆.

The window was too small for Mr. McGregor, and he was tired of running after Peter. He went back to his work.  
这窗户对麦格雷戈先生来说太小了,并且他厌烦了追着彼得跑. 于是继续去工作了.

![40](http://www.gutenberg.org/files/14838/14838-h/images/peter40.jpg)

---

Peter sat down to rest; he was out of breath and trembling with fright, and he had not the least idea which way to go.  
彼得坐了下来休息,气喘吁吁地,害怕的直发抖,他确实不知道该怎么走了.

Also he was very damp with sitting in that can.  
而且他身上湿乎乎的,因为之前藏在那个水壶中的缘故.

After a time he began to wander about, going lippity—lippity—not very fast, and looking all around.  
过了一会儿,他又开始到处蹦跶,蹦的不是很快,并且还在不停的四处张望着.

![43](http://www.gutenberg.org/files/14838/14838-h/images/peter43.jpg)

---

He found a door in a wall; but it was locked and there was no room for a fat little rabbit to squeeze underneath.  
他在墙上找到了一扇上了锁的门,门下的空间很小,彼得根本无法挤过去.

An old mouse was running in and out over the stone doorstep, carrying peas and beans to her family in the wood. Peter asked her the way to the gate but she had such a large pea in her mouth she could not answer. She only shook her head at him.  
一只老鼠在石阶上进进出出, 正在给她树林里的家人送豆子. 彼得问她去大门的路怎么走,但她的嘴里咬着豌豆,无法回答. 她只是对他摇了摇头.

Peter began to cry.  
然后,彼得哭了.

![44](http://www.gutenberg.org/files/14838/14838-h/images/peter44.jpg)

---

Then he tried to find his way straight across the garden, but he became more and more puzzled. Presently he came to a pond where Mr. McGregor filled his water-cans. A white cat was staring at some gold-fish; she sat very, very still, but now and then the tip of her tail **twitched** as if it were alive. Peter thought it best to go away without speaking to her.  
之后,他试图找到穿过菜园的路,但是他越来越困惑. 现在,他来到了一处池塘边,麦格雷戈先生正在那给水壶装水. 一只白色的猫正盯着一些金鱼,她正在非常安静地坐着,但是她的尾巴时不时的**扭动着**,看起来好像只有她的尾巴是活的一样. 彼得认为最好还是不要跟她说话,然后就走开了.

He had heard about cats from his cousin, little Benjamin Bunny.  
他是从他的表哥 Benjamin Bunny 那里听说的,关于猫这种动物的.

![47](http://www.gutenberg.org/files/14838/14838-h/images/peter47.jpg)

---

He went back towards the tool-shed, but suddenly, quite close to him, he heard the noise of a hoe—scr-r-ritch, scratch, scratch, scritch.  
他又回到了工具棚,但是突然,在他附近,他听到了些刮擦,刮擦,乱挠的声音.

Peter **scuttered** underneath the bushes, but presently as nothing happened.  
彼得吓得**跑进了**灌木丛,但是什么事也没有发生.

he came out and climbed upon a **wheelbarrow** and peeped over.  
他从灌木丛中出来后,爬进了一个**手推车**,偷偷的往外看去.

The first thing he saw was Mr. McGregor **hoeing** onions. His back was turned towards Peter and beyond him was the gate!  
他看到麦格雷戈先生正在背对着他**刨**洋葱. 并且在麦格雷戈先生的远处,他看到了进来时的大门!

![48](http://www.gutenberg.org/files/14838/14838-h/images/peter48.jpg)

---

Peter got down very quietly off the wheel-barrow and started running as fast as he could go, along a straight walk behind some black currant bushes. Mr. McGregor caught sight of him at the corner, but Peter did not care. He slipped underneath the gate and was safe at last in the wood outside the garden.  
彼得非常小心的下了手推车,沿着黑醋栗灌木丛后面的一条小路,开始拼命的跑. 麦格雷戈先生在拐角处看见了他,但彼得并不关心. 他随后就从大门下钻了出去, 最终安全地跑进了菜园之外的树林.

![51](http://www.gutenberg.org/files/14838/14838-h/images/peter51.jpg)

---

Mr. McGregor hung up the little jacket and the shoes for a **scare-crow** to frighten the blackbirds.  
麦格雷戈先生用彼得的小外套和小鞋子做了一个"**稻草人**",以此用来吓唬那些黑鹂.

Peter never stopped running or looked behind him, till he got home to the big fir-tree.  
彼得不停地奔跑,全然不顾他的身后,直到他回到那棵大枞树下(他才停下来休息的).

![52](http://www.gutenberg.org/files/14838/14838-h/images/peter52.jpg)

He was so tired that he **flopped** down upon the nice soft sand on the floor of the rabbit hole, and shut his eyes. His mother was busy cooking; she wondered what he had done with his clothes.  
他太累了,所以**一屁股坐进了**柔软的兔洞土堆中,并且闭上了他的眼睛. 他母亲正忙着做饭,她很好奇他的衣服哪里去了(他对他的衣服做了什么).

It was the second little jacket and pair of shoes that Peter had lost in a **fortnight**!  
这已经是彼得**两周**内丢掉的第二件小外套和第二双小鞋子了!

![55](http://www.gutenberg.org/files/14838/14838-h/images/peter55.jpg)

---

I am sorry to say that Peter was not very well during the evening. His mother put him to bed and made some **camomile tea**; and she gave a dose of it to Peter! "One **teaspoonful** to be taken at bedtime."  
我很抱歉这么说,彼得在夜晚的时候并不怎么舒服. 他的妈妈把他放到了床上,并煮了些**甘菊茶**,然后给他喂了一剂,并说着 "吃了药就要好好睡觉觉.(睡前服用**一勺**.)"

![57](http://www.gutenberg.org/files/14838/14838-h/images/peter57.jpg)

---

But Flopsy, Mopsy and Cotton-tail had bread and milk and blackberries for supper.  
但是, Flopsy, Mopsy and Cotton-tail 在晚饭的时候吃了面包,牛奶和黑莓.

![58](http://www.gutenberg.org/files/14838/14838-h/images/peter58.jpg)
