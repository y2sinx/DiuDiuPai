[Introduction.]()    

June 30th. 96.    
6 月 30 日. 96.    

My Dear Tuer    
我亲爱的图尔    

Here is the last of the Gutter snipe drawings and sorry I am to leave them! Children of the gutter roam about free and are often hungry, but what would one give for such appetites? You and I smoke big cigars while they all too soon, poor little chaps smoke what you and I and others throw away. Sometimes I wonder whether they don't lead the happier lives?    
这是最终的一些贫民区图画,并且我很抱歉我离开了他们! 贫民区的孩子们到处流浪,常常忍饥挨饿,但对于这样的欲望,人们会付出什么呢? 我们把还未抽完的烟和雪茄扔掉后,那些贫穷而可怜的人才会有的抽. 有时我想知道他们是否能够过上更幸福的生活?    

Yours always    
你永远的    

Phil May    
菲尔⋅梅
