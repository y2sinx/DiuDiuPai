## The Owl and The Pussy-Cat.    

The Owl and the Pussy-Cat went to sea    
In a beautiful pea-green boat.    
They took some honey, and plenty of money Wrapped up in a **five-pound** note.    
猫头鹰和猫咪乘着一条豆绿色的漂亮小船出海了. 他们带了一些蜂蜜,以及包在一张**五磅纸币**内的很多钱.

![1](http://www.gutenberg.org/files/20113/20113-h/images/image_01.jpg)    
********************************    

The Owl looked up to the stars above,    
And sang to a small guitar,    
"O lovely Pussy! O Pussy, my love.    
What a beautiful Pussy you are,    
You are, You are!    
What a beautiful Pussy you are!"    
猫头鹰仰望着星空,并弹着小吉他唱道: "噢,可爱的猫咪! 噢,猫咪,我的爱人. 你是一只多么漂亮的猫咪啊! 你是,你是! 你是最漂亮的猫咪!"    

![2](http://www.gutenberg.org/files/20113/20113-h/images/image_02.jpg)    
********************************    

Pussy said to the Owl, "You elegant fowl!    
How **charmingly** sweet you sing!    
O let us be married, too long we have **tarried**:    
But what shall we do for a ring?"    
猫咪对猫头鹰说: "你这只优雅的小鸟! 你唱的是如此的**动听**! 哦,就让我们结婚吧! 我们已经**耽搁**的太久了: 但是,我们应该(到哪)去找一个戒指呢?"  

![3](http://www.gutenberg.org/files/20113/20113-h/images/image_03.jpg)    
********************************    

They sailed away for a year and a day,    
To the land where the Bong-tree grows.    
And there in a wood a **Piggy-wig** stood, With a ring at the end of his nose.    
His nose, His nose,    
With a ring at the end of his nose.    
他们航行了一年零一天,来到了奉树生长着的土地. 在一片树林里,他们遇到了一位带着**假发的猪**,在他的鼻端挂着一个戒指. 看他的鼻子,看他的鼻子,他的鼻端挂着一个戒指.    

![4](http://www.gutenberg.org/files/20113/20113-h/images/image_04.jpg)    
********************************    

"Dear Pig, are you willing to sell your ring for one **shilling**?"    
Said the Piggy, "I will."    
猫咪问: "亲爱的猪,我可以用**一先令**买你的戒指么?" 猪说,"可以."    

![5](http://www.gutenberg.org/files/20113/20113-h/images/image_05.jpg)    
********************************    

So they took it away, and were married next day    
By the Turkey who lives on the hill.    
They dined on **mince**, and slices of **quince** Which they ate with a **runcible** spoon.    
他们把戒指带走了,并在第二天结了婚,婚礼是由住在山上的火鸡(牧师)主持的. 他们用**三齿勺**吃了一些**肉碎**和**榅桲(wēn po)片**.    

![6](http://www.gutenberg.org/files/20113/20113-h/images/image_06.jpg)    
********************************    

And hand in hand, on the edge of the sand,    
They danced by the light of the moon.    
The moon, The moon,    
They danced by the light of the moon.    
他们手牵着手,在月色的沙滩上,跳起了舞. 在月色中,在月色中,他们跳起了舞.    

![7](http://www.gutenberg.org/files/20113/20113-h/images/image_07.jpg)    
********************************    

## The Duck and The Kangaroo.    

Said the Duck to the Kangaroo,    
"Good gracious! how you **hop**    
Over the fields and the water too,    
As if you never would stop!    
鸭子对袋鼠说: "你在田野和水洼边**跳**的好漂亮啊! 就好像你永远都不会停止似的!    

![8](http://www.gutenberg.org/files/20113/20113-h/images/image_08.jpg)    
********************************    

My life is a **bore** in this nasty pond,    
And I long to go out in the world beyond!  I wish I could hop like you!"    
我这一生都在这个肮脏的小池塘里,我的生活好**无聊**啊,我好想去外面的世界看看呐! 我渴望能像你一样跳跃!"    

![9](http://www.gutenberg.org/files/20113/20113-h/images/image_09.jpg)    
********************************    

"Please give me a ride on your back!" Said the Duck to the Kangaroo.    
"I would sit quite **still**, and say nothing but '**Quack**,'    
The whole of the long day through!    
鸭子对袋鼠说: "求求你,让我在你的背上搭个车吧! 我会坐在你的背上**一动不动**,除了说一个'**嘎**'啥也不说. 漫长的一天然后就会结束.    

![10](http://www.gutenberg.org/files/20113/20113-h/images/image_10.jpg)    
********************************    

And we'd go to the Dee, and the Jelly Bo Lee, over the land, and over the sea.    
Please take me a ride! O do!"    
让我们跨过陆地和海洋,去看'迪和杰利波利'吧. 让我搭一下吧,求求你!"    

![11](http://www.gutenberg.org/files/20113/20113-h/images/image_11.jpg)    
********************************    

"This requires some little **reflection**;    
Perhaps on the whole it might bring me luck,    
And there seems but one **objection**,    
Which is, if you'll let me speak so **bold**,    
Your feet are unpleasantly wet and cold,    
And would probably give me the roo-matiz." Said the Kangaroo.    
袋鼠对鸭子说: "我容我**考虑**一下下. 或许在整体上这件事可能会给我带来好运,但是仍然有一个**拒绝的理由**,如果你允许我这么**大胆**地说的话,那就是,你的脚又湿又冷,可能会让我感到很不舒服."    

![12](http://www.gutenberg.org/files/20113/20113-h/images/image_12.jpg)    
********************************    

Said the Duck,"As I sate on the rocks, I have thought over that completely, and I bought four pairs of **worsted** socks which fit my **web-feet** neatly.    
鸭子说: "当我坐在石头上时,我就已经完全考虑过这件事了. 我买了四双**精纺的**袜子,正好适合我的**脚**.    

![13](http://www.gutenberg.org/files/20113/20113-h/images/image_13.jpg)    
********************************    

And to keep out the cold I've bought a **cloak**, and every day a cigar I'll smoke.    
All to follow my own dear true love of a Kangaroo!"    
为了防寒,我还买了一件**斗篷**,并且我每天还会抽一支雪茄. 而其他所有的事都遵从我亲爱而真诚的袋鼠朋友!"    

![14](http://www.gutenberg.org/files/20113/20113-h/images/image_14.jpg)    
********************************    

Said the Kangaroo,"I'm ready!    
All in the moonlight pale;    
But to balance me well, dear Duck, sit steady!    
And quite at the end of my tail!"    
袋鼠说: "我准备好了! 世间的万物都披上了银白色的月光. 但是要平衡好我,亲爱的鸭子,请在我的尾巴端上坐好!"    

![15](http://www.gutenberg.org/files/20113/20113-h/images/image_15.jpg)    
********************************    

So away they went with a hop and a bound, And they hopped the hole world three times round;    
And who so happy, —O who,    
As the Duck and the Kangaroo?    
他们一蹦一跳的走了, 他们围着整个世界跳了三圈. 现在,还有谁能像鸭子和袋鼠一样那么开心呢?    
********************************
