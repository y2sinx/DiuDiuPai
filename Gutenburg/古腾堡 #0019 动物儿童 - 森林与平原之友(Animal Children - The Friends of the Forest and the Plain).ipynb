### 前言  

When God made the world he planted the flowers and the grass and the trees to make things beautiful to look upon; He **swung** the sun and the moon and the stars in the sky to make things bright; He put the birds in the trees to fill the air with music, and when He made the animals we believe that he intended them to be the friends of man.  
上帝创造世界之时,为了让万物变得漂亮,他种下了花草树木. 为了使万物明亮,他在天上**挂了**太阳(以及月亮和星星). 为了使空中充满音乐,他把鸟儿放在了树上. 我们相信,他创造了动物,是为了让他们成为我们的朋友.  

Why, isn't the dog the best playmate that a boy can have? Did any one ever hear of Towser or Gyp being false friends? And the soft, **dainty/秀丽的**, **cunning/精巧的** bit of a fluffy ball of a kitten who comes rubbing its **downy/毛绒绒的** sides against the tiny girl's skirts begging for a return **caress/爱抚**, is there a play-fellow more lovable? And the squirrel who comes begging at the window for nuts; the bunny rabbit who **snuggles/依偎着** its delicate nose, trustingly, under the little boy's chin; the horse who has been man's friend in times of trouble and of peace, bearing his burdens or **scampering/奔跑** with him over the fields and roads in play; the cow who has sent her good milk to the babies of all time; the sheep and the goats who have given of their wool to keep us warm, —we love them all dearly.  
为什吗? 难道一个 BOY 不喜欢可爱的狗狗,一个 GIRL 不喜欢毛绒绒的小猫,而更喜欢 "抠脚大汉" 吗? 难道小松鼠,小兔子不讨人喜欢,小马不忠诚吗? 难道奶牛产的奶不香,绵羊产的毛不暖吗? 当然,我们喜欢动物.  

In this volume we have tried to make friends and playmates of all of the animals. You have loved the "Flower Children" and the "Bird Children" whom the publisher already has made your playmates. We feel that you are going to be just as happy to know the "Animal Children." Therefore we add to "The Little Cousins of the Field and Garden" and "The Little Playmates of the Flower Children," this volume—"The Friends of the Forest and the Plain."  
你或许已经与 "Flower Children" 或 "Bird Children" 中的朋友见过面了,现在让我们一起来与 "Animal Children" 中的动物成为朋友.
