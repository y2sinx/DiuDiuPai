# The Three Little Kittens    

By    

**Kate Greenaway**    

---    

Three little kittens    
Lost their **mittens**,    
And they began to cry,    
"Oh, mammy dear,    
We sadly fear    
Our mittens we have lost!"    
"What! lost your mittens,    
You naughty kittens;    
Then you shall have no pie!"    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
三只小猫    
弄丢了**手套**,    
哭着对妈妈说:    
"哦,亲爱的妈妈,    
我们好担心啊    
我们把手套给弄丢了!"    
"啥! 把手套弄丢了,    
你们这群顽皮的小猫;    
那今天晚上你们就别想吃饭了!"    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

The three little kittens    
Then sought their mittens,    
Upon the table high;    
Indoors and out    
They **scampered about**,    
For they were very **spry**;    
Now high, now low,    
The three in a row,    
And oh! how they made things fly.    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
三只小猫,    
开始找她们的手套,    
在桌子上找,    
在屋的里里外外找,    
她们**四处蹦跶**,    
因为她们都很**敏捷**.    
一会儿蹿上,一会儿蹿下,    
一会排成一行,    
哈! 她们把东西都弄飞了.    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

![ill_003_th](http://www.gutenberg.org/files/31342/31342-h/images/ill_003_th.jpg)    

The three little kittens    
Found their mittens,    
And they began to cry,    
"Oh mammy dear,    
See here, see here,    
Our mittens we have found!"    
"What! found your mittens,    
You darling kittens;    
Then you shall have some pie."    
Purr, **purr**, purr, purr,    
Purr, purr, purr, purr.    
三只小猫    
发现了她们的手套,    
她们高兴地对妈妈说,    
"噢,亲爱的妈妈,    
快看这里,看这里,    
我们找到了弄丢的手套!"    
"啥! 找到了你们的手套,    
我的宝贝猫猫;    
那今天晚上你们可以吃饭了."    
咕噜,**咕噜**,咕噜,咕噜,    
咕噜,咕噜,咕噜,咕噜.    

The three little kittens    
Put on their mittens,    
And soon ate up the pie.    
"Oh mammy dear,    
We sadly fear    
Our mittens we have **soiled**."    
"What! soiled your mittens,    
You naughty kittens!"    
Then they began to sigh,    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
三只小猫    
戴上了手套,    
然后很快就吃完了馅饼.    
"哦,亲爱的妈妈,    
我们好担心啊    
我们把手套给**弄脏了**."    
"啥! 弄脏了你的手套,    
你们这群顽皮的小猫!"    
然后她们开始叹气,    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

The three little kittens    
Washed their mittens,    
And hung them up to dry.    
"Oh, mammy dear,    
Look here, look here,    
Our mittens we have washed!"    
"What! washed your mittens,    
You good little kittens!    
But I smell a rat close by!    
Hush! hush!" Miew, miew,    
Miew, miew, miew, miew.    
三只小猫    
洗了她们的手套,    
把它们挂起来然后给晾干了.    
"哦,亲爱的妈咪,    
看这里,看这里,    
我们把弄脏的手套给洗了!"    
"啥! 洗了你们的手套,    
你们真是妈妈的好猫猫!    
嘘!嘘! 我好像听到附近有一只大老鼠!"    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

![ill_004_th](http://www.gutenberg.org/files/31342/31342-h/images/ill_004_th.jpg)    

These kittens so gay    
Were invited one day    
To feast by a running stream,    
Where they had as much meat    
As they wanted to eat,    
And plenty of nice ice-cream;    
And each went to sleep    
Curled up in a heap    
And had a most lovely dream.    
Purr, purr, purr, purr.    
Purr, purr, purr, purr.    
三只小猫非常的开心    
有一天被邀请    
去一个流动的小溪旁参加盛宴,    
她们吃了好多肉,    
因为想吃多少有多少,    
她们还吃了好多好多好吃的冰淇淋;    
吃饱后她们就都睡觉了    
缩成了一堆    
然后做了一个非常香甜的美梦.    
咕噜,咕噜,咕噜,咕噜,    
咕噜,咕噜,咕噜,咕噜.    

One night in the Fall    
They went to a ball,    
And danced to a lively tune,    
With a leap and a bound    
And a **merry-go-round**,    
And the sound of a big **bassoon**;    
And with holes in their mittens    
These careless kittens    
Came home by the light of the moon.    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
秋天的一晚    
她们参加了舞会,    
在欢快的音乐中跳舞,    
做着蹦蹦跳跳的姿势,    
还做着"**旋转木马**"的姿势,    
还聆听着**大低音管**的声音.    
然后把她们的手套都给磨出洞了    
这些粗心的小猫    
最后被月光送回了家.    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

These kittens 'twas said    
Were soon to be wed;    
The cards had been out some days;    
And **cat-birds**, no doubt,    
Spread the news about    
As they flew o'er the great **high-ways**;    
And cats, one and all,    
The great and the small,    
Were loud in the kittens' praise.    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
据说,这些小猫    
很快就要结婚了;    
把婚礼邀请卡都给发出去了.    
毫无疑问,是**猫鹊**    
散布了这个消息    
当他们飞越这些**公路**的时候.    
所有的猫猫,    
无论大的还是小的,    
都在大声的称赞.    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

At last came the day,    
And in splendid array    
The guests soon began to arrive,    
The aunts and the cousins    
By sixes and dozens,    
All buzzing like bees in a hive;    
And among them Sir Rouser,    
A famous old **mouser**,    
And the handsomest **Maltese** alive.    
Purr, purr, purr, purr,    
Purr, purr, purr, purr.    
这一天终于来了,    
以一种华丽的阵势    
客人们很快就到了,    
阿姨和堂兄妹们    
一批一批的,    
就像蜂巢里的蜜蜂一样嗡嗡嗡;    
在这些人当中,罗瑟爵士,    
是一个著名的**捕鼠达人**,    
当然,潇洒的**马尔他犬**也来了.    
咕噜,咕噜,咕噜,咕噜,    
咕噜,咕噜,咕噜,咕噜.    

![ill_007_th](http://www.gutenberg.org/files/31342/31342-h/images/ill_007_th.jpg)    

Then after the marriage    
Each groom called his carriage,    
And, oh, they **rode off** in fine style;    
The brides **beaming** brightly,    
And bowing politely,    
To friends every once in a while,    
Who kept up a **squalling**    
And great **caterwauling**    
That might have been heard for a mile.    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
婚礼结束后,    
每位新郎都叫来了自己的马车    
然后带着新娘开心的**离开了**.    
新娘们**灿烂地**笑着 ,    
有礼貌地鞠躬,    
跟每一位亲人朋友告别,    
他们**哭哭啼啼**的    
**幸福的抱怨着**    
这些声音在一英里之外都可以听到.    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

![ill_008_th](http://www.gutenberg.org/files/31342/31342-h/images/ill_008_th.jpg)    

The guests kept on dancing,    
Now leaping and **prancing**;    
The band still continued to play;    
And "Puss-in-the-corner,"    
And "Little Jack Homer,"    
Were games very much in their way;    
With singing and **screeching**,    
And laughter far-reaching,    
They had a good time, I dare say.    
Miew, miew, miew, miew,    
Miew, miew, miew, miew.    
客人们继续跳着舞,    
**蹦蹦跳跳的**;    
乐队依然继续演奏着.    
他们还玩着"抢位置游戏,"    
和"小杰克 ⋅ 荷马游戏,"    
他们都以自己的方式游戏着.    
我敢说,    
伴着歌声和**尖叫声**,    
还有深远的笑声,    
他们度过了愉快的时光.    
喵,喵,喵,喵,    
喵,喵,喵,喵.    

The three pretty brides,    
And their husbands besides    
Took rooms in a very nice flat;    
Not a rat nor a mouse    
Was e'er seen in the house,    
Nor any one heard to cry Scat!    
So they lived and looked pleased—    
They were **petted** not teased—    
Now what do you think of that?    
Purr, purr, purr, purr,    
Purr, purr, purr, purr.    
三个漂亮的新娘,    
与她们的夫君相伴    
住进了一间非常漂亮的公寓;    
公寓里既没有大老鼠,也没有小老鼠,    
也没吵吵闹闹!    
她们的生活看起来很幸福—    
她们从此被人**宠爱**而不再被人取笑—    
现在,你觉得怎么样呢?    
咕噜,咕噜,咕噜,咕噜,    
咕噜,咕噜,咕噜,咕噜.
