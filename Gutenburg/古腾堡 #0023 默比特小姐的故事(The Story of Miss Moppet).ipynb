![m10](http://www.gutenberg.org/files/14848/14848-h/images/m10.jpg)    

**T**his is a Pussy called Miss **Moppet**, she thinks she has heard a mouse!    
这是一只名叫**默比特**的猫猫小姐,她觉得自己听到了老鼠的声音!    

---    

![m13](http://www.gutenberg.org/files/14848/14848-h/images/m13.jpg)    

This is the Mouse peeping out behind the cupboard, and making fun of Miss Moppet. He is not afraid of a kitten.    
这是一只在柜后偷瞄的老鼠,他正在取笑默比特小姐. 他觉得自己一点都不怕猫猫.    

---    

![m14](http://www.gutenberg.org/files/14848/14848-h/images/m14.jpg)    

This is Miss Moppet jumping just too late; she misses the Mouse and hits her own head.    
默比特小姐扑得太晚了,她错过了老鼠并撞到了自己的脑袋.    

---    

![m17](http://www.gutenberg.org/files/14848/14848-h/images/m17.jpg)    

She thinks it is a very hard cupboard!    
她觉得这是一个非常硬的柜子!    

---    

![m18](http://www.gutenberg.org/files/14848/14848-h/images/m18.jpg)    

The Mouse **watches** Miss Moppet from the top of the cupboard.    
这只老鼠又在柜子的顶部**打量着**默比特小姐.    

---    

![m21](http://www.gutenberg.org/files/14848/14848-h/images/m21.jpg)    

Miss Moppet **ties up her head in a duster**, and sits before the fire.    
默比特小姐坐在火炉前,**用抹布把自己的脑袋绑了起来**.    

---    

![m23](http://www.gutenberg.org/files/14848/14848-h/images/m23.jpg)    

The Mouse thinks she is looking very ill. He comes sliding down the bell-pull.    
老鼠认为猫猫伤的很严重. 他顺着铃铛绳滑了下来.    

---    

![m24](http://www.gutenberg.org/files/14848/14848-h/images/m24.jpg)    

Miss Moppet looks worse and worse. The Mouse comes a little nearer.    
默比特小姐的情况看起来越来越糟. 然后,老鼠稍微靠近了一点.    

---    

![m26](http://www.gutenberg.org/files/14848/14848-h/images/m26.jpg)    

Miss Moppet holds her poor head in her paws, and looks at him through a hole in the duster. The Mouse comes very close.    
默比特小姐用爪爪抱住了自己可怜的小脑袋,然后透过抹布上的一个洞洞观察着老鼠. 现在老鼠已经靠的非常的近了.    

---    

![m29](http://www.gutenberg.org/files/14848/14848-h/images/m29.jpg)    

And then all of a sudden— Miss Moppet jumps upon the Mouse!    
然后,突然之间— 默比特小姐又扑向了老鼠(这一次,成功的捉住了他)!    

---    

![m30](http://www.gutenberg.org/files/14848/14848-h/images/m30.jpg)    

And because the Mouse has teased Miss Moppet— Miss Moppet thinks she will tease the Mouse; which is not at all nice of Miss Moppet.    
由于老鼠取笑了默比特小姐— 所以默比特小姐觉得自己也应该取笑他一番. 当然,默比特小姐这样的行为一点也不友好.    

---    

![m33](http://www.gutenberg.org/files/14848/14848-h/images/m33.jpg)    

She ties him up in the duster, and tosses it about like a ball.    
猫猫小姐把老鼠绑在了抹布中,像个球一样将他抛到了空中.    

---    

![m35](http://www.gutenberg.org/files/14848/14848-h/images/m35.jpg)    

But she forgot about that hole in the duster; and when she untied it— there was no Mouse!    
但是她忘记了抹布上的那个洞. 当她解开抹布时— 这里面竟然没了老鼠!    

---    

![m36](http://www.gutenberg.org/files/14848/14848-h/images/m36.jpg)    

He has **wriggled out** and run away; and he is dancing a **jig** on the top of the cupboard!    
因为老鼠早已**钻了出去**,并逃跑了. 你看,他现在正在柜子上嘚瑟的跳着**吉格舞**呢!
