turtle中常见的坐标有三种:  
1. 电脑屏幕中如何确定窗口的大小和位置.
    * `setup(width,height,startx,starty)`: 
用于确定绘图窗口的大小和位置. 窗口左上角,相对于屏幕左上角的进行定位.  

1. 如何确定海龟在窗口中位置.  
    * `home()`: 海龟在绘图窗口中的原点, 位于绘图窗口的中心.  
    * `goto(x,y)`: 将海龟移动到绘图窗口中的指定位置.  

1. 海龟相对于自身的移动方向和位置.  
    * `setheading(degree)`: 设置海龟的移动方向
    * `forward(dis)`: 设置海龟前行的移动步数

如果保证弧与初始方向"正向相切",只画角度为正的弧即可(不用考虑角度的正负号,只关心半径的正负号).  
* 半径为正,逆时针旋转;
* 半径为负,顺时针旋转;  


1. 确定其实方向线
1. 判断圆弧在方向线的左右  
半径为正,弧在原始方向的左侧;  
半径为负,弧在原始方向的右侧;  
2. 判断圆心相对于圆弧的位置
角度为正, 圆心在弧的左侧;  
角度为负,圆心在弧的右侧;  
3. 检查圆弧的顺逆
乘积为正,逆时针画弧; 乘积为负时顺时针画弧;  
4. 由现实的弧线逆向确定半径与角度的**正负**

## 正多边形

```python
def draw_polygon(num=3,radius=200,length=None):
    """按边长或圆内切绘制正多边形"""
    tt.screensize(500,500)
    tt.pensize(3)
    tt.pencolor('red')
    tt.penup()
    tt.goto(-100,-250)
    tt.pendown()    
    
    if length!= None:
        for _ in range(num):
            tt.forward(length)
            tt.left(360/num)
    else: 
        tt.circle(radius=radius,steps=num)
        tt.pencolor('black')
        tt.circle(radius=radius)
        
    tt.done()
```

```python
import turtle as t
def draw_pinwheel():
    tt.pensize(5)
    tt.speed(1)
    tt.color('black', 'blue')

    for i in range(4):
        tt.setheading(0+90*i)

        tt.begin_fill()
        tt.forward(150)
        tt.right(90)
        tt.circle(-150,45)
        tt.right(90)
        tt.forward(150)
        tt.end_fill()
        
    tt.done()
    
def pinwheel2():
    """风车"""
    tt.pensize(5)
    tt.speed(1)
    tt.color('green', 'blue')
    
    for _ in range(4):
        tt.forward(150)  
        tt.right(90)
        tt.circle(-150,45)
        tt.right(90)
        tt.forward(150)
        tt.left(135)

    tt.done()
```

## 绘制爱心

```python
def draw_heart(r=50):
    """假设半径为r,则上直角边长为r,下直角边长为3r(直径加上直角的一个边长)"""
    tt.pensize(5)
    tt.seth(0)
    tt.color('red')
    tt.begin_fill()
    tt.left(135)

    tt.forward(3*r)
    tt.right(180)#画笔掉头
    tt.circle(r,-180)
    tt.backward(r)
    tt.right(90)
    tt.forward(r)
    tt.circle(-r,180)
    tt.forward(3*r)
    tt.end_fill()
    tt.hideturtle()
    tt.done()
```

## 小蛇

```python
import turtle as tt
def draw_snake( rad=70,angle=80,length=2,neckrad=15):
    tt.reset()
    tt.setup(1200, 500, 0, 0)
    tt.pensize(30) # 画笔尺寸 
    tt.pencolor("green") 
    tt.seth(-40) # 前进的方向 
    
    for _ in range(length): 
        tt.circle(rad, angle) 
        tt.circle(-rad, angle) 
    tt.circle(rad, angle/2) 
    tt.forward(rad/2) # 直线前进 
    tt.circle(neckrad, 180) 
    tt.forward(rad/4) 
    tt.done()
```

## 简化亭桥

```python
from turtle import *
from math import *

G = (sqrt(5)-1)/2

def goto_point(x=0,y=0,show_line=True):
    """从当前点直线前进到另一点,笔的默认状态为落下"""
    if show_line==True: goto(x,y)
    else:
        penup()
        goto(x,y)
        pendown()

def locating_point(tag=0,size=10,color='red'):
    """定位点"""
    dot(size,color)
    print("locating point {}: {}".format(tag,position()))

def sind(degree):
    return sin(deg2rad(degree))

def cosd(degree):
    """当参数为角度时直接转化为弧度进行计算"""
    return cos(deg2rad(degree))

def tand(degree):
    return tan(deg2rad(degree))

def skip(fun):
    penup()
    fun
    pendown()
    
def forward_skip(dis):
    penup()
    forward(dis)
    pendown()
def home_skip():
    penup()
    home()
    pendown()
    
def coord(r=300):
    # 坐标系
    pencolor('red')
    
    tracer(False)
    forward(300)
    backward(300)
    left(90)
    forward(300)
    backward(300)
    tracer(True)
    right(90)
    pencolor('black')

def pillar(x,y,dx,dy):
    #第一根柱子
    goto_point(0,0,False)
    
    color('black','red')
    begin_fill()
    goto_point(x-dx,y,False)
    goto_point(x+dx,y)
    goto_point(x+dx,y+dy)
    goto_point(x-dx,y+dy)
    goto_point(x-dx,y)
    end_fill()
    
    goto_point(0,0,False)

    
def ddpy(R=100,times=2,alpha=180,theta=45,dx=2):
    """
    R:大孔半径
    times:大孔半径与小孔半径的比
    alpha: 大小桥孔的圆心角
    theta: 亭子的底角
    dx: 柱子的半径"""
    
    r = R/times # 小孔半径
    if theta == None: theta = rad2deg(atan(1-2*G**2)) #theta=13.283时,长纵比等于1
    
    
    ## d表示横,h表示纵
    d = 2*R*sind(alpha/2) # 大石的长
    h = d*G # 大石的高
    
    d2= (h-2*r*sind(alpha/2))/2 # 大小孔的间距
    d3= h # 柱间距等于大石的高 
    
    h2= d3*G # 柱子的高
    h3= d/2*tand(theta) #亭子的高      
                     
    x = d3/2                  # 右柱的下心x=柱间距的一半
    y = h                     # 右柱的下心y等于大石的高
    
    coord()
    color('black','gray')    
    begin_fill()
    
    ########## 桥 ##########
    # >1 
    goto_point(d/2,0,show_line=False)
    # >2
    forward(d2)
    # >3
    left(alpha/2)
    circle(-r,alpha)
    # >4
    seth(0)  
    forward(d2)
    # >5
    left(90)
    circle(h,90)
    # >6
    forward(d)
    # >7
    circle(h,90)
    # >8
    seth(0)
    forward(d2)
    # >9
    seth(alpha/2)
    circle(-r,alpha)
    # >10
    seth(0)
    forward(d2)
    # >1
    seth(alpha/2)
    circle(-R,alpha)
    end_fill()
    # >0
    home_skip()
    
    ########## 亭子 ##########
    

    ## 柱子左
    pillar(-x,y,dx,h2) 
    ## 柱子右
    pillar(x,y,dx,h2) 
    
    ## 亭顶三角形
    ## 假设右弧的右半径,与三角形的底边垂直,且相交于右下角点
    
    ## 右弧: 
    color('black','yellow')
    begin_fill()
    r3=h3/2/sind(theta)**2
    
    goto_point(0,h+h2,False) # 三角形的底边中心
    forward(d/2)             # 三角形的右下角点
    seth(180)
    circle(-r3,2*theta)
    
    ##左弧
    seth(0)
    goto_point(0,h+h2,False) # 三角形的底边中心
    backward(d/2)             # 三角形的左下角点
    seth(0)
    circle(r3,2*theta)      
    end_fill()
    
    ## 回到初始位置
    goto_point(0,0,0)
    seth(0)
    
    ## LOGO长纵比
    print("长纵比: {}".format((d/2+h)/(h+h2+h2)))   

def main():
    setup(1000,700,0,0)
    pensize(2)
    ddpy()
    mainloop()
    
if __name__ == "__main__":
    main()
```
