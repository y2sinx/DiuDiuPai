**The Adventures    
Of    
Samuel And Selina**    

By    
**Jean C. Archer**    

![emblem](http://www.gutenberg.org/files/24611/24611-h/images/emblem.jpg)    

*London, Grant Richards, 1902.*    

---    

In Spring,    
While softly **cooed**    
The Dove,    
春,鸽子在轻柔地**咕咕叫着**,    

Sam    
Told Selina of    
His Love.    
山姆,对瑟琳娜诉说了对她的爱意.    

![1](http://www.gutenberg.org/files/24611/24611-h/images/i001.jpg)    

---    

The Summer Moon smiled on them both,    
Selina **plighted** him her Troth.    
夏夜的月光中,瑟琳娜向山姆**允诺了**同他的婚约.    

![2](http://www.gutenberg.org/files/24611/24611-h/images/i002.jpg)    

---    

But Autumn brought a gayer **Swain**—    
Selina broke it off again.    
秋天,瑟琳娜结识了一位使她更快乐的**求爱者**, 瑟琳娜打破了与山姆的誓言.    

![3](http://www.gutenberg.org/files/24611/24611-h/images/i003.jpg)    

---    

'Tis Winter now— Selina's slack—    
She'd give her thumbs to have him back.    
冬,瑟琳娜踌躇着,她想同山姆重拾婚约.    

![4](http://www.gutenberg.org/files/24611/24611-h/images/i004.jpg)    

---    

Yet—    
When they met    
She tossed her head;    
然而,当他们相遇的时候,瑟琳娜却扬起了头(选择了对山姆视而不见).    

He    
Stared at her and    
Cut her dead!    
山姆虽盯了瑟琳娜一会儿,但亦选择了视而不见.    

![5](http://www.gutenberg.org/files/24611/24611-h/images/i005.jpg)    

---    

But Fate at last to them was kind:    
It sent a Roaring, Raging Wind!    
但是命运对他们非常的仁慈: 给他们送来了一阵呼啸的狂风!    

Which,    
Just as Sam was passing by,    
Blew off Selina's Hat!    
Oh! My!    
正当山姆与瑟琳娜擦肩的时候,天哪(好巧),大风竟然吹落了瑟琳娜的帽子!    

![6](http://www.gutenberg.org/files/24611/24611-h/images/i006.jpg)    

---    

Sam    
Caught it— by a daring jump.    
山姆英勇的接住了瑟琳娜的帽子.    

Selina's    
Heart    
went    
**Thump**! Thump!! Thump!!!    
瑟琳娜的心突然开始**砰砰砰**的直跳!!!    

"Oh, Sam!" she cried;    
Tears dimmed her sight—    
And after that it all came right.    
Thump! Thump!! Thump!!!    
"哦,山姆!" 她叫到; 泪水润湿了她的眼睛— 如此一来一切都变得明朗了起来. 他们的心开始一起砰砰砰的直跳.    

![7](http://www.gutenberg.org/files/24611/24611-h/images/i007.jpg)    

---    

They made it up— and very soon    
They started on their Honeymoon.    
他们结了婚— 并很快开始了他们的蜜月之旅.    

![8](http://www.gutenberg.org/files/24611/24611-h/images/i008.jpg)    

---    

Selina proved a model wife,    
Her Sam was all her joy in life;    
She fetched his shoes and **darned** his hose,    
And sympathized with all his **woes**.    
时间证明,瑟琳娜是一位好妻子,她将山姆作为生活中的全部快乐.    
她不仅会为他操持家务(递鞋子,**缝补裤袜**),还会为他一起分担**忧愁**.    

![9](http://www.gutenberg.org/files/24611/24611-h/images/i009.jpg)    

---    

And,    
As she let him have his say,    
He loved her more from day to day;    
And— on her birthday— for a **spree**,    
Took her to the **Menagerie**.    
由于瑟琳娜从不干涉山姆的作为,山姆对瑟琳娜的爱与日俱增.    
为了**庆祝**瑟琳娜的生日,山姆便带她去了**动物园**.    

![10](http://www.gutenberg.org/files/24611/24611-h/images/i010.jpg)    

---    

She **revelled** in the **Monkey Walk**,    
Where Apes, of **motley hue**,    
Each jumped— upon a yellow stick—    
All shining and brand new.    
瑟琳娜对 "**猴子走步**" **入了迷**,那些猴子们,**混杂着各种叫声**,从一根黄色的柱子上,跳到另一根黄色的柱子上,每次的跳跃都非常精彩.    

![11](http://www.gutenberg.org/files/24611/24611-h/images/i011.jpg)    

---    

And picture, children, how the Snarks    
Rejoiced her **frugal mind**;    
They ate the Buns, they ate the Bag,    
And even stale cheese **rind**.    
哈哈,孩子们,看下面这张图,蛇鲨把**勤俭持家的**瑟琳娜逗乐了.    
它们不仅吃了面包,还把包给吃了,甚至还吃了变质的奶酪**壳**.    

![12](http://www.gutenberg.org/files/24611/24611-h/images/i012.jpg)    

---    

The **Jub-jub birds** Selina fed,    
But they were rude and rough;    
They fought and **scratched**,    
Nor would they stop    
When they had had enough.    
在这张图中,瑟琳娜正在给 "**Jub鸟们**" 投食呢,但是这些鸟儿们都很粗鲁.    
它们互相打着架,**撕扯着对方**,直到它们全都吃饱了才消停了下来.    

![13](http://www.gutenberg.org/files/24611/24611-h/images/i013.jpg)    

---    

At last,    
When happy, hot and tired,    
They found no more to see,    
Sam took her to a shady spot    
And treated her to tea.    
他们虽然又累又热,但是很幸福,由于没有其他可以看得了,    
山姆带瑟琳娜来到了一个凉快的地方,请她喝茶.    

Selina's hat and dress he praised,    
She clapped his **feeblest puns**;    
It was a perfect carnival    
Of **sentiment and Buns**!    
山姆称赞瑟琳娜的帽子与衣服,而瑟琳娜则称赞山姆那**拙劣的俏皮话**.    
这真是一个关于"**面包与情感**"的绝美嘉年华.    

![14](http://www.gutenberg.org/files/24611/24611-h/images/i014.jpg)    

---    

Much time, alas! they cannot spare,    
Since holidays are few;    
Soon, hand in hand, they start afresh    
To seek adventures new.    
哎,他们不能浪费更多的时间了,因为假期余额所剩不多了.    
不久,他们就手牵着手重新出发了,去寻找新的冒险.    

And all about along the walks    
Stern "Cautions" they **espy**;    
"You need not fear," said Samuel,    
"While I, my love, am nigh."    
沿着人行道到处走的时候,他们**发现了**一处钉着 "请小心" 的警告牌.    
"我的爱人,请勿怕," 塞缪尔说, "我时刻陪在你的身边."    

![15](http://www.gutenberg.org/files/24611/24611-h/images/i015.jpg)    

---    

Alas! how brief are mortal joys;    
There comes an awful **burbling** noise!    
哎,快乐时光停留地是多么短暂啊!    
突然传过来了一阵**糟糕的**声音.    

![16](http://www.gutenberg.org/files/24611/24611-h/images/i016.jpg)    

---    

As, terror-struck, he turns to fly,    
Too late he hears her anguished cry,    
山姆受到了惊吓,四下逃奔,并没有听到瑟琳娜痛苦的呼喊,    

"O Samuel!    
O Samuel!!    
"哦,山姆! 哦,山姆!!    

Beware!    
The awful    
Camuel"!!!    
小心那个糟糕的骆驼!!"    

![17](http://www.gutenberg.org/files/24611/24611-h/images/i017.jpg)    

---    

The Camel rushed!    
The Camel flew!    
Till all its spots were **streaks** of blue;    
To Samuel it seemed to be    
Itself a whole    
Menagerie!    
骆驼急速的飞了过来,它身上的斑点此时也变成了**一条条**蓝色的线.    
而此时的山姆,好像他本身也成了动物园的一项展览.    

![18](http://www.gutenberg.org/files/24611/24611-h/images/i018.jpg)    

---    

The Camel chased him round and round;    
He sank— exhausted— on the ground;    
The Camel never noticed that,    
But pranced along—    
with Sammy's hat.    
骆驼追着山姆跑了一圈又一圈,他最终精疲力尽,一屁股坐在了地面上.    
骆驼并未注意到山姆停下来,而是蒙着山姆的帽子,仍在不停的奔腾.    

![19](http://www.gutenberg.org/files/24611/24611-h/images/i019.jpg)    

---    

And— when it found its victim gone,    
Imagine how the brute went on;    
It **bucked** and **reared**    
and kicked and **shied**,    
Till, finally,    
It Bust!    
and died.    
当骆驼发现山姆跑了,可以想象到它是多么的生气.    
他**弓着背跳跃着**,还**用后腿直立着**,然后不停地踢蹬并且**倒退**,直到最终猛的一下,爆裂而亡了.    

![20](http://www.gutenberg.org/files/24611/24611-h/images/i020.jpg)    

---    

When Sammy heard the loud report    
And saw the pieces fly,    
He felt that **sure as eggs was eggs**,    
He, too, must surely die.    
当山姆听到爆裂的声音并看到四散的碎块后,他**确信自己也一定会**死去了.    

But brave Selina, though her tears    
Fell all the while like rain,    
Washed off the dirt and set him up    
Upon his feet again.    
但是勇敢的瑟琳娜(虽然泪如雨下)给他清理了身上的泥土,并把他扶了起来.    

![21](http://www.gutenberg.org/files/24611/24611-h/images/i021.jpg)    

---    

She found the remnants of his hat,    
And led him to the gate;    
But there the Camel's owner stood    
As large and grim as fate.    
瑟琳娜找到了山姆破损的帽子,并领着他来到了动物园门口.    
而骆驼的主人正在这里站着,看起来十分的无情.    

Before they left, that greedy man    
Took all the cash they had,    
And turned their pockets inside-out    
(Which made Selina mad).    
在他们离开之前,那个贪婪的人拿走了他们所有的钱.    
这使得瑟琳娜感到非常地生气.    

![22](http://www.gutenberg.org/files/24611/24611-h/images/i022.jpg)    

---    

How different their coming home    
From their gay start at morn;    
They creep along— **a sorry sight**—    
**Bedraggled** and **forlorn**.    
与他们早上离家时的快乐心情相比,他们回家时的心情是多么的糟糕啊!    
他们沿着**一副悲惨的光景**"爬行",**全身脏兮兮的**,好像那些**被遗弃的**人.    

He knows he showed a want of pluck,    
Whatever she may say;    
She feels that it was all her fault    
For having a birthday.    
山姆知道自己缺乏勇气,无论瑟琳娜怎么数落他都不为过.    
但是,瑟琳娜却认为这全是她自己的过错,因为这全都是为了庆祝她的生日.    

![23](http://www.gutenberg.org/files/24611/24611-h/images/i023.jpg)    

---    

But— once at home— the **ruddy blaze**    
Each **drooping spirit cheers**;    
Sam sets Selina by the fire    
And wipes away her tears.    
但是— 一旦回到家— 他们立马就从**垂头丧气**变得**容光焕发了**.    
山姆陪着瑟琳娜坐在火堆旁,为她擦去了眼角的眼泪.    

He draws her closer to his side;    
He **tootles** on a comb,    
And sings her, as her sobs **subside**,    
A verse of "Home, Sweet Home."    
山姆把瑟琳娜拥入怀中,开始**安慰**着她.
当瑟琳娜从啜泣中平静下来,对她轻轻唱着,"家,甜蜜的家."   

![24](http://www.gutenberg.org/files/24611/24611-h/images/i024.jpg)
