__The    
Illustrated    
Alphabet Of Birds__    

![Small Bird](http://www.gutenberg.org/files/11073/11073-h/images/titlebird.jpg)    

_Boston    
Wm. Crosby & H.P. Nichols    
1851_    

---    

__A__ is an Auk,    
Of the **Arctic** sea,    
He lives on the ice,    
Where the winds blow free.    
A 是一只生活在**北冰洋**的海雀,    
那里的风整日肆无忌惮的吹着.    

![auk](http://www.gutenberg.org/files/11073/11073-h/images/auk.jpg)    

---    

B is a Blue Bird.    
In early spring,    
How sweet his songs    
Through the forest ring.    
B 是一只蓝鸟.    
若在早春的时候,能透过森林听到他的歌声,该是多么如此的美妙.    

![bluebird](http://www.gutenberg.org/files/11073/11073-h/images/bluebird.jpg)    

---    

C is a **Condor**,    
On the Andes' height,    
He **plumes** his wings    
For a lofty flight.    
C 是一只可以飞越安第斯山脉的**神鹰**,    
他**扑棱着**翅膀,只为寻找一个更高的地方.    

![**condor**](http://www.gutenberg.org/files/11073/11073-h/images/condor.jpg)    

---    

D is a Duck.    
Of the canvas back sort,    
To shoot at a flock    
Is considered fine sport.    
D 是一种背如帆布的野鸭.    
在英国人眼里,猎杀野鸭是一项非常好的运动.    

![duck](http://www.gutenberg.org/files/11073/11073-h/images/duck.jpg)    

---    

E's a **Bald Eagle**,    
So bold and so free;    
On the flag of our country    
He spans land and sea.    
E 是一只自由且勇敢的**白头鹫**;    
他们飞翔在英国的陆地和海洋之上.    

![eagle](http://www.gutenberg.org/files/11073/11073-h/images/eagle.jpg)    

---    

F is a Fish Hawk,    
Who lives on the shore:    
He **catches** his prey    
Mid the ocean's roar.    
F 是一只生活在海岸边的鱼鹰,    
他在咆哮的海洋之上**捕捉**他的猎物.    

![fishhawk](http://www.gutenberg.org/files/11073/11073-h/images/fishhawk.jpg)    

---    

G is a Goose;    
His feathers we take    
And put them in **sacking**    
Our beds to make.    
G 是一只鹅;    
他们的绒毛经常被人们装进**布袋**做成床上用品.    

![goose](http://www.gutenberg.org/files/11073/11073-h/images/goose.jpg)    

---    

H is a Humming Bird,    
Sporting mid flowers    
And brightly enjoying    
The sunny hours.    
H 是一只奔忙在花丛之中的蜂鸟,    
他们非常喜欢阳光明媚的时光.    

![**hummingbird**](http://www.gutenberg.org/files/11073/11073-h/images/hummingbird.jpg)    

---    

I is an **Ibis**,    
Who wanders in bogs,    
And lives upon lizards,    
And fishes and frogs.    
I 是一只在沼泽中漫步的**朱鹭**,    
他们以蜥蜴,鱼和青蛙为食.    

![**ibis**](http://www.gutenberg.org/files/11073/11073-h/images/ibis.jpg)    

---    

J is a **Jay**,    
With his blue and white coat,    
With a **crest** on his head,    
And a ring round his throat.    
J 是一只**松鸡**,他的羽毛蓝白相间,    
他的头顶戴着**鸟冠**,他的喉咙附近有个小圈.    

![**jay**](http://www.gutenberg.org/files/11073/11073-h/images/jay.jpg)    

---    

K is a King Bird,    
**Pugnacious** and bold:    
A hero in fight,    
And a terrible scold.    
K 是一只国王鸟,**脾气暴躁**而大胆,    
他是战斗中的英雄,也是喜欢招惹是非的坏怂(huài sǒng).    

![**kingbird**](http://www.gutenberg.org/files/11073/11073-h/images/kingbird.jpg)    

---    

L is a Lark,    
A sociable bird;    
His song in the meadow    
Is frequently heard.    
L 是一只善于交际的云雀,    
在草地之上,可以经常听到他们的歌声.    

![lark](http://www.gutenberg.org/files/11073/11073-h/images/lark.jpg)    

---    

M is a Magpie,    
He lives at the west,    
Steals and scolds and eats **carrion**;    
He's none of the best.    
M 是一只生活在西部的喜鹊,    
这种鸟不是很好,他喜欢偷东西,招惹是非,以及吃**腐肉**.    

![magpie](http://www.gutenberg.org/files/11073/11073-h/images/magpie.jpg)    

---    

N is a Night **Heron**,    
Of fishes quite fond;    
He looks for them now,    
As he stands by the pond.    
N 是一只喜欢鱼的**夜鹭**,    
他现在正站在池塘边找鱼吃呢.    

![nightheron](http://www.gutenberg.org/files/11073/11073-h/images/nightheron.jpg)    

---    

O is an Owl,    
Who hides through the day;    
And comes out at night,    
To seek for his prey.    
O 是一只猫头鹰,他会在白天的时候躲起来;    
而只有在晚上的时候出来寻找猎物.    

![owl](http://www.gutenberg.org/files/11073/11073-h/images/owl.jpg)    

---    

P is a Pigeon    
So rapid in flight,    
That before you can shoot him    
He's gone out of sight.    
P 是一只鸽子,可以飞的很快,    
当你准备开枪打他的时候,他早已飞的不见踪影.    

![pigeon](http://www.gutenberg.org/files/11073/11073-h/images/pigeon.jpg)    

---    

Q is a Quail,    
Who hides in a tree,    
And whistles "**Bob-White**"    
With lively glee.    
Q 是一只躲在树上的鹌鹑(ān chún),    
欢快的叫着"**鲍勃⋅怀特**",就像在吹口哨.    

![quail](http://www.gutenberg.org/files/11073/11073-h/images/quail.jpg)    

---    

R is the Robin,    
So kind and so good,    
Who covered with leaves,    
The poor babes in the wood.    
R 是一只知更鸟,他非常的友好,    
经常藏在树叶之中,是树林中可怜的小东西.    

![robin](http://www.gutenberg.org/files/11073/11073-h/images/robin.jpg)    

---    

S is the Swallow,    
She darts through the air    
To catch little insects,    
Her favorite fare.    
S 是一只燕子,喜欢在空中猛冲,    
这是为了捕捉小昆虫,她们最喜欢的食物.    

![swallow](http://www.gutenberg.org/files/11073/11073-h/images/swallow.jpg)    

---    

T is a Turkey,    
A fine dashing beau,    
By his fuming and **strutting**,    
His pride you may know.    
T 是火鸡,一种非常华丽的鸟类,    
从他们的怒气冲冲和**昂首阔步**,可以看出他们很骄傲.    

![turkey](http://www.gutenberg.org/files/11073/11073-h/images/turkey.jpg)    

---    

U's the Upupa,    
Or **Hoopoe**. His crest    
He can raise or lower,    
As suits him best.    
U 是一只**戴胜鸟**,    
他的鸟冠可以随意的升起或落下,直到最适合他们.    

![upupa](http://www.gutenberg.org/files/11073/11073-h/images/upupa.jpg)    

---    

V is a **Vulture**,    
Who feeds on the dead,    
When the dark battle-field,    
With corpses is spread.    
V 是一只**秃鹫**,    
他在尸横遍野的战场之上以死者为食,    

![vulture](http://www.gutenberg.org/files/11073/11073-h/images/vulture.jpg)    

---    

W's a Woodpecker,    
Who with his long bill,    
Bores holes in a tree    
And of worms eats his fill.    
W 是一只啄木鸟,他的鸟喙(huì)很长,    
他喜欢在树上钻孔,然后吃掉里面的蠕虫.    

![woodpecker](http://www.gutenberg.org/files/11073/11073-h/images/woodpecker.jpg)    

---    

X is Xanthoronus,    
Or Baltimore Bird.    
Oft in our orchards,    
His music is heard.    
X 是巴尔的摩黄鹂,    
在我们的果园里可以经常听到他们的歌声.    

![xanthoronus](http://www.gutenberg.org/files/11073/11073-h/images/xanthoronus.jpg)    

---    

Y is a Yellow Bird,    
With feathers so bright,    
Who sings all the day,    
And sleeps all the night.    
Y 是一只金莺,他的羽毛非常的靓丽,    
他们在白天会唱一天的歌,在晚上会睡一晚的觉.    

![yellowbird](http://www.gutenberg.org/files/11073/11073-h/images/yellowbird.jpg)    

---    

Z stands for none    
of the feathered race.    
It must serve as a **roost**,    
Or lose the last place.    
Z 不是指鸟而是指他们的**巢穴**,    
否则他们将失去最后的栖身之所.    

![zroost](http://www.gutenberg.org/files/11073/11073-h/images/zroost.jpg)
