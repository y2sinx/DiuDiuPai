# 辽阔的天空    

南希·楚宁 | 奥尔加·鲍默特    

[sm]()    

When Katharine Lee Bates(凯瑟琳·李·贝茨) was very young, the Civil War raged(内战肆虐). Some of her earliest memories were of men trudging home(拖着沉重的步伐回家) in tattered blue(穿着破破烂烂的蓝色衣服). Then when Abraham Lincoln(亚伯拉罕·林肯) was shot, Katharine's mother wept(哭了). A hush suffocated the streets of her village(她的小村庄安静的令人窒息). The country's heart was ripped in two(被撕成了两块).    

[4-5]()    

Katharine sat for hours, gazing at the sea. But all that the waves stirred up(波浪所激起的) was empty shells(空空的贝壳), which she scooped in her apron(用围裙抱起). What could a girl growing up in the small village of Falmouth, Massachusetts, do(一个在小村庄长大的女孩子能做什么)?    

No one expected much of girls in 1865(没人会期盼一个女孩子做什么大事). The boys she knew grew up to be fishermen or studied to become doctors or lawyers or businessmen. Girls learned to mend and cook(缝缝补补和烧菜做饭).    

Katharine didn't think that was fair.    

At nine, she wrote in her diary: "Sewing(缝缝补补) is always expected of girls. Why not of boys?"    

[6-7]()    

Katharine was lucky. Her widowed(守寡的) mother took in sewing and washing, and grew and sold vegetables to help Katharine go to school.    

Some people said girls couldn't learn as well as boys(无法像男孩一样学习). Others said she should be thinking about marriage, not books.    

[8]()    

Katharine didn't agree. She marveled at(对...惊奇) the power of words to make people think and feel differently. When she heard of a new college that opened just for women, she studied hard to earn a place there so she could keep writing.    

[9]()    

At Wellesley(韦尔斯利), everyone ate the same food and lived in identical rooms(相同的房间). Katharine and her classmates were taught they could do anything that men could do.    

But outside her school, the rich didn't mix with the poor, who worked long hours for little pay.    

Women couldn't become doctors or lawyers or own businesses. They couldn't even vote(投票).    

Katharine sat for hours, gazing at Lake Waban(瓦班湖) on her college campus. What could a young woman do?    

[10-11]()    

She helped establish(帮助建立了) a settlement house(安置房) for immigrants(移民) and college women in Boston. She spoke up for women's suffrage(妇女选举权) and world peace.    

She wrote a novel, Rose and Thorn(玫瑰与荆棘), about people who labored in sweatshops(血汗工厂) and struggled to care for their sick children. Her book inspired talk about how to help the poor. Katharine's words were making a difference(产生了重大影响), and she had so much more to say.    

Katharine became a professor. And she kept writing. Poems(诗歌). Books(书籍). Essays(散文).    

[12-13]()    

In 1893, Katharine took a job teaching a summer class in Colorado. She traveled by train from Boston, excited to see the country on the way.    

In New York, she marveled at the cascading waters of Niagara Falls(她惊叹于尼亚加拉大瀑布的壮美). In Chicago, she admired the alabaster city(雪花石膏城) of the future at the World's Fair(世界博览会). So much beauty. So much hope.    

[14-15]()    

And yet, from her work as a reformer, she knew people struggled across the country. Some groaned in dark mines(在黑暗的矿井之中痛苦的叹息); some picked crops in blistering sun(在炎炎的烈日之下采摘玉米); others choked for air in smoke-stained factories(在烟雾弥漫的工厂里忍受令人窒息的空气).    

As an economic depression(经济萧条) spread dark wings(伸展黑暗的羽翼) across the country, people filled the streets, feeling scared and angry and alone.    

[16-17]()    

On July 4, 1893, the country's birthday(国庆日), Katharine sat on the train, gazing through the window. She saw the amber stalks of wheat swaying in the Kansas fields(琥珀色的麦梗在堪萨斯的田野上摇曳). The way they moved flooded her mind(麦浪涌上了她的心头) with memories of the ocean(连同关于海洋的记忆).    

Suddenly she wondered: Could the people who sail powerful waves of water(航行水浪的人) and the people who tend mighty waves of grain(照料谷浪的人) be more alike than we realize(比我们所意识到的还要相像)?    

[18-19]()    

Katharine's mind might have been dancing with that idea when she arrived in Colorado and visited Pikes Peak(派克峰). She rode a horse-drawn wagon(马车) up the mountain to 14,000 feet above sea level. The horse snorted(气汹汹的喷着鼻子) and stopped.    

Katharine wanted to keep going. A mule(骡子) took the wagon up, up, up the final 110 feet of crumbled pink granite road(粉色的花岗岩碎石路) to the summit.    

[20-21]()    

"O beautiful!" As Katharine gazed from that height, she saw America as one unbroken country(完整的国家), with no walls between people.    

It had been a long journey up the mountain. But Katharine was uplifted by(被升了起来) the wonder of America. "Most glorious scenery I ever beheld" she wrote in her notebook.    

The first lines of a poem rushed to her head. She wrote and rewrote in her hotel room, trying to get each word right.    

[22-23]()    

Two years later, on July 4, 1895, a national magazine published Katharine's poem. People turned to one another, proudly sharing the words that described America united as a family. The poem was so popular it was published several more times, and each time Katharine revised it.    

By the fourth published version, she was satisfied with the poem that had come to be called "America the Beautiful(美丽的美国)."    

In 1910, people started singing the poem to a melody(旋律) by composer(作曲家) Samuel A. Ward(塞缪尔·沃德). Now Katharine's words soared even higher(飞得更高了,在人群中散播的更远了).    

[24-25]()    

"America the Beautiful" was the most famous poem Katharine ever wrote, but she never accepted money for it. It was her gift to America. Americans, rich and poor, men and women, from sea to shining sea, flooded her with thank-you letters.    

She wrote back to every one, handwriting copies of what she jokingly(打趣地) called her "A the B" poem. She served coffee and cake to visitors who came to the home she shared with Katharine Coman, another professor at Wellesley. And she continued to speak up for the poor, for the rights of women, and for peace in the country and across the world.    

[26-27]()    

Nine years after "America the Beautiful" became popular as a song, Congress passed the Nineteenth Amendment(国会通过了第十九修正案), which gave women the right to vote. In 1920 Katharine strode to the ballot box(大步走向投票箱), filled with joy.    

Katharine never did like needles and thread(针线活). But she had crafted a poem(创作了一首诗) that sewed the dreams of a diverse nation together(把一个多元化国家的"统一的"梦想缝合了起来).    

[28-29]()    

America still had a long way to go toward mending the differences(弥合分歧) that pushed people apart. But on the day Katharine cast her ballot(投票) alongside and equal to men, America was one stitch closer(更进一"针") to becoming the country she knew it could be— beautiful.
