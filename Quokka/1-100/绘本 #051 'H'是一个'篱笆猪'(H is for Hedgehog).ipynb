### 'H'是一个'篱笆猪'    

利维·高斯林    

![sm.png](https://ae04.alicdn.com/kf/U0969f433e55949009f5a444eadac2ae0B.jpg)    

---

There are many different types(类型) of hedgehogs(刺猬).    

African pygmy hedgehog(非洲小刺猬), European hedgehog(欧洲刺猬), Long-eared hedgehog(长耳刺猬), Brandt's hedgehog(勃兰特刺猬), Indian long-eared hedgehog(印度长耳刺猬), Amur hedgehog(远东刺猬/东北刺猬).    

![02-03](https://s.pc.qq.com/tousu/img/20210304/8676741_1614846266.jpg)    

---

Some hedgehogs live in hedgerows(灌木篱笆墙) in the countryside(乡下).    

![04](https://s.pc.qq.com/tousu/img/20210304/1868133_1614846259.jpg)    

---

Others live in the hot desert(沙漠).    

![05](https://s.pc.qq.com/tousu/img/20210304/9280503_1614846264.jpg)    

---

Hedgehogs eat all sorts of(各种各样的) creepy crawlies(恐怖小爬虫).    

Caterpillar(毛毛虫), beetle(甲壳虫), worm(蠕虫), slug(鼻涕虫), earwig(蠼螋(qú sōu)).    

![06-07](https://s.pc.qq.com/tousu/img/20210304/9046696_1614846266.jpg)    

---

Hedgehogs spines(体刺) are hard and spiky(硬而尖的).    

![08](https://s.pc.qq.com/tousu/img/20210304/7963569_1614846267.jpg)    

---

A curled-up(卷起来的) hedgehog looks like a prickly ball(多刺的球).    

![09](https://s.pc.qq.com/tousu/img/20210304/4373214_1614846264.jpg)    

---

Baby hedgehogs are called hoglets(小刺猬).    

![10-11](https://s.pc.qq.com/tousu/img/20210304/2291761_1614846268.jpg)    

---

Hedgehogs hibernate(冬眠), which means(意味着) they sleep through winter(整个冬天都在睡觉).    

![12](https://s.pc.qq.com/tousu/img/20210304/9282127_1614846262.jpg)    

---

Hedgehogs have little legs, but they can run fast(腿小跑的快).    

![13](https://s.pc.qq.com/tousu/img/20210304/4914849_1614846256.jpg)    

---

H is for... hide and seek(捉迷藏)! Can you spot(找到) the hedgehogs?
