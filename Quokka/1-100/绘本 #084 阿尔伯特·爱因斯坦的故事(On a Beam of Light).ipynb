### 阿尔伯特·爱因斯坦的故事    

珍妮佛·伯恩 | 弗拉基米尔·拉顿斯基    

![035](https://s.pc.qq.com/tousu/img/20210427/3329873_1619536435.jpg)    

Over 100 years ago, as the stars swirled in the sky(当星星在天上打着旋), as the Earth circled the sun(地球在围着太阳转), as the March winds blew through a little town by a river(而三月的风拂过一个河边小镇的时候), a baby boy was born. His parents named him Albert(阿尔伯特).    

![002](https://s.pc.qq.com/tousu/img/20210427/3392674_1619536356.jpg)    

Albert turned one year old(一岁了). And didn't say a word(一句话都不会说).    

![004](https://s.pc.qq.com/tousu/img/20210427/6164280_1619536356.jpg)    

Albert turned two. And didn't say a word.    

![005](https://s.pc.qq.com/tousu/img/20210427/1407203_1619536356.jpg)    

Albert turned three. And hardly said a word at all.    

![006](https://s.pc.qq.com/tousu/img/20210427/8899085_1619536357.jpg)    

He just looked around with his big, curious eyes(好奇的大眼睛). Looked and wondered. Looked and wondered.    

![007](https://s.pc.qq.com/tousu/img/20210427/2369376_1619536356.jpg)    

His parents worried. Little Albert was so different; was there something wrong? But he was their baby, so they loved him... no matter what.    

![008](https://s.pc.qq.com/tousu/img/20210427/5487137_1619536356.jpg)    

One day, when Albert was sick in bed, his father brought him a compass(指南针)— a small round case with a magnetic needle inside(一个内含磁性指针的小盒子).    

![009](https://s.pc.qq.com/tousu/img/20210427/5748080_1619536356.jpg)    

No matter which way Albert turned the compass, the needle always pointed north, as if held by an invisible hand(无形的手). Albert was so amazed his body trembled(如此激动,以至于他的身体都颤抖了).    

Suddenly he knew there were mysteries(神秘而难以理解的事物) in the world— hidden and silent(隐秘而又沉静), unknown and unseen(不可知,不可见).    

He wanted, more than anything, to understand those mysteries.    

![010](https://s.pc.qq.com/tousu/img/20210427/6788217_1619536362.jpg)    

Albert started asking questions. Questions at home. Questions at school. But Albert didn't want to be like the other students.    

So many questions that some of his teachers told him he was a disruption to his class. They said he would never amount to anything(他会一事无成) unless he learned to behave like all the other students(除非他像其他学生那样端正自己的行为).    

![011](https://s.pc.qq.com/tousu/img/20210427/6735762_1619536360.jpg)    

But Albert didn't want to be like the other students. He wanted to discover the hidden mysteries in the world.    

![012](https://s.pc.qq.com/tousu/img/20210427/9523720_1619536360.jpg)    

One day, as Albert was zipping through the countryside on his bicycle(骑着自行车"窜过"(飞速穿过)村子的时候), he looked up at the beams of sunlight(太阳光线) speeding from the sun(飞速远离太阳) to the Earth. 

He wondered, what would it be like to ride one of those beams?    

And in his mind, right then and there(就在此时此地), Albert was no longer on his bicycle, no longer on the country road...  he was racing through space on a beam of light(骑着光飞越太空). 

It was the biggest, most exciting thought Albert had ever had. And it filled his mind with questions.    

![013](https://s.pc.qq.com/tousu/img/20210427/1281144_1619536359.jpg)    

Albert began to read and study. He read about light and sound(光和声). About heat and magnetism(热和磁).    

And about gravity(重力), the invisible force(一种看不见的力) that pulls us down toward our planet(将我们拉向我们的星球), and keeps the moon from floating away into outer space(阻止月亮跑到外太空).    

![014](https://s.pc.qq.com/tousu/img/20210427/8236278_1619536359.jpg)    

And he read about numbers. Albert loved numbers. They were like a secret language for figuring things out.    

But all that reading still didn't answer all of Albert's questions. So he kept on reading. Wondering. And learning.    

![015](https://s.pc.qq.com/tousu/img/20210427/5567074_1619536360.jpg)    

When Albert graduated from college, he wanted to teach the subjects(科目) he loved— all the things he had read about all those years. But Albert couldn't find a job as a teacher. So he got another job.    

![016](https://s.pc.qq.com/tousu/img/20210427/7071590_1619536374.jpg)    

A simple, quiet job in a government office(政府部门). An office where he worked with other people's ideas and inventions. He did his work very well and very quickly... so quickly that he had lots of extra time(额外时间) to think and wonder.    

![017](https://s.pc.qq.com/tousu/img/20210427/1041310_1619536363.jpg)    

Albert watched a lump of sugar(一块糖) dissolve and disappear into(溶解,然后消失进) his hot tea. How could this happen?    

He watched the smoke from his pipe(烟斗) swirl and disappear into the air. How could one thing disappear into another?    

![018](https://s.pc.qq.com/tousu/img/20210427/6963411_1619536363.jpg)    

Then he began to figure it out. He thought about the idea that everything is made out of(由...组成) teeny, tiny, moving bits of stuff— far too tiny to see— little bits called "atoms(原子)."    

![019](https://s.pc.qq.com/tousu/img/20210427/2619793_1619536352.jpg)    

Some people didn't believe that atoms(原子) existed, but Albert's figuring helped prove that everything in the world is made of atoms...    

![020](https://s.pc.qq.com/tousu/img/20210427/4627583_1619536433.jpg)    

Even sugar and tea, even smoke and air. Even Albert and you.    

![021](https://s.pc.qq.com/tousu/img/20210427/8774499_1619536444.jpg)    

Then Albert thought about motion(运动). He realized that everything is always moving. Moving through space, moving through time.    

Even sound asleep(即使是在酣睡中) we're moving(我们也在运动), as our planet circles the sun, and our lives travel into the future.    

Albert saw time and space as no one ever had before(阿尔伯特对时间和空间的认识超越了之前所有的人).    

![022](https://s.pc.qq.com/tousu/img/20210427/7601811_1619536436.jpg)    

Albert wrote down his new ideas, put them into envelopes(信封), and sent them to science magazines(科学杂志). The magazines printed everything Albert sent.    

He hoped that scientists and professors would be interested. And they were. Very interested indeed! They asked Albert to come work with them and teach with them.    

For the first time in his life, people started to say, "Albert is a genius(天才)!" Now Albert could spend all his days doing what he loved— imagining, wondering, figuring and thinking.    

![023](https://s.pc.qq.com/tousu/img/20210427/8656550_1619536441.jpg)    

Albert thought about very, very big things. Like the size and shape of the entire universe.    

![024](https://s.pc.qq.com/tousu/img/20210427/3201837_1619536439.jpg)    

He thought about very, very small things. Like what goes on inside the atoms that everything is made of.    

He thought about mysterious forces(力), like magnetism and gravity. He discovered whole new ways to understand how all these things work.    

![025](https://s.pc.qq.com/tousu/img/20210427/3206196_1619536433.jpg)    

Everywhere Albert went he would think and think. One of Albert's favorite thinking places was his little sailboat(帆船). He loved to let his mind wander as the wind blew him across the water.    

![026](https://s.pc.qq.com/tousu/img/20210427/1059340_1619536436.jpg)    

Sometimes, when Albert was having a tough time with a tricky problem(棘手的问题), he would put it aside(把它放在一边) and play his violin.    

Music made Albert happy. He said it helped him think better.    

![027](https://s.pc.qq.com/tousu/img/20210427/6827016_1619536435.jpg)    

Albert even chose his clothes for thinking. His favorites were his comfy, old saggy-baggy sweaters and pants(宽松下垂的毛衣和裤子). And shoes without socks.    

He said now that he was grown up, no one could tell him to put on his socks.    

![028](https://s.pc.qq.com/tousu/img/20210427/6627808_1619536440.jpg)    

In the town where he lived, he became known for wandering around(四处闲逛,徘徊), deep in thought(沉思). Sometimes eating an ice-cream cone(冰激凌锥,甜筒).    

Always recognizable with his long, wild hair(他乱蓬蓬的长头发很显眼(具有可辨识性)), which by then had become quite white.    

![029](https://s.pc.qq.com/tousu/img/20210427/8184511_1619536437.jpg)    

Everywhere Albert went he tried to figure out the secrets of the universe. And he never forgot about the beam of light that he rode so long ago in his imagination.    

Albert figured out that no person, no thing, could ever zoom through(快速通过) space as fast as a beam of light(一束光).    

![030](https://s.pc.qq.com/tousu/img/20210427/1378535_1619536440.jpg)    

He figured that if he could travel near the speed of light(光速), crazy things would happen! Only minutes would pass for Albert, while years and years went by for the rest of us!    

![032](https://s.pc.qq.com/tousu/img/20210427/3079892_1619536436.jpg)    

This idea was so amazing, people didn't believe it at first, but scientists today have proven that it's true.    

![033](https://s.pc.qq.com/tousu/img/20210427/6172957_1619536433.jpg)    

Albert thought and figured until the very last minute of the very last day of his life.    

He asked questions never asked before. Found answers never found before. And dreamed up ideas never dreamt before.    

![037](https://s.pc.qq.com/tousu/img/20210427/6328169_1619536438.jpg)    

Albert's ideas helped build spaceships(宇宙飞船) and satellites(人造卫星) that travel to the moon and beyond. His thinking helped us understand our universe as no one ever had before.    

But still, Albert left us many big questions. Questions that scientists are working on today. Questions that someday you may answer... by wondering, thinking, and imagining.
