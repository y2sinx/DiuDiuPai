### 好饿好饿的毛毛虫    

一本被各大"推广或销售平台"大肆宣传的绘本, 主要讲述了一只饥饿的毛毛虫吃了好多好多的东西, 然后变成了一只会飞的大蛾子(hú dié)的故事.    

现在让我来给你好好讲讲这个 "开头一个蛋(egg), 结局全靠吃" 的故事.    

在某个星期天(Sunday)的早上, 一颗吸收了日月精华的卵(egg), 蹦出来了(pop)一个又小又饿的毛毛虫(caterpillar). 因为他太饿了, 所以他见啥吃啥...    

星期一(Monday)他吃了一个苹果, 星期二(Tuesday)他吃了两个梨子, 星期三(Wednesday)他吃了三个李子(plums), 星期四(Thursday)他吃了四个草莓, 星期五(Friday)他吃了五个橙子(oranges), 然后星期六(Saturday)他吃的就比较多了...    

![29.png](https://ae02.alicdn.com/kf/H4425ae24b830448d87e8dcc25bc74fdc3.jpg)    

On Saturday he ate through one piece of chocolate cake(一块巧克力蛋糕), one ice-cream cone(一个冰激凌筒), one pickle(一根腌黄瓜), one slice of Swiss cheese(一片瑞士奶酪), one slice of salami(一片"萨拉米"风干肉肠), one lollipop(一根棒棒糖), one piece of cherry pie(一块樱桃派), one sausage(一根香肠), one cupcake(一个纸杯蛋糕), and one slice of watermelon(一片西瓜).    

That night he had a stomachache(那天晚上他吃的肚子都疼了)!    

现在又到了星期天啦, 他臃肿的(因为他现在又大又胖)趴在一片树叶子上, 然后开始了这颗残酷的星球上最令人匪夷所思的"异形进化"—— 他变成了一颗椭圆形的蛋蛋茧(cocoon).    

在接下来至少两周的时间里, 他在茧中闭关修炼, 先是将自己"溶解", 然后再照着蝴蝶姐姐的那个样子, 最后把自己捏成了一只大胖蛾子...    

![29-2.png](https://ae02.alicdn.com/kf/H7b28f717c739454fa401a48be4ff974as.jpg)    

噗~ 吃了那么多就变成了个蛾子, 感觉一点都不遵守能量守恒啊! 我还等着他变成大胖鹅呢!!! 额~ 串了, 串了, 那好像是"丑小鸭"的故事!

### 他自己的颜色    

有个小变色龙(chameleon)有点儿痛苦(distressed), 因为他不像其他的动物那样拥有自己"特有的"颜色.    

比如, 鹦鹉(parrot)的特有色是绿色的, 大象的特有色是灰色的, 但是无论他走到哪里, 他的颜色都会变成周围环境的颜色...    

直到有一天, 他遇到了另一只变色龙: "咱俩为啥不在一起呢? 虽然我们的颜色仍然会改变, 但是至少咱俩会一起改变."    

Why don't we stay together? We will still change color wherever we go, but you and I will always be alike.    

然后他们俩从此就过上了幸福快乐的生活(lived happily ever after).
