# 摘要:

Up in the garden, the world is full of green— leaves(叶片) and sprouts(菜苗), growing vegetables, ripening fruit.

But down in the dirt there is a busy world of earthworms digging, snakes hunting, skunks burrowing(挖掘), and all the other animals that make a garden their home.

In this exuberant book(热情洋溢的书), discover the wonder and activity that lie hidden between the stalks(茎秆), under the shade of leaves ... and down in the dirt. 

— Provided by publisher.

# 作者注解(note)

Every garden is a community garden. Do you know why? You may work hard planting seeds and pulling weeds(拔草), but plants can't thrive without the help of all those smaller gardeners(将泥土中的一些小昆虫比作园艺大师) down in the dirt.

Some help by controlling pests(控制害虫) that would like to gobble up(吞掉) your broccoli and tomatoes before you have a chance to eat them yourself. Some help by creating tunnels(创造隧道), which bring air into the soil and let water drain(排水) more easily. And believe it or not, some help just by going to the bathroom(上厕所) right there next to the green beans. This adds important nutrients(营养素) to the soil(就是小动物上个厕所,也能为土壤提供营养素).

A healthy organic garden isn't a garden without bugs— it's packed with lots and lots of living things, all working together with the human gardeners to nurture(培植) healthy plants.

# 相关动物(About the Animals)

The animals you met in this book really do live, eat, and work in many vegetable gardens. Some are helpful. Others ... not so much! 

1. You might think of **chickens** as farm animals, but more and more families are keeping them at home. These backyard pets do double duty(双重任务), laying eggs for their owners and eating bugs, kitchen scraps(残羹剩饭), and weeds. Chicken manure(粪肥) makes excellent compost.

1. **Pill bugs(西瓜虫)** aren't really insects at all but terrestrial(陆地生物), or land-based(栖息于陆地的), crustaceans(甲壳纲动物). Their scientific name is Armadillidium vulgare(卷甲虫), but most people call them pill bugs or roly-poly bugs(胖墩虫), nicknames that come from the creature's habit of rolling up into a ball when it feels threatened. Pill bugs eat decaying plants, which helps to add important nutrients to the soil.

1. **The tomato hornworm(天蛾)** is the colorful larva of an insect called the five-spotted hawk moth(五点天蛾), and it's a dreaded(可怕的) garden pest. Adult moths often lay their eggs on tomato leaves, where the larvae emerge to eat. When it's fully grown, a caterpillar burrows into the soil(钻到土里去) to pupate(化蛹), and an adult moth emerges about two weeks later to start the process over again.

1. **Honeybees** and bumblebees help to pollinate(对...授粉) plants in a garden. When they fly from flower to flower, collecting nectar and pollen(花蜜和花粉), they also transfer pollen from one flower to another, leading to fertilization(受精). Without an exchange of pollen, the blossoms on plants like cucumbers, squash, and watermelon would never develop into the fruits and vegetables we eat.

1. **Earthworms** are garden heroes(菜园英雄)! Their tunnels bring air into the soil and make it easier for water to drain(排水). Too much water is bad for plants because it can lead to mold(发霉) and mildew(霉病) and prevent roots from getting the oxygen they need. As earthworms burrow, they ingest soil(吞下土壤) and eat the decomposing roots(正在分解的根) and leaves. When the dirt comes out of the other end of the worm(一头吃土,一头拉营养物), as castings(铸件), it is finer and softer dirt, full of nutrients to help plants grow.

1. **Long-legged spiders** that we see in gardens are often Opiliones(盲蛛), or harvestmen(盲蛛,收割者), which are related to true spiders but differ in that they don't have two distinct body sections(明显的身体分区). Harvestmen don't have venom(毒液) or silk glands(丝腺) either. They eat decomposing plants as well as some pesky(讨厌的) garden insects like aphids and mites(蚜虫和螨虫).

1. **The robin**, or Turdus migratorius(旅鸫), is just one of many species of birds that might come to the garden to snack on bugs(吃点小虫点心). This is helpful because crickets, beetles, grubs(蛆), and slugs(蛞蝓(kuò yú)) can all damage plants. Some gardeners go out of their way to attract insecteating birds(吃昆虫的鸟) by providing them with small shrubs for cover and bird baths(鸟浴,供鸟戏水的水盆) for drinking and bathing.

1. **Bats** are also good friends to gardeners. Like birds, they eat many insects that damage plants as well as the mosquitoes(蚊子) that pester(使...烦扰) you while you're trying to pull weeds(拔草) at dusk. A single brown bat can devour(吞食) thousands of insects in a single night.

1. **June bugs(六月虫)**, or Phyllophaga, are so named because they're common in the months of May and June. The June bug is actually a double threat to gardens: the larval grubs can damage the roots of plants, while the adults devour leaves. Predators(捕食者) like toads and frogs help keep June bugs under control.

1. Like bats, **skunks(臭鼬)** are nighttime predators that gobble garden pests after dark. Skunks love grubs and slugs. However, they have smelly reputations(臭名昭著) and sometimes dig holes while they're searching for insects, so not all gardeners welcome them.

1. **Cutworms(切根虫)** aren't really worms; they're moth larvae(蛾子幼虫) that hide during the day and feast at night. Cutworms feed on young plants by eating right through the stems, near the base of the soil, cutting the plant down like a tree. Many gardeners dig up cutworms from the soil and remove them by hand. Predators like skunks help, too.

1. **The praying mantis(螳螂)** is a welcome garden guest because it has an appetite for(对...有胃口) aphids— notorious little bugs that suck the sap(植物的液汁) out of plants, weakening or killing them. The praying mantis also eats beetles and grasshoppers(草蜢) as it grows larger. In the fall, the female mantis mates and lays up to four hundred eggs in a tough brown capsule(胶囊/螵蛸(piāo xiāo)). If it survives the winter, there will be more hungry praying mantises in the spring! 

1. **Grasshoppers(蚱蜢)** may be fun to catch with a butterfly net(捕蝶网), but they're no friends to garden plants. They chow down(大快朵颐) on the leaves of most vegetables, but carrots, beans, and lettuce are among their favorites. Many gardeners hand pick grasshoppers from their plants and count on garden predators to take care of the rest.

1. **Garter snakes(袜带蛇)** are not venomous and won't hurt people, but they're great soldiers in the garden grasshopper war. They tend to live under fallen leaves or in wood piles(柴火堆) and don't disturb the garden at all.

1. Many **orb-weaver spiders(圆蛛)** build a new web each evening to capture insects that fly around gardens at night. Most of these webs are made of two kinds of silk— sticky capture threads(黏性捕捉网) that trap prey and nonsticky threads(非黏性网) that the spider uses to get around.

1. **Ants** may not be welcome at your harvest picnic(丰收野餐), but they can be beneficial in the garden. Ants often crawl(爬) from plant to plant looking for nectar(花蜜), so they can help pollinate plants. Like earthworms, they also help to bring air into the soil with their tunnels so plant roots can grow down more easily.
