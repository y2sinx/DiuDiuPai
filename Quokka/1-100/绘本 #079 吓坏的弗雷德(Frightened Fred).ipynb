### 吓坏的弗雷德 

佩塔·考普兰斯

![sm](https://sc04.alicdn.com/kf/Uec1ce2ba362d4b89adc8155f272e6fbbu.jpg)

---

Fred was often frightened(受到惊吓). 

![04-05](https://ae02.alicdn.com/kf/Uebbcb902376d494ebd1172d7a56a17aaL.jpg)

---

He couldn't help it(他总是控制不住自己). 

He came from a very nervous family(他来自一个精神兮兮的家庭). 

![06-07](https://ae03.alicdn.com/kf/U89423994d52f41e889ede17c28b5e26aI.jpg)

---

He was too scared to(太害怕以至于不能) help his sister check for monsters under the bed(检查床底下是否有怪兽). 

He was too scared to watch his brother's caterpillars wiggle(不敢看毛毛虫来回扭动). 

![08-09](https://ae04.alicdn.com/kf/Ub8d1d5bc6345420e99bbabb087f43967y.jpg)

---

When his cousin Jack came to play, Fred showed him his list of Scary Things(可怕物名单). 

Fred's mother said maybe they should take a nice, gentle walk to the park(去公园散个优雅的小步). 

![10-11](https://ae04.alicdn.com/kf/U16cfc489aa6d4e36a20a4f017c27e792l.jpg)

---

"I'm not scared," said Jack. "I can climb trees(爬树)." 

"So can I(倒装)," said Fred. 

![12-13](https://ae03.alicdn.com/kf/U94f410bc7c4b41b5b057957e940c9f1dC.jpg)

---

"I can jump across this river," said Jack. 

"Me too," said Fred. 

"I'm not scared of(害怕) rivers," said Jack. 

![14-15](https://ae03.alicdn.com/kf/U1515636cf75049d5a4fd84f1addeef81b.jpg)

---

"Only big rivers full of crocodiles(鳄鱼)." 

![16-17](https://ae02.alicdn.com/kf/U06ce832bed9f4dd6a0a8754dafdf91f8V.jpg)

---

"Spooky(令人感觉怕怕的), isn't it?" said Fred in the woods. 

"Not too spooky for me," said Jack. "I'm never scared." 

![18-19](https://ae04.alicdn.com/kf/Uecee861c38474597bc328b5f1d0e0866Z.jpg)

---

Some cats from Fred's street(来自弗雷德的街区的,定语/状语?) were coming up the hill. 

They were big. They were tough(强壮的,硬汉(tough guy)). They were getting closer. 

![20-21](https://ae02.alicdn.com/kf/U6a0ba27d5f2d4795876bc2046a741e2aX.jpg)

---

And closer... Jack turned and ran. 

Fred couldn't move(杰克转身跑了,但是弗雷德却没有). 

![22-23](https://ae04.alicdn.com/kf/U4198242d00da40329a41404fe6aa110fn.jpg)

---

"Is this the way to the river(这是去河边的路吗)?" asked the biggest cat. 

"Yes," said Fred. 

"Thanks," said the cats. "See you around(回见)." 

![24-25](https://ae03.alicdn.com/kf/U00f771c3dd1d41b8af61aed7b8b34d73M.jpg)

---

"Hey Fred," said Jack when he was sure they'd gone. "Those cats looked so mean(看起来有点刻薄) - weren't you scared?" 

"Scared?" said Fred. "WHO, ME?(用反问的语气,表示不怕)"
