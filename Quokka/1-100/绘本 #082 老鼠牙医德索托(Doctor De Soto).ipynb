### 老鼠牙医德索托    

威廉·斯蒂格    

![03](https://s.pc.qq.com/tousu/img/20210421/2796958_1619018892.jpg)    

Doctor De Soto(德索托), the dentist(牙医), did very good work, so he had no end of patients(数不尽的病人). Those close to his own size— moles(鼹鼠), chipmunks(花栗鼠), et cetera(等等)— sat in the regular dentist's chair.    

Larger animals sat on the floor, while Doctor De Soto stood on a ladder(梯子).    

![05](https://s.pc.qq.com/tousu/img/20210421/5578056_1619018904.jpg)    

For extra-large animals, he had a special room. There Doctor De Soto was hoisted up to the patient's mouth(被吊起来送进病人的嘴里) by his assistant, who also happened to be his wife.    

![06](https://s.pc.qq.com/tousu/img/20210421/7272377_1619018907.jpg)    

![07](https://s.pc.qq.com/tousu/img/20210421/9219747_1619018905.jpg)    

Doctor De Soto was especially popular with the big animals. He was able to work inside their mouths, wearing rubbers(橡胶) to keep his feet dry; and his fingers were so delicate(他的手指是如此的灵巧), and his drill so dainty(他的钻工是如此的精致), they could hardly feel any pain(病人几乎感觉不到一丁点儿的疼痛).    

![08](https://s.pc.qq.com/tousu/img/20210421/9162824_1619018900.jpg)    

Being a mouse, he refused to treat animals dangerous to mice(拒绝招待对老鼠们有危险的动物), and it said so on his sign. When the doorbell(门铃) rang, he and his wife would look out the window. They wouldn't admit even the most timid-looking cat(即使是最羞怯的猫猫,他们也不会招待).    

![09](https://s.pc.qq.com/tousu/img/20210421/4127773_1619018904.jpg)    

One day, when they looked out, they saw a well-dressed fox with a flannel bandage(法兰绒绷带) around his jaw(上下颌,下巴).    

![10](https://s.pc.qq.com/tousu/img/20210421/5853213_1619018903.jpg)    

![11](https://s.pc.qq.com/tousu/img/20210421/6840734_1619018913.jpg)    

"I cannot treat you(不能招待你), sir!" Doctor De Soto shouted. "Sir! Haven't you read my sign?"    

"Please!" the fox wailed(恸哭着). "Have mercy(求求你了), I'm suffering(我好痛啊)!" And he wept so bitterly it was pitiful to see(哭的满脸都是,看起来很是可怜).    

![12](https://s.pc.qq.com/tousu/img/20210421/7312568_1619018900.jpg)    

"Just a moment," said Doctor De Soto. "That poor fox," he whispered to his wife. "What shall we do?"    

"Let's risk it," said Mrs. De Soto. She pressed the buzzer(蜂鸣器) and let the fox in.    

![13](https://s.pc.qq.com/tousu/img/20210421/6422190_1619018905.jpg)    

He was up the stairs in a flash(立马就上了楼梯). "Bless your little hearts," he cried, falling to his knees. "I beg you, do something! My tooth is killing me(我的牙要疼死了)."    

![14](https://s.pc.qq.com/tousu/img/20210421/4934910_1619018903.jpg)    

"Sit on the floor, sir," said Doctor De Soto, "and remove the bandage, please."    

![15](https://s.pc.qq.com/tousu/img/20210421/5549591_1619018912.jpg)    

Doctor De Soto climbed up the ladder and bravely entered the fox's mouth. "Ooo-wow!" he gasped(喘着气). The fox had a rotten bicuspid(一颗烂的二尖齿) and unusually bad breath(口臭).    

"This tooth will have to come out," Doctor De Soto announced. "But we can make you a new one."    

![16](https://s.pc.qq.com/tousu/img/20210421/2987386_1619018898.jpg)    

"Just stop the pain," whimpered the fox, wiping some tears away(一边擦着他的眼泪,一边啜泣着说).    

Despite his misery(尽管他很疼痛), he realized he had a tasty little morsel(一口好吃的老鼠肉肉) in his mouth, and his jaw began to quiver(颤动->开始吧唧起嘴来).    

"Keep open!" yelled Doctor De Soto. "Wide open!" yelled his wife.    

![17](https://s.pc.qq.com/tousu/img/20210421/1981570_1619018898.jpg)    

"I'm giving you gas(麻醉剂) now," said Doctor De Soto. "You won't feel a thing when I yank that tooth(当我给你拔牙的时候,你一点儿都不会疼的).    

"Soon the fox was in dreamland(梦境). "M-m-m, yummy(真好吃)," he mumbled(含糊的说着). "How I love them raw(生的,新鲜的)... with just a pinch of salt(撒上一撮盐), and a... dry... white wine(再喝点儿白葡萄酒)."    

![18](https://s.pc.qq.com/tousu/img/20210421/4774029_1619018901.jpg)    

They could guess what he was dreaming about. Mrs. De Soto handed her husband a pole(撑牙杆) to keep the fox's mouth open.    

![19](https://s.pc.qq.com/tousu/img/20210421/4696754_1619018895.jpg)    

Doctor De Soto fastened his extractor(收紧他的拔牙器) to the bad tooth. Then he and his wife began turning the winch(转动绞盘). Finally, with a sucking sound(吮吸声), the tooth popped out(砰的一下蹦了出来) and hung swaying in the air(在空中悬挂着摇来摇去).    

"I'm bleeding!" the fox yelped(因疼痛而大声尖叫着) when he came to(清醒过来).    

Doctor De Soto ran up the ladder and stuffed some gauze in the hole(往洞里塞了些纱布). "The worst is over," he said. "I'll have your new tooth ready tomorrow. Be here at eleven sharp(十一点整).    

![20](https://s.pc.qq.com/tousu/img/20210421/6009348_1619018893.jpg)    

"The fox, still woozy(晕晕乎乎的), said goodbye and left. On his way home, he wondered if it would be shabby of him to eat the De Sotos(寻思着"把医生吃掉的话"是不是有点卑鄙) when the job was done.    

![21](https://s.pc.qq.com/tousu/img/20210421/8149721_1619018901.jpg)    

After office hours, Mrs. De Soto molded a tooth of pure gold(浇铸了一颗纯金的牙齿) and polished it(并把它抛了光). "Raw with salt, indeed(竟然想用盐生腌我)," muttered Doctor De Soto. "How foolish to trust a fox!"    

"He didn't know what he was saying," said Mrs. De Soto. "Why should he harm us? We're helping him."    

"Because he's a fox!" said Doctor De Soto. "They're wicked, wicked creatures(邪恶的,不道德的生物)."    

![22](https://s.pc.qq.com/tousu/img/20210421/2817820_1619018897.jpg)    

That night the De Sotos lay awake worrying. "Should we let him in tomorrow?" Mrs. De Soto wondered.    

"Once I start a job," said the dentist firmly, "I finish it(要干就要干到底). My father was the same way."    

"But we must do something to protect ourselves," said his wife. They talked and talked until they formed a plan(想出了一个主意). "I think it will work," said Doctor De Soto. A minute later he was snoring(打呼噜了).    

![23](https://s.pc.qq.com/tousu/img/20210421/4351870_1619018901.jpg)    

The next morning, promptly at eleven(刚到十一点), a very cheerful fox turned up. He was feeling not a particle of pain(感觉一点都不疼了).    

![24](https://s.pc.qq.com/tousu/img/20210421/5751051_1619018962.jpg)    

When Doctor De Soto got into his mouth, he snapped it shut(啪的一下把嘴巴闭了起来) for a moment, then opened wide and laughed. "Just a joke!" he chortled(咯咯地笑着).    

"Be serious," said the dentist sharply. "We have work to do." His wife was lugging(使劲拖着) the heavy tooth up the ladder.    

"Oh, I love it!" exclaimed(说) the fox. "It's just beautiful.    

"Doctor De Soto set the gold tooth in its socket(牙槽) and hooked it up(把它钩在...的上面) to the teeth on both sides.    

![25](https://s.pc.qq.com/tousu/img/20210421/1144103_1619018964.jpg)    

The fox caressed the new tooth with his tongue(用舌头触摸着他的新牙). "My, it feels good," he thought. "I really shouldn't eat them. On the other hand, how can I resist?"    

![26](https://s.pc.qq.com/tousu/img/20210421/8755342_1619018965.jpg)    

"We're not finished," said Doctor De Soto, holding up a large jug. "I have here a remarkable preparation developed only recently by my wife and me. With just one application(只需一次服用), you can be rid of toothaches forever(便可永久摆脱牙痛). How would you like to be the first one to receive this unique treatment?"    

![27](https://s.pc.qq.com/tousu/img/20210421/7176640_1619018961.jpg)    

"I certainly would!" the fox declared(说). "I'd be honored." He hated any kind of personal pain."    

You will never have to see us again," said Doctor De Soto.    

"No one will see you again," said the fox to himself. He had definitely made up his mind to(下定决心去) eat them— with the help of his brand-new tooth(崭新的牙齿).    

![28](https://s.pc.qq.com/tousu/img/20210421/8840179_1619018964.jpg)    

Doctor De Soto stepped into the fox's mouth with a bucket of secret formula(一桶秘密配方) and proceeded to paint each tooth(一个接一个地涂到了每一颗牙齿上). He hummed(哼着小曲儿) as he worked. Mrs. De Soto stood by on the ladder, pointing out spots he had missed. The fox looked very happy.    

![29](https://s.pc.qq.com/tousu/img/20210421/8529269_1619018964.jpg)    

When the dentist was done, he stepped out. "Now close your jaws tight," he said, "and keep them closed for a full minute."    

The fox did as he was told. Then he tried to open his mouth— but his teeth were stuck together(粘在了一起)!"    

Ah, excuse me, I should have mentioned," said Doctor De Soto, "you won't be able to open your mouth for a day or two. The secret formula must first permeate the dentine(这种秘密配方必须首先渗进牙质). But don't worry. No pain ever again!"    

![30](https://s.pc.qq.com/tousu/img/20210421/7929326_1619018960.jpg)    

The fox was stunned(目瞪口呆的). He stared at Doctor De Soto, then at his wife. They smiled, and waited. All he could do was say, "Frank oo berry mush(灰常感蟹)" through his clenched teeth(被紧紧粘住的牙齿), and get up and leave. He tried to do so with dignity(留有尊严地).    

![31](https://s.pc.qq.com/tousu/img/20210421/3779031_1619018970.jpg)    

Then he stumbled down the stairs(跌跌撞撞地走下了楼梯) in a daze(晕晕乎乎地).    

![32](https://s.pc.qq.com/tousu/img/20210421/4233087_1619018961.jpg)    

Doctor De Soto and his assistant had outfoxed(比...还狐狸(机智)) the fox(用其机智战胜了狐狸). They kissed each other and took the rest of the day off.
