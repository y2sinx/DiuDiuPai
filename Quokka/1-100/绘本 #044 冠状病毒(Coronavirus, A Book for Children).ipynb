### 什么是冠状病毒?    

Viruses(病毒) are tiny germs(细菌) that are so small that you can't see them.    

![4-1.png](https://s.pc.qq.com/tousu/img/20210206/8016250_1612598381.jpg)    

They are so light that they can float through the air(在空气中流动) in tiny drops of water(微小的水滴), and they can sit on your skin(粘在你的皮肤上) without you feeling them.    

![4-2.png](https://s.pc.qq.com/tousu/img/20210206/7759884_1612598380.jpg)    

infect: /ɪnˈfekt/, 感染.    

a snotty nose: 一个流鼻涕的鼻子.    

![4-3.png](https://s.pc.qq.com/tousu/img/20210206/8292274_1612598380.jpg)    

### 人们是如何感染冠状病毒的?    

Coronavirus germs live in people's throats(喉咙) and mouths.    

![5-1.png](https://s.pc.qq.com/tousu/img/20210206/3012357_1612598382.jpg)    

a cloud of steam: (天冷时呼出的)气团.    

![5-2.png](https://s.pc.qq.com/tousu/img/20210206/7384746_1612598374.jpg)    

If the person with the coronavirus germs on their hands uses a door, the invisible(不可见的) germs can live on the handle(门把手) for hours.    

![5-3.png](https://s.pc.qq.com/tousu/img/20210206/6045576_1612598375.jpg)    

不要轻易用手摸自己的鼻子或嘴巴.    

![5-4.png](https://s.pc.qq.com/tousu/img/20210206/4723357_1612598376.jpg)    

### 如果感染了冠状病毒会发生什么?    

Some people, particularly most children, hardly feel ill at all(有些人, 尤其是孩子, 在感染冠状病毒之后可能并不会生病) when they get the coronavirus.    

![6-1.png](https://s.pc.qq.com/tousu/img/20210206/9567865_1612598372.jpg)    

coughs(咳嗽), high temperature(高温), headache(头痛), aches(疼痛).    

![6-2.png](https://s.pc.qq.com/tousu/img/20210206/1910266_1612598378.jpg)    

Tiny cells(细胞) in your blood(血液) make antibodies(抗体) to fight each different virus invader(病毒侵略者).    

![6-3.png](https://s.pc.qq.com/tousu/img/20210206/1890255_1612598375.jpg)    

### 人们为什么会害怕感染冠状病毒?    

老年人或身患疾病的人很难对抗病毒, 因此他们需要额外的小心(extra-careful).    

![7-1.png](https://s.pc.qq.com/tousu/img/20210206/7610198_1612598373.jpg)    

ventilator: /ˈventɪleɪtə(r)/, 呼吸器.    

![7-2.png](https://s.pc.qq.com/tousu/img/20210206/8355831_1612598378.jpg)    

away from anywhere they might catch the coronavirus: 远离任何他们可能会感染冠状病毒的地方.    

![7-3.png](https://s.pc.qq.com/tousu/img/20210206/3661756_1612598376.jpg)    

不要去拜访别人(to go and visit sb), 也不要接待拜访者(visitors).    

![7-4.png](https://s.pc.qq.com/tousu/img/20210206/9679666_1612598383.jpg)
