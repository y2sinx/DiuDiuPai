# 哞哞叫的棕牛    

贾基⋅伍德(作) | 罗格⋅邦纳(画)    

---

![02_.jpg](https://ae04.alicdn.com/kf/H6992bcce3a874025bc3e17750740d0366.jpg)    

哞哞, 棕色牛, 你有任何小牛吗?    
有的, 小猫猫, 我有的, 我有一只带斑点的小牛(calf).    

---

![03_.jpg](https://ae04.alicdn.com/kf/He03750c3a96f4d8ca1f5534522e17e41m.jpg)    

咩咩, 黑绵羊, 你有任何小羊吗?    
有的, 小猫猫, 我有的, 我有两只毛茸茸的小羊(lamb).    

---

![04_.jpg](https://ae04.alicdn.com/kf/Hc2324b32bfa84cff8027d0abd1b6b321J.jpg)    

咩咩, 黄山羊, 你有任何孩子吗?    
有的, 小猫猫, 我有的, 我有三个睡意正浓的孩子.    

---

![05_.jpg](https://sc01.alicdn.com/kf/He121614780ee45ae8f0b378164b8209f2.jpg)    

嘎嘎, 白鸭子, 你有任何小鸭子吗?    
有的, 小猫猫, 我有的, 我有四只松蓬蓬的小鸭子(duckling).    

---

![06_.jpg](https://ae04.alicdn.com/kf/Haa7d8eecbe0d467bbb7898e1df6ac46dT.jpg)    

嘎嘎, 大蓝鹅, 你有任何小鹅吗?    
有的, 小猫猫, 我有的, 我有五只小胖鹅(gosling).    

---

![07_.jpg](https://ae02.alicdn.com/kf/Hfc7efaf82bc34110bcb2c1ed55e18106i.jpg)    

咯咯, 橘色鸡, 你有任何的小鸡吗?    
有的, 小猫猫, 我有的, 我有六只吵闹的小鸡(chick).    

---

![08_.jpg](https://sc03.alicdn.com/kf/Ha21374fbc58c476ea954dc1784adb275H.jpg)    

哼哼, 粉色猪, 你有任何的小猪吗?    
有的, 小猫猫, 我有的, 我有七只饥饿的小猪(piglet).    

---

![09_.jpg](https://ae03.alicdn.com/kf/Hbef19d1df58846f1bd5390ce94d0978cb.jpg)    

呱呱, 绿色蛙, 你有任何的小青蛙吗?    
有的, 小猫猫, 我有的, 我有八只正在蠕动的小青蛙(froglet).    

---

![10_.jpg](https://ae04.alicdn.com/kf/H857b25155e5d4c97af63c8c4c45e9a1da.jpg)    

汪汪, 红色狗, 你有任何的小狗吗?    
有的, 小猫猫, 我有的, 我有九只乱叫的小狗(pup).    


---

![11_.jpg](https://sc01.alicdn.com/kf/H21fc69b37b8d457dbbd002825fa85390u.jpg)    

咕噜咕噜, 虹鳟鱼, 你有任何的小鱼苗(fry)吗?    
有的, 小猫猫, 我有的, 你能把它们十个全都数过来吗?    

---

![12_.jpg](https://sc01.alicdn.com/kf/H265e0130069c455bbc3b57d190d12b7e8.jpg)    

喵喵, 小猫猫, 你有任何小猫咪(kitten)吗?    

---

![13_.jpg](https://sc03.alicdn.com/kf/H3edcbbc6e5c5485cb77e76052e8bc301t.jpg)    

没有小猫咪, 我没有——    
但是我有很多很多的朋友.
