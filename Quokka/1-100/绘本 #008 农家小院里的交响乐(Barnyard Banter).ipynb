### 中文大意  

牛儿在牧场(pasture)上,哞(mōu)哞哞~  
公鸡在院子里,喔(wō)喔喔~  
母鸡在鸡舍里,咯(gē])咯咯~  
猪猪在泥坑(wallow)里,哼哼哼~  
但是大鹅跑到哪里去了呢?  

小猫在草棚(hayloft)里,喵喵喵~  
鸽子在屋梁(rafter)上,咕咕咕~  
老鼠在粮箱里,吱吱吱~  
孔雀在钢丝柱上,哇(wā)哇哇~  
但是大鹅跑到哪里去了呢?  

驴子在围栏(paddock)里,昂昂昂~  
乌鸦在玉米地里,呀呀呀~  
蟋蟀在石缝里,啾(jiū)啾啾~  
青蛙在池塘里,呱(guā)呱呱~  
但是大鹅跑到哪里去了呢?  

此处有一曲无声的农场交响乐...  

但是大鹅跑到哪里去了呢?  
啊啊啊,原来大鹅在这里啊——  
嘎嘎嘎~
