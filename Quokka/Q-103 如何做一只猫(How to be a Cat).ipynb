#### 如何做一只猫猫    

丽莎·斯威林 | 拉尔夫·拉扎尔    

![sm](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/3077dcc4-8e0c-474b-9697-8b0a75249f2c.png)    

Just be adorable.    

![03](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/3504e33d-d416-4392-aa1b-ec1beb8a44be.png)    

Refuse cuddles(拥抱) at a reasonable time. Demand them at 7 a.m.    

![04](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/f68796a1-b649-43b2-9f26-d67830f30862.png)    

Consider it important to throw up(呕吐) at 4 a.m. at least twice a month. Preferably on the carpet.    

![07](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/256a5b94-a816-4a2e-9581-0373bb41cbcc.png)    

Never let your human pay more attention to their book than to you.    

![09](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/b967551d-3224-4109-872e-d6a5d4208db7.png)    

Jump on anything that moves on the bed.    

![11](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/fe1cd9dd-ae4b-4731-b507-f426b30f3e57.png)    

Dash for the door as soon as it gets opened(一旦门被打开,立马冲向门口).    

![14-15](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/bbb99ae9-54c4-4859-b94e-1d043893e4f7.png)    

Sleep anywhere.    

![16](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/975dedaf-db12-4aa1-9a2e-ec2c310b982c.png)    

A daily yoga(瑜伽) session is very important.    

![17](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/dc01f8ad-e6e2-457d-87b8-9e4ae18245f8.png)    

Stand and meow until your human figures out what you want. This can be hours of fun!    

![20](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/218e7de6-8f31-4d1f-8cfb-e7d2b4f9f030.png)    

Be curious.    

![21](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/9222e924-ddea-4d79-85be-868e9459a3a6.png)    

Beg for belly rubs(求着摸肚子), then attack your human after exactly three rubs.    

![22](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/d7926027-278f-4342-a493-10ad42bbb371.png)    

Make yourself comfortable on your human's lap(大腿) and remain there even after their legs go to sleep.    

![24](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/f00ceafc-9d2b-4098-bdf0-a4f7ec8837ba.png)    

Sleep, sleep, and then sleep some more. The more you sleep, the catter you are(你睡得越多,就越像猫).    

![25](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/f63c8d00-4ed2-48ff-81f1-cd5e12b151a6.png)    

When walking down the stairs in front of your human, stop suddenly to maximize their chance of tripping(最大化他们被绊倒的几率).    

![28-29](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/9146e366-b43c-4994-98ee-f23371b43eea.png)    

Plants must be tasted(一定要尝尝"花草"的味道).    

![31](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/25a5d323-49f0-4991-8f44-6d6d4b33ed90.png)    

When your photo is being taken, under no circumstances stand still(绝对不要站着不动).    

![33](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/f289492d-7228-449c-a4fc-ac8ca39c950c.png)    

Assume every tin being opened is full of tuna, just for you(所有打开的金枪鱼罐头都是你的).    

![34](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/b7a3af95-b69a-4575-bee2-b6a09681f28d.png)    

Plead for food. Once you have it, forget it.    

![37](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/989b8581-9929-4607-864e-f72287375264.png)    

From up high, watch your human frantically(当你的主人疯狂找你的时候) look for you. Say nothing(静静看着,不要让她发现).    

![38](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/a2ea3752-f6a8-4d84-aacf-0bf80097b0ff.png)    

When your human puts new sheets on the bed, add your own paw print design. Preferably muddy.    

![40-41](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/a6ba9fbe-98a0-4f48-9ec9-c2e61b119c00.png)    

Be desperate to go outside, but when the door is finally opened, refuse.    

![44](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/28a104ac-3f88-4f0d-a748-463dc4eca340.png)    

Ignore the obvious scratching post(忽略磨爪柱). Claw expensive furniture instead(专挑贵家具).    

![50](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/bf9e4ea3-5ccb-43c0-bd8c-e74e398beb73.png)    

Randomly surprise attack your human when they're gardening.    

![51](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/ff64fba2-097a-4d06-b8ec-d67a0d978575.png)    

Refuse to be picked up when your human tries to show guests how loving and cuddly you are.    

![52](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/c6b23d6a-8e22-4b45-8016-77b550dd134e.png)    

Find box, Stalk box(追踪箱子), Squeeze into box, Destroy box, Wait for next box.    

![60-61](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/85918705-f992-41db-b8bc-0335f10a8db7.png)    

When called, go in the opposite direction.    

![70](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/24b165e5-6f4c-4171-bc7c-4580701263d5.png)    

Must get laser dot(激光点). Must be destroyed.    

![73](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/2ea755b2-639e-4c49-84a4-4fae4bc4913e.png)    

Keep new visitors under CONSTANT surveillance(持续监视).    

![74](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/6eae5c68-52fc-4c00-8a91-3e237c62fd40.png)    

Sit and gaze at your food bowl as though(仿佛) you have not been fed for days (even if it is full).    

![77](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/4ef9fc9f-3e46-47b2-adea-32efa5f92d12.png)    

As soon as someone gets up, grab their place and act as if you've been there for ages.    

![80-81](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/af98b05c-e6e7-4ac0-bc87-a1774d628cf8.png)    

Take center stage at mealtimes.    

![82](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/88da0c0d-34e8-435a-b405-72edf9e4eaf4.png)    

Seek out the warmest lap.    

![83](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/85a016cf-fbb3-4ef0-85e9-425f55077fab.png)    

Observe.    

![86-87](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/40ea509e-1f76-4c06-8cdf-5d2336a6959a.png)    

Roll over as if you want your tummy scratched, but then scratch the human who attempts to do so.    

![89](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/84060412-59ab-4a26-907a-9eab24d88796.png)    

Wait for clothes to be perfectly ironed(烫平), then it's very important to lie on them.    

![90](https://zebra-common-system-pre.oss-cn-shanghai.aliyuncs.com/bmdev/506716b7-b95b-422e-b4e9-11812b85bfac.png)    

Rest whenever possible, you deserve it.
